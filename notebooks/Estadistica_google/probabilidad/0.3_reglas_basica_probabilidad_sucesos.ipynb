### Las reglas básicas de la probabilidad y los sucesos

Hasta ahora, nos hemos centrado en calcular la probabilidad de eventos individuales. Muchas situaciones, tanto en la vida cotidiana como en el análisis de datos, implican más de un evento. 

Como futuro profesional de datos, con frecuencia tendrá que lidiar con la probabilidad de varios eventos. 

- la regla del complemento, 
- la regla de la suma y 
- la regla de la multiplicación. 

Estas reglas ayudan a comprender mejor la probabilidad de que se produzcan varios eventos. 

tipos diferentes de eventos: 
- eventos que se excluyen mutuamente 
- eventos independientes.


### Notacion de la probabilidad 

Primero, analicemos la notación de probabilidad, que es la forma estándar de simbolizar los conceptos de probabilidad.

- La letra P indica la probabilidad de un evento. Por ejemplo, si se trata de dos eventos, puede etiquetar un evento como A y el otro como B. La notación para la probabilidad del evento A es la letra P seguida de la letra A entre paréntesis.

- P(A) = Probability of event A
- P(B) = Probability of event B
- P(A')= Probability of not event A

Para la probabilidad del evento B, es la letra P, seguida de la letra B entre paréntesis. Si quieres hablar de la probabilidad de que el evento A no ocurra, agrega un apóstrofe después de la letra A. También puedes decir que esta es la probabilidad de que no ocurra A.

### Regla del complemento

Ahora, veamos nuestra primera regla básica, la regla del complemento. En las estadísticas, el complemento de un evento es el evento que no ocurre. 

Por ejemplo, o llueve o no llueve. O ganas la lotería o no la ganas. El complemento de la lluvia es la ausencia de lluvia. El complemento de ganar no es ganar.

Lo importante a tener en cuenta es que las dos probabilidades, la probabilidad de que ocurra un evento y la probabilidad de que no ocurra, deben sumarse a una. 

Recuerde que la probabilidad de uno es lo mismo que decir que hay una certeza del 100 por ciento de que ocurra un evento. Otra forma de verlo es que hay un 100 por ciento de probabilidades de que ocurra un evento u otro. Puede haber un 30 por ciento de probabilidades de que llueva mañana, pero hay un 100 por ciento de probabilidades de que mañana llueva o no llueva. 

La regla del complemento dice que la probabilidad de que el evento A no ocurra es de 1 menos la probabilidad del evento A. 

- P(A') = 1 - P(A)

Por ejemplo, si el pronóstico del tiempo dice que hay un 30 por ciento de probabilidades de que llueva mañana, hay una probabilidad de 0.3. Puedes usar la regla del complemento para calcular la probabilidad de que no llueva mañana. 

La probabilidad de que no llueva es igual a 1 menos la probabilidad de lluvia. Esto es 1 menos 0.3 igual a 0.7 o 70 por ciento. 

- P(no rain) = 1 - P(rain) = 1 - 0.3 = 0.7


### Eventos excluyentes mutuamente

Sin embargo, la regla del complemento y nuestra siguiente regla, la regla de la suma, aplicaron dos eventos que se excluyen mutuamente. 

- Dos eventos se excluyen mutuamente si no pueden ocurrir al mismo tiempo. 

Por ejemplo, no puedes visitar Argentina y China al mismo tiempo, ni girar a la izquierda y a la derecha al mismo tiempo. 

La regla de la suma dice que si los eventos A y B se excluyen mutuamente, entonces la probabilidad de que ocurra A o B es la suma de las probabilidades de A y B. 

Addition rule (for mutually exclusive events)

P(A or B) = P(A) + P(B)

Veamos un ejemplo con un dado de seis caras. Supongamos que quieres averiguar la probabilidad de sacar un dos o un cuatro en una sola tirada del dado. Estos dos eventos se excluyen mutuamente. Puedes sacar un dos o un cuatro, pero no los dos al mismo tiempo. La regla de la suma dice que para encontrar la probabilidad de que ocurra cualquiera de los eventos, debes sumar sus probabilidades. 

P(rolling a 2 or rolling a 4) = P(rolling a 2) + P(rolling a 4) = 1/6 + 1/6 = 1/3

Las probabilidades de sacar cualquier número en un dado de seis caras son 1/6. La probabilidad de sacar un dos es 1/6, y la probabilidad de sacar un cuatro es 1/6. 1/6 más 1/6 es igual a 1/3. La probabilidad de sacar un dos o un cuatro es de 1/3 o 33 por ciento. 


La regla de adición se aplica a los eventos que se excluyen mutuamente. 


### Independent events 

Si quieres calcular la probabilidad de eventos independientes, puedes usar la regla de multiplicación. 

- Dos eventos son independientes si la ocurrencia de un evento no cambia la probabilidad del otro evento. Esto significa que un evento no afecta al resultado del otro evento.

Por ejemplo, sacar un libro de la biblioteca local no afecta el clima de mañana. Beber café por la mañana no afecta la entrega de su correo por la tarde. Estos eventos son separados e independientes.

La regla de la multiplicación dice que si los eventos A y B son independientes, entonces la probabilidad de que ocurran tanto A como B es la probabilidad de A multiplicada por la probabilidad de B. 

Multiplication rule (for independet events)

P(A and B) = P(A) * P(B)

P(1st toss tails and 2nd toss heads) = P(1st toss tails) * P(2nd toss heads) 
                            P(tails and heads) = (0.5) * (0.5) = 0.25

Por ejemplo, imagine dos lanzamientos consecutivos. Supongamos que quieres saber la probabilidad de que salgan las colas en el primer lanzamiento y las cabezas en el segundo. 

Primero, averigua a qué eventos te enfrentas y, a continuación, aplica la regla correspondiente. Los lanzamientos de dos monedas son eventos independientes.

El primer lanzamiento no afecta al resultado del segundo lanzamiento. Para cualquier lanzamiento, la probabilidad de obtener cara o cruz siempre permanece entre el 1/2 o el 50 por ciento.

Usarías la regla de multiplicación para este evento. La probabilidad de conseguir cruz y cara es la probabilidad de conseguir cruz, multiplícala por la probabilidad de conseguir cara. La probabilidad de cada evento es de 0,5 o 50 por ciento.

Ahora, introduce los números, 0,5 veces 0,5 es igual a 0,25 o 25 por ciento. La probabilidad de sacar cruz en el primer lanzamiento y cara en el segundo es del 25 por ciento. 



Para resumir, comparemos las reglas de suma y multiplicación y enumeremos sus diferencias.

Será útil tener en cuenta estas diferencias para saber cuándo usar las dos reglas.

- La regla de la suma suma las probabilidades de los eventos y la regla de la multiplicación las multiplica.

- La regla de adición se aplica a los eventos que se excluyen mutuamente.
- La regla de la multiplicación se aplica a los eventos que son independientes.
- Las reglas básicas de probabilidad ayudan a describir eventos que son mutuamente excluyentes o independientes.
