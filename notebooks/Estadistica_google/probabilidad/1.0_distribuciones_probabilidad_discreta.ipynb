### Distribuciones de probabilidad discretas

Hace poco, aprendi que los profesionales de datos utilizan distribuciones de probabilidad para modelar diferentes tipos de conjuntos de datos y para identificar patrones significativos en sus datos. Recuerde que una distribucin de probabilidad describe la probabilidad de los posibles resultados de un suceso aleatorio. Las distribuciones de probabilidad discretas representan variables aleatorias discretas o sucesos discretos. A menudo, los resultados de los sucesos discretos se expresan como nmeros enteros que se pueden contar. Por ejemplo, lanzar un dado puede dar como resultado un 2 o un 3, pero no un valor decimal como 2,575 o 3,184.

En esta lectura, obtendr una visin general de los principales atributos de cuatro distribuciones de probabilidad discreta comunes:

### Distribuciones de probabilidad discreta

**Distribucin uniforme**

La distribucin uniforme describe sucesos cuyos resultados son igualmente probables.

Por ejemplo, lanzar un dado puede dar seis resultados: 1, 2, 3, 4, 5  6. La probabilidad de cada resultado es la misma. La probabilidad de cada resultado es la misma: 1 de 6, es decir, aproximadamente el 16,7%.

Puedes visualizar una distribucin con un grfico, como un histograma. Para una distribucin discreta, la variable aleatoria se traza a lo largo del eje x, y la probabilidad correspondiente se traza a lo largo del eje y. En este caso, el eje x representa cada resultado posible de una tirada de dado y el eje y representa la probabilidad de cada resultado.



**Nota:** Los profesionales de los datos suelen utilizar la distribucin uniforme como parte de mtodos estadsticos ms complejos, como las simulaciones de Montecarlo. Una discusin detallada de estos mtodos est fuera del alcance de este curso.

Nota: La distribucin uniforme se aplica tanto a variables aleatorias discretas como continuas.


**Distribucin binomial**

La Distribucin binomial modela la probabilidad de sucesos con slo dos resultados posibles: xito o fracaso. Estos resultados son mutuamente excluyentes y no pueden ocurrir al mismo tiempo.

Esta definicin supone lo siguiente:

- Cada suceso es independiente, es decir, no afecta a la probabilidad de los dems.

- Cada suceso tiene la misma probabilidad de xito.

Recuerde que xito y fracaso son etiquetas que se utilizan por comodidad. Por ejemplo, si lanzas una moneda, slo hay dos resultados posibles: cara o cruz. En funcin de las necesidades de su anlisis, puede elegir entre cara o cruz como resultado exitoso.

La Distribucin binomial representa un tipo de suceso aleatorio llamado experimento binomial. Un experimento binomial tiene los siguientes atributos:

- El experimento consiste en un nmero de ensayos repetidos.

- Cada ensayo tiene slo dos resultados posibles.

- La probabilidad de xito es la misma en cada ensayo.

- Y cada ensayo es independiente.

Un ejemplo de experimento binomial es lanzar una moneda 10 veces seguidas. Se trata de un experimento binomial porque tiene las siguientes caractersticas:

- El experimento consiste en 10 ensayos repetidos, o lanzamientos de monedas.

- Cada prueba tiene slo dos resultados posibles: cara o cruz.

- Cada prueba tiene la misma probabilidad de xito. Si se define el xito como cara, la probabilidad de xito de cada lanzamiento es la misma: 50%.

- Cada prueba es independiente. El resultado de un lanzamiento no afecta al resultado de otro lanzamiento.

**Distribucin de Bernoulli**

La distribucin Bernoulli es similar a la distribucin binomial, ya que tambin modela sucesos que slo tienen dos resultados posibles (xito o fracaso). La nica diferencia es que la distribucin de Bernoulli se refiere a un nico ensayo de un experimento, mientras que la binomial se refiere a ensayos repetidos. Un ejemplo clsico de ensayo Bernoulli es el lanzamiento de una moneda.

**Distribucin de Poisson**

La Distribucin de Poisson modela la probabilidad de que se produzca un determinado nmero de sucesos durante un periodo de tiempo especfico.

**Nota:** _La Distribucin de Poisson tambin se puede utilizar para representar el nmero de sucesos que ocurren en un espacio especfico, como una distancia, un rea o un volumen. En este curso, nos centraremos en el tiempo._

La Distribucin de Poisson representa un tipo de experimento aleatorio llamado experimento de Poisson. Un experimento de Poisson tiene los siguientes atributos:

- El nmero de sucesos del experimento se puede contar.

- Se conoce el nmero medio de sucesos que ocurren durante un periodo de tiempo especfico.

- Cada suceso es independiente.

Por ejemplo, imagine que tiene una pgina web en la que publica contenidos. Su sitio web tiene una media de dos visitas por hora. Quiere determinar la probabilidad de que su sitio web reciba un cierto nmero de visitas en una hora determinada.

Se trata de un experimento de Poisson porque

- Se puede contar el nmero de eventos del experimento. Se puede contar el nmero de visitas.

- Se conoce el nmero medio de eventos que se producen durante un periodo de tiempo determinado. Hay una media de dos visualizaciones por hora.

- Cada resultado es independiente. La probabilidad de que una persona vea tu pgina web no afecta a la probabilidad de que otra persona la vea.
