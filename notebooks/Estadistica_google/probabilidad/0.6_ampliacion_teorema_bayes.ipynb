El teorema es el siguiente. La probabilidad del evento A dado el evento B es igual a la probabilidad de B dado A, multiplicada por la probabilidad de A dividida por la probabilidad de B dado A, por la probabilidad de A ms la probabilidad de B dado A, multiplicada por la probabilidad de no A. 

Puedes usar las dos versiones del teorema de Bayes para tratar diferentes tipos de problemas. 

A veces, por ejemplo, no se conoce la probabilidad del evento B, que est en el denominador de la ecuacin del teorema bsico de Bayes. En ese caso, puedes usar la versin expandida del teorema de Bayes, porque no necesitas conocer la probabilidad del evento B para usar la versin expandida.

Al evaluar la precisin de una prueba, el teorema de Bayes puede tener en cuenta la probabilidad de errores de prueba conocidos como falsos positivos y falsos negativos. 

**Un falso positivo es** el resultado de una prueba que indica que algo est presente cuando en realidad no lo est. Por ejemplo, un filtro de spam puede identificar incorrectamente un correo electrnico legtimo como spam. 

Los falsos positivos, con frecuencia se refieren a pruebas mdicas, pero tambin se aplican a otras reas, como las pruebas de software. Por ejemplo, el software antivirus puede indicar que un archivo (informtico) es un virus , aunque el archivo sea normal. 

**Un falso negativo** es el resultado de una prueba que indica que algo no est presente cuando realmente lo est.

Por ejemplo, un filtro de spam puede identificar incorrectamente un correo no deseado como legtimo. Los falsos negativos tambin se aplican a todo tipo de pruebas. En la fabricacin, por ejemplo, una prueba de control de calidad puede identificar incorrectamente una pieza defectuosa como una pieza aceptable.
