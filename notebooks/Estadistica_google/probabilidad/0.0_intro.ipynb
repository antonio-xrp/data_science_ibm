### Probability

- La probabilidad es la rama de las matemáticas que se ocupa de medir y cuantificar la incertidumbre. En otras palabras, la probabilidad usa las matemáticas para describir la probabilidad de que algo suceda.

Por ejemplo, la posibilidad de que llueva mañana o de ganar la lotería.

- Los profesionales de datos utilizan la probabilidad para ayudar a los líderes empresariales a tomar decisiones basadas en datos en situaciones de incertidumbre. 

Nadie puede saber el resultado de los acontecimientos futuros con total certeza. 
Lo que pueden hacer los profesionales de datos es utilizar todos los datos disponibles para hacer predicciones razonables basadas en la probabilidad.

Por ejemplo, imagine que trabaja con la parte interesada de una gran empresa aeroespacial. Tienen que decidir si invierten o no en una nueva tecnología para mejorar el proceso de producción de sus motores a reacción. Como profesional de datos, puede estimar la probabilidad de que la nueva tecnología tenga un impacto positivo y predecir cuáles podrían ser sus posibles costos y beneficios. La parte interesada puede usar esta información para tomar una decisión informada sobre lo que es mejor para la organización.
