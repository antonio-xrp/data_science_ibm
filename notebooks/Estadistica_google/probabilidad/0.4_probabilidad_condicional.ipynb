### Probabilidad Condicional 

Hasta ahora, has aprendido a calcular la probabilidad de un solo evento y de dos o más eventos independientes. Recuerda que dos eventos son independientes si un evento no afecta el resultado del otro evento, como el lanzamiento de dos monedas. Aprenderás a calcular la probabilidad de dos o más eventos dependientes. Este tipo de probabilidad se conoce como probabilidad condicional.

- La probabilidad condicional se refiere a la probabilidad de que ocurra un evento dado que ya ha ocurrido otro evento. La probabilidad condicional se usa en muchos campos diferentes, como las finanzas, los seguros, la ciencia y el aprendizaje automático.

Por ejemplo, una agencia que vende seguros de vida podría usar la probabilidad condicional para decidir qué tan riesgoso es asegurar a alguien que hace paracaidismo para ganarse la vida. 

Los profesionales de datos, como los que trabajan en modelos de aprendizaje automático, utilizan la probabilidad condicional para hacer predicciones precisas sobre conjuntos de datos complejos. 

### Dependent events

- Antes de empezar a calcular la probabilidad condicional, repasemos el concepto de dependencia. 

- Dos eventos son dependientes si la ocurrencia de un evento cambia la probabilidad del otro evento. Esto significa que el primer evento afecta al resultado del segundo evento. 

Por ejemplo, si quieres visitar un sitio web, primero necesitas acceso a Internet. La visita a un sitio web depende de que tenga acceso a Internet. Si desea viajar a otro país, primero debe obtener un pasaporte. Viajar a otro país depende de que tengas un pasaporte. En cada caso, podemos decir que el segundo evento depende o está condicionado al primer evento. 

Veamos un ejemplo de dependencia que se acerca más a la teoría de la probabilidad. 

Imagina que tienes dos eventos. El primer evento consiste en sacar un as de una baraja de cartas estándar, y el segundo evento consiste en sacar otro as de la misma baraja. Hay cuatro ases en una baraja de 52 cartas. Para el primer empate, la probabilidad de conseguir un as es de cuatro sobre 52, es decir, un 7,8 por ciento.

Pero en el segundo sorteo, la probabilidad de conseguir un as cambia porque has retirado una carta del mazo. Ahora, hay tres ases en una baraja de 51 cartas. Para el segundo sorteo, la probabilidad de conseguir un as es de tres sobre 51, o alrededor del 5,8 por ciento. Conseguir un as ahora es menos probable. Estos dos eventos son dependientes porque conseguir un as en el primer sorteo cambia la probabilidad de conseguir un as en el segundo sorteo.


Ahora tiene una mejor comprensión de los eventos dependientes. Volvamos a la probabilidad condicional y veamos la fórmula. 

No es necesario que memorice la fórmula, pero personalmente, me parece que revisar la fórmula a menudo me ayuda a entender mejor el concepto. Por eso lo comparto contigo. 

- Conditional probability P(A and B) = P(A) * P(B|A)
- P (A and B) = "probability of event A and event B"
- P(A) = "probability of event A"
- P(B|A) = "probability of event B givent event A"

La fórmula dice que para dos eventos dependientes, la probabilidad de que ocurran los eventos A y B es igual a la probabilidad del evento A multiplicada por la probabilidad del evento B dado A.

Puede observar que tenemos una nueva notación en esta fórmula: la **barra vertical entre las letras B y A significa que el evento B depende de que ocurra el evento A**.Decimos esto como la probabilidad de B dado A.


La fórmula también se puede expresar como la probabilidad del evento B dado que el evento A es igual a la probabilidad de que A y B ocurran dividida por la probabilidad de A. 

- P(B|A) = P(A and B) / P (A)
- P(A and B) = P(A) * P (B|A)

Estas son solo dos formas de representar la misma ecuación. 

Dependiendo de la situación o de la información que se le dé por adelantado, puede ser más fácil usar uno u otro. 

Podemos aplicar la fórmula de probabilidad condicional a nuestro ejemplo de cómo robar una carta de juego de as. 

La probabilidad del evento A o de conseguir un as en el primer sorteo es de cuatro sobre 52.

La probabilidad del evento B dado el evento A, o de conseguir un as en el segundo sorteo, es de tres sobre 51. 

Introduzcamos estos números en la fórmula. 

La probabilidad de los eventos A y B, o de conseguir dos ases seguidos, es de 4 sobre 52 multiplicada por tres sobre 51. Si haces los cálculos, esto equivale a uno sobre 221. 



- P(A and B) = P(A) * P(B|A)
- P(A) = ace on first draw = 4/52
- P(B|A) = ace on second draw = 3/51
- P(A and B) = ace on first draw and second draw = 4/52 * 3/51 = 1/221 = 0.5 %

La probabilidad de conseguir dos ases seguidos es igual a uno sobre 221, o alrededor del 0,5 por ciento.

Hace poco aprendió que la probabilidad condicional se refiere a la probabilidad de que ocurra un suceso dado que ya ha ocurrido otro. La probabilidad condicional le permite describir la relación entre sucesos dependientes, o cómo afecta la ocurrencia del primer suceso a la probabilidad del segundo.

En esta lectura, aprenderá a calcular la probabilidad condicional para dos o más sucesos dependientes. Antes de hablar del cálculo de la probabilidad condicional, repasaremos el concepto de dependencia.

### Probabilidad condicional
Anteriormente, calculó la probabilidad para un único suceso y para dos o más sucesos independientes, como dos lanzamientos consecutivos de una moneda. La probabilidad condicional se aplica a dos o más sucesos dependientes.

### Sucesos dependientes
Anteriormente, aprendió que dos sucesos son independientes si el primero no afecta al resultado del segundo ni cambia su probabilidad. Por ejemplo, dos lanzamientos consecutivos de una moneda son sucesos independientes. Obtener cara en el primer lanzamiento no afecta al resultado del segundo.

Por el contrario, dos sucesos son dependientes si la ocurrencia de uno de ellos cambia la probabilidad del otro. Esto significa que el primer suceso afecta al resultado del segundo.

Por ejemplo, si quiere sacar una buena nota en un examen, primero tiene que estudiar el material del curso. Sacar una buena nota depende del estudio. Si quiere comer en un restaurante popular sin esperar mesa, tiene que llegar pronto. Evitar una espera depende de llegar pronto. En cada caso, se puede decir que el segundo acontecimiento depende o es condicional al primero.

Ahora que entiende mejor los sucesos dependientes, volvamos a la probabilidad condicional y repasemos la fórmula.

### Fórmula de la probabilidad condicional
La fórmula dice que para dos sucesos dependientes A y B, la probabilidad de que ocurran el suceso A y el suceso B es igual a la probabilidad de que ocurra el suceso A, multiplicada por la probabilidad de que ocurra el suceso B, dado que el suceso A ya ha ocurrido.

### Probabilidad condicional

P(A y B) = P(A) * P(B|A)

En notación de probabilidad, la barra vertical entre las letras B y A indica dependencia, o que la ocurrencia del suceso B depende de la ocurrencia del suceso A. Se puede decir esto como "la probabilidad de B dado A".

La fórmula también se puede expresar como la probabilidad del suceso B dado el suceso A es igual a la probabilidad de que ocurran tanto A como B dividida por la probabilidad de A.

### Probabilidad condicional

P(B|A) = P(A y B) / P(A)

Éstas son sólo dos formas de representar la misma ecuación. Dependiendo de la situación, o de la información que se le facilite de antemano, puede ser más fácil utilizar una u otra.

**Nota:** La fórmula de la probabilidad condicional también se aplica a sucesos independientes. Cuando A y B son sucesos independientes, P(B|A) = P(B). Por lo tanto, la fórmula se convierte en P(A y B) = P(A) * P(B). Esta fórmula es también la regla de multiplicación que aprendió anteriormente en el curso.

**Ejemplo: jugar a las cartas**

Exploremos un ejemplo de probabilidad condicional que trata de una baraja estándar de 52 naipes.

Imagine dos sucesos:

- El primer suceso es sacar un corazón de la baraja.

- El segundo suceso es sacar otro corazón de la misma baraja.

Supongamos que quiere averiguar la probabilidad de sacar dos corazones seguidos. Estos dos sucesos son dependientes porque obtener un corazón en la primera extracción cambia la probabilidad de obtener un corazón en la segunda.

Una baraja estándar incluye cuatro palos diferentes: corazones, diamantes, picas y tréboles. Cada palo tiene 13 cartas. Para la primera extracción, la probabilidad de obtener un corazón es de 13 sobre 52, es decir, del 25%. Para la segunda extracción, la probabilidad de obtener un corazón cambia porque ya ha elegido un corazón en la primera extracción. Ahora, hay 12 corazones en una baraja de 51 cartas. Para la segunda extracción, la probabilidad de obtener un corazón es de 12 de 51, o alrededor del 23,5%. Conseguir un corazón es ahora menos probable: la probabilidad ha pasado del 25% al 23,5%.

Apliquemos ahora la fórmula de la Probabilidad condicional:

P(A y B) = P(A) * P(B|A)

Desea calcular la probabilidad de que se produzcan tanto el suceso A como el suceso B. Llamemos al suceso A 1er corazón , que se refiere a obtener un corazón en el primer sorteo. Llamemos al suceso B 2º corazón, que se refiere a obtener un corazón en la segunda extracción, dado que se extrajo un corazón la primera vez. La probabilidad del suceso A es 13/52, o el 25%. La probabilidad del suceso B es 12/51, o el 23,5%.

Introduzcamos estos números en la fórmula:

P(1er corazón y 2º corazón) = P(1er corazón) * P(2º corazón | 1er corazón) = 13/52 * 12/51 = 1/17 = 0,0588, o alrededor del 5,9%

Por lo tanto, hay un 5,9% de probabilidades de sacar dos corazones seguidos de una baraja estándar.

**Ejemplo: compras en línea**

Exploremos otro ejemplo. Imagine que es usted un profesional de los datos que trabaja para una tienda minorista en línea. Tiene datos que le dicen que el 20% de los clientes que visitan la página web de la tienda realizan una compra de 100 $ o más. Si un cliente gasta 100 dólares, tiene derecho a recibir una tarjeta regalo. La tienda concede aleatoriamente tarjetas regalo al 10% de los clientes que gastan al menos 100 $.

Quiere calcular la probabilidad de que un cliente gaste 100 $ y reciba una tarjeta regalo. Recibir una tarjeta regalo depende de que primero se gasten 100 $. Por lo tanto, se trata de una probabilidad condicional porque tiene que ver con dos sucesos dependientes.

Apliquemos la fórmula de la probabilidad condicional:

P(A y B) = P(A) * P(B|A)

Quiere calcular la probabilidad de que se produzcan tanto el suceso A como el suceso B. Llamemos al suceso A  100 $ y al suceso B tarjeta regalo. La probabilidad del suceso A es 0,2, o el 20%. La probabilidad del suceso B es 0,1, o el 10%.

P(100 $ y tarjeta regalo) = P(100 $) * P(tarjeta regalo dada 100 $) = 0,2 * 0,1 = 0,02, o 2%

Por lo tanto, la probabilidad de que un cliente gaste 100 $ o más y reciba una tarjeta regalo es de 0,2 * 0,1 = 0,02, o 2%.
