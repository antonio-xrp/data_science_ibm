Recientemente, ha aprendido sobre el teorema del límite central y cómo puede ayudarle a trabajar con una amplia variedad de conjuntos de datos. Los profesionales de los datos utilizan el teorema del límite central para estimar los parámetros de población de los datos en economía, ciencia, negocios y muchos otros campos.

En esta lectura, aprenderá más sobre el teorema del límite central y cómo puede ayudarle a estimar la media de la población para diferentes tipos de datos. Repasaremos la definición del teorema, las condiciones que deben cumplirse para aplicarlo y veremos un ejemplo del teorema en acción.

### Teorema del límite central

**Definición**

El teorema del límite central establece que la distribución muestral de la media se aproxima a una distribución normal a medida que aumenta el tamaño de la muestra. En otras palabras, a medida que aumenta el tamaño de la muestra, la distribución de muestreo adopta la forma de una curva de campana. Y, a medida que se muestrean más observaciones de una población, la media muestral se acerca más a la media de la población. Si se toma una muestra suficientemente grande de la población, la media muestral será aproximadamente igual a la media poblacional.

Por ejemplo, imaginemos que queremos estimar el peso medio de una determinada clase de vehículos, como las camionetas ligeras. En lugar de pesar millones de camionetas, puede obtener datos de una muestra representativa de camionetas. Si el tamaño de la muestra es lo suficientemente grande, el peso medio de la muestra será aproximadamente igual al peso medio de la población (de acuerdo con la ley de los grandes números).



**Nota:** El Teorema del límite central es válido para cualquier población. No es necesario conocer de antemano la forma de la distribución de la población para aplicar el teorema: la distribución puede tener forma de campana, estar sesgada o tener otra forma. Si recoge suficientes muestras de tamaño suficiente, la forma de la distribución de sus medias seguirá una distribución normal.

**Condiciones**

Para aplicar el teorema del límite central, deben cumplirse las siguientes condiciones:

- **Aleatoriedad**: Los datos de la muestra deben ser el resultado de una selección aleatoria. La selección aleatoria significa que todos los miembros de la población tienen las mismas posibilidades de ser elegidos para la muestra.

- **Independencia**: Los valores de la muestra deben ser independientes entre sí. Independencia significa que el valor de una observación no afecta al valor de otra observación. Normalmente, si sabe que los individuos o elementos de su conjunto de datos se seleccionaron aleatoriamente, también puede asumir la independencia.

    - **10%:** Para ayudar a garantizar que se cumple la condición de independencia, el tamaño de la muestra no debe ser superior al 10% de la población total cuando la muestra se extrae sin reemplazo (que suele ser el caso).

        - **Nota:** En general, se puede muestrear con o sin reemplazo. Cuando un elemento de la población sólo puede seleccionarse una vez, se trata de un muestreo sin sustitución. Cuando un elemento de la población puede seleccionarse más de una vez, se trata de un muestreo con sustitución. Más adelante aprenderá más sobre este tema.

- **Tamaño de la muestra:** El tamaño de la muestra debe ser suficientemente grande.

Analicemos la condición del tamaño de la muestra con más detalle. No existe una regla exacta sobre el tamaño que debe tener una muestra para que se aplique el teorema del límite central. La respuesta depende de los siguientes factores:

- **Requisitos de precisión**: Cuanto mayor sea el tamaño de la muestra, más se parecerá la distribución muestral a una distribución normal y más precisa será la estimación de la media de la población.

- **La forma de la población**: Si la distribución de la población tiene forma de campana y ya se parece a una distribución normal, la Distribución de muestreo de la media muestral se acercará a una distribución normal incluso con un tamaño de muestra pequeño.

En general, muchos estadísticos y profesionales de los datos consideran que un tamaño de muestra de 30 es suficiente cuando la distribución de la población tiene forma de campana o es aproximadamente normal. Sin embargo, si la población original no es normal -por ejemplo, si está muy sesgada o tiene muchos valores atípicos-, los profesionales de datos suelen preferir que el tamaño de la muestra sea un poco mayor. El Análisis exploratorio de datos puede ayudarle a determinar el tamaño de muestra necesario para un conjunto de datos determinado.

**Ejemplo: Salario anual**

Exploremos un ejemplo para hacernos una mejor idea de cómo funciona el teorema del límite central.

Imaginemos que estamos estudiando los datos salariales anuales de profesionales en activo de una gran ciudad como Buenos Aires, El Cairo, Delhi o Seúl. Supongamos que la población profesional en la que estás interesado incluye 10 millones de personas. Quiere conocer el salario medio anual de un profesional que viva en la ciudad. Sin embargo, no tiene tiempo ni dinero para encuestar a millones de profesionales y obtener datos completos sobre cada salario.

En lugar de encuestar a toda la población, recoges los datos de la encuesta a partir de muestras aleatorias repetidas de 100 profesionales. Con estos datos, calcula el salario medio anual en dólares de la primera muestra: 40.300 dólares. Para la segunda muestra, el salario medio es de 41.100 dólares. Encuestas una tercera muestra. El salario medio es de 39.700 dólares. Y así sucesivamente. Debido a la variabilidad del muestreo, la media de cada muestra será ligeramente diferente.




En teoría, se podría tomar una muestra muy grande y aumentar el tamaño de la muestra hasta haber encuestado a los 10 millones de personas sobre su salario. El teorema del límite central dice que, a medida que aumenta el tamaño de la muestra, la forma de la distribución de muestreo se parecerá cada vez más a una curva de campana.

Si toma una muestra suficientemente grande de la población, la media de su distribución de muestreo será aproximadamente igual a la media de la población. A partir de esta muestra de la población, se puede estimar con precisión el salario medio anual de toda la población de profesionales.

**Nota:** En la práctica, los profesionales de los datos suelen tomar una sola muestra. El tamaño concreto de la muestra que eligen depende de factores como el presupuesto, el tiempo, los recursos y el nivel de confianza deseado para su estimación.

### Puntos clave

El teorema del límite central puede ayudarle a inferir parámetros de población como la media, incluso si sólo dispone de datos de una parte de la población. Cuanto mayor sea el tamaño de la muestra, más precisa será la estimación de la media de la población. Tanto si mide el peso de un vehículo como su salario anual, el teorema del límite central es un método útil para comprender mejor sus datos.
