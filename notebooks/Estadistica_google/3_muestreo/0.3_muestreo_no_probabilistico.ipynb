Una muestra elegida al azar significa que todos los miembros de la población tienen las mismas posibilidades de ser incluidos. Por el contrario, los métodos de muestreo no probabilístico no utilizan la selección aleatoria, por lo que no suelen generar muestras representativas. De hecho, los métodos no probabilísticos suelen dar lugar a muestras sesgadas. El sesgo del muestreo se produce cuando algunos miembros de la población tienen más probabilidades de ser seleccionados que otros.

En esta lectura, obtendrá más información sobre cuatro métodos de muestreo no probabilístico y aprenderá cómo el sesgo del muestreo puede afectar a cada método. También analizaremos por qué el muestreo no probabilístico puede ser útil en determinadas situaciones.

### Métodos de Muestreo no probabilístico

Los muestreos no probabilísticos utilizan métodos de selección no aleatorios, por lo que no todos los miembros de una población tienen las mismas posibilidades de ser seleccionados. Por eso, los métodos no probabilísticos tienen un alto riesgo de sesgo de muestreo. Sin embargo, los métodos no probabilísticos suelen ser menos costosos y más cómodos de llevar a cabo para los investigadores. A veces, debido a la limitación de tiempo, dinero u otros recursos, no es posible utilizar el muestreo probabilístico. Además, los métodos no probabilísticos pueden ser útiles para estudios exploratorios, que buscan desarrollar una comprensión inicial de una población, en lugar de hacer inferencias sobre la población en su conjunto.

Repasaremos cuatro métodos de muestreo no probabilístico:

- Muestreo por conveniencia

- Muestreo de respuesta voluntaria

- Muestreo de bola de nieve

- Muestreo intencional

Analicemos cada método con más detalle.

### Muestreo por conveniencia

En el muestreo por conveniencia, se eligen miembros de una población con los que sea fácil ponerse en contacto o a los que sea fácil llegar. Por ejemplo, para realizar una encuesta de opinión, un investigador podría situarse a la entrada de un centro comercial durante el día y encuestar a las personas que pasen por allí.



Dado que estas muestras se basan en la conveniencia para el investigador, y no en una muestra más amplia de la población, las muestras de conveniencia a menudo sufren el sesgo de subcobertura. El sesgo de infracobertura se produce cuando algunos miembros de una población no están suficientemente representados en la muestra. Por ejemplo, la muestra anterior no representará a las personas a las que no les gusta comprar en centros comerciales, o prefieren comprar en un centro comercial diferente, o no visitan el centro comercial porque carecen de transporte.

El Muestreo por conveniencia suele ser rápido y barato, pero no es una forma fiable de obtener una muestra representativa.

### Muestreo de respuesta voluntaria

Una muestra de respuesta voluntaria está formada por miembros de una población que se ofrecen voluntariamente a participar en un estudio. Al igual que una muestra de conveniencia, una muestra de respuesta voluntaria suele basarse en el acceso conveniente a una población. Sin embargo, en lugar de que el investigador seleccione a los participantes, éstos se ofrecen voluntarios por su cuenta.

Por ejemplo, supongamos que los administradores de una universidad quieren saber qué opinan los estudiantes sobre la comida que se sirve en el campus. Les envían por correo electrónico un enlace a una encuesta en línea sobre la calidad de la comida y les piden que la rellenen si tienen tiempo.



Las muestras de respuesta voluntaria tienden a sufrir el sesgo de falta de respuesta, que se produce cuando determinados grupos de personas tienen menos probabilidades de proporcionar respuestas. Es probable que las personas que responden voluntariamente tengan opiniones más firmes, ya sean positivas o negativas, que el resto de la población. En este caso, es posible que sólo los alumnos a los que realmente les gusta o les disgusta la comida se sientan motivados para rellenar la encuesta. La encuesta puede omitir a muchos estudiantes que tienen opiniones más suaves sobre la comida, o que son neutrales. Esto hace que los estudiantes voluntarios no sean representativos del conjunto de la población estudiantil. 

### Muestreo de bola de nieve

En un muestreo de bola de nieve, los investigadores reclutan a los participantes iniciales para que participen en un estudio y luego les piden que recluten a otras personas para que participen en el estudio. Como en una bola de nieve, el tamaño de la muestra aumenta a medida que se unen más participantes. Los investigadores suelen utilizar el muestreo de bola de nieve cuando es difícil acceder a la población que quieren estudiar.

Por ejemplo, imaginemos que un investigador estudia a personas con una enfermedad poco frecuente. Por razones de confidencialidad, puede ser difícil para el investigador obtener información de contacto de los miembros de esta población a través de hospitales u otras fuentes oficiales. Sin embargo, si el investigador puede encontrar a un par de personas dispuestas a participar, es posible que estas dos personas conozcan a otras con la misma afección. Los participantes iniciales podrían entonces reclutar a otros compartiendo los beneficios potenciales del estudio.




El muestreo de bola de nieve puede llevar mucho tiempo, y los investigadores deben confiar en los participantes para continuar con éxito el proceso de reclutamiento y construir la "bola de nieve" Este tipo de reclutamiento también puede provocar un sesgo del muestreo. Dado que los participantes iniciales reclutan a más participantes por su cuenta, es probable que la mayoría de ellos compartan características similares, y estas características podrían no ser representativas de la población total objeto de estudio.

### Muestreo intencional

En el muestreo intencional, los investigadores seleccionan a los participantes en función del propósito de su estudio. Como los participantes se seleccionan para la muestra en función de las necesidades del estudio, se rechaza a los solicitantes que no se ajustan al perfil.

Por ejemplo, imaginemos que una empresa de desarrollo de videojuegos quiere realizar un estudio de mercado sobre un nuevo videojuego antes de su lanzamiento público. El equipo de investigación sólo quiere incluir en su muestra a expertos en videojuegos. Para ello, encuestan a un grupo de jugadores profesionales para que les den su opinión sobre posibles mejoras.



En el muestreo intencional, los investigadores suelen excluir intencionadamente a determinados grupos de la muestra para centrarse en un grupo específico que consideran más relevante para su estudio. En este caso, el investigador excluye a los jugadores aficionados. Los jugadores aficionados pueden comprar el nuevo juego por motivos diferentes a los de los jugadores profesionales y disfrutar de características del juego que no atraen a los profesionales. Esto podría dar lugar a resultados sesgados, ya que es probable que los profesionales de la muestra no sean representativos del conjunto de la población de jugadores.

El muestreo intencional se utiliza a menudo cuando un investigador desea obtener información detallada sobre una parte específica de una población, o cuando la población es muy pequeña y todos sus miembros tienen características similares. El muestreo intencional no es eficaz para hacer inferencias sobre una población grande y diversa.

### Puntos clave

El Muestreo no probabilístico es útil para recopilar datos en situaciones en las que se dispone de poco tiempo, presupuesto y otros recursos. El Muestreo no probabilístico también es útil para la investigación exploratoria, cuando se desea obtener una comprensión inicial de una población, en lugar de hacer inferencias sobre la población en su conjunto. Sin embargo, es importante recordar que los métodos de muestreo no probabilístico tienen un alto riesgo de sesgo de muestreo.

Como profesional de los datos, tiene que pensar en el sesgo y la imparcialidad desde el momento en que empieza a recoger los datos de la muestra hasta el momento en que presenta sus conclusiones. Una vez que conozca algunas formas comunes de sesgo, podrá permanecer alerta ante cualquier forma de sesgo.
