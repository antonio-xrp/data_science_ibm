### Como afecta el muestro a sus datos ​

​Como profesional de los datos, suelo ​trabajar con datos de muestra ​para hacer predicciones informadas ​sobre los ingresos por ventas futuros o el rendimiento de los productos. ​Comprender cómo el muestreo afecta a ​sus datos, tanto positiva como negativamente, ​será importante en su futura carrera ​en el análisis de datos. ​

Por ejemplo, los profesionales de datos unidireccionales utilizan ​las estadísticas de muestreo para estimar los parámetros de la población. 

​- **una estadística es una característica de ​una muestra y un parámetro ​es una característica de una población.** ​

### Statistic vs Parameter

Por ejemplo, 

- el peso medio de ​una muestra aleatoria de 100 pingüinos es una estadística.

- ​El peso medio de la población total de ​10 000 pingüinos es un parámetro. 

​Un profesional de datos podría usar ​el peso medio de la muestra de ​100 pingüinos para estimar ​el peso medio de la población. ​

Este tipo de estimación se denomina estimación puntual. 

### Point estimate 

​Una estimación puntual utiliza ​un único valor para estimar un parámetro de población.


### Sampling distribution 

​Una distribución muestral es ​una distribución de probabilidad de una estadística muestral. 

​Recuerde que una distribución de probabilidad ​representa los posibles resultados de una variable aleatoria, ​como lanzar una moneda o tirar un dado. ​Del mismo modo, una distribución ​muestral representa los posibles resultados de ​una estadística muestral, como la media.

​Imagine que toma muestras aleatorias simples repetidas ​del mismo tamaño de una población. 

​Dado que cada muestra es aleatoria, ​el valor medio variará de una ​muestra a otra de una manera que no se puede predecir con certeza. En este ​momento, esto puede parecer un poco abstracto. ​Para tener una mejor idea de ​la distribución muestral de la media, ​continuemos con nuestro ejemplo de pingüinos. 

​Imagina que estás estableciendo una población ​de 10 000 pingüinos azules, ​que son las especies de pingüinos más pequeñas conocidas. ​Quieres saber el peso medio de ​un pingüino azul en esta población. Como llevaría demasiado tiempo localizar ​y pesar a cada uno de los pingüinos, ​en su lugar se recopilan datos de muestra de la población. 

​Supongamos que tomas ​muestras aleatorias simples y repetidas de 10 pingüinos, ​cada uno de la población. ​En otras palabras, ​eliges al azar 10 pingüinos del grupo ​, los pesas y luego repites ​este proceso con un grupo diferente de 10 pingüinos. ​



- Para su primera muestra, ​encuentra que el peso medio de ​los 10 pingüinos es de 3.1 libras. ​- Para la segunda muestra, ​el peso medio de los 10 pingüinos es de 2.9 libras.
- ​Para la tercera muestra, ​el peso medio es de 2,8 libras, y así sucesivamente. ​

Imagina que el peso medio real de ​un pingüino en esta población es de tres libras.

​Aunque en la práctica, no lo sabrías ​a menos que pesaras a todos los pingüinos. ​Cada vez que tomes una muestra de 10 pingüinos, ​es probable que el peso medio de los pingüinos de la ​muestra se acerque a ​la media de la población de 3 libras, ​pero no exactamente 3 libras.

​De vez en cuando, puede obtener una muestra llena de ​pingüinos más pequeños que el promedio con ​un peso promedio de 2.5 libras o menos. ​O bien, puede obtener una simple combinación de ​pingüinos más grandes que el promedio con un peso promedio ​de 3.5 libras o más. 



​El peso medio variará aleatoriamente de una muestra a otra. ​L

### Sampling Variability 

La variabilidad del muestreo se refiere a cuánto ​varía una estimación entre las muestras.

​Puede usar una distribución muestral para representar ​la frecuencia de todas sus diferentes medias muestrales. 

​Me parece que ayuda ​visualizar estas muestras como un histograma. ​Vamos a trazar 10 muestras aleatorias simples de 10 pingüinos cada una. ​La media de muestreo más frecuente ​será de alrededor de tres libras. 



​Las medias de muestreo menos frecuentes serán ​las formas más extremas, como 2.3 o 3.7 libras. 

​A medida que aumente el tamaño de una muestra, ​el peso medio de los datos de la muestra se ​acercará al peso medio de la población. 

​Si toma una muestra de toda la población, es decir, ​si realmente pesó los 10 000 pingüinos, la ​media de la muestra sería la ​misma que la media de la población. ​Sin embargo, para obtener una estimación precisa de la media de la población, ​no es necesario pesar 10 000 pingüinos. 

​Si tomas un tamaño de muestra lo suficientemente grande de una población, ​digamos 100 pingüinos, la ​media de la muestra será ​una estimación precisa de la media de la población. 

​Este punto se basa en el **teorema del límite central**, ​que exploraremos ​con más detalle más adelante en el curso. ​Por ahora, solo debes saber que si tu muestra es lo suficientemente grande, la ​media de la muestra será aproximadamente igual a la media de la población. 

​Por ejemplo, imagina que ​recolectas una muestra de 100 pingüinos ​y descubres que el peso promedio de tu muestra es de 3 libras. ​Esto significa que su mejor estimación del peso medio de ​toda la población de pingüinos también es de 3 libras. 

​También puede usar su muestreo ​para estimar con qué precisión ​el peso medio de cualquier muestra determinada ​representa el peso medio de la población. ​Es útil saberlo porque ​la media que varía de una muestra a otra en ​cualquier muestra determinada no es ​necesariamente un reflejo exacto de la media de la población. 

​Por ejemplo, el peso medio real de ​la población de pingüinos podría ser de tres libras. ​El peso medio de cualquier muestra de ​pingüinos puede ser de 3.3 libras, ​2.8 libras, 2.4 libras, etc. 

​Mientras mayor sea la variabilidad de los datos de ​la muestra, es menos probable que la media de la muestra ​sea una estimación precisa de la media de la población. ​Los profesionales de datos utilizan la desviación estándar ​de las medias de la muestra para medir esta variabilidad. ​

Recuerde que la ​desviación estándar mide la variabilidad ​de los datos o la dispersión de los valores de los datos. ​Cuanto mayor sea la dispersión entre los valores de los datos, ​mayor será la desviación estándar. 

### Standard Error 

​En estadística, la desviación estándar de ​una estadística muestral se denomina error estándar. 

​- El error estándar de la media mide la ​variabilidad entre todas las medias de la muestra. 

- ​Un error estándar mayor indica ​que las medias de la muestra están más dispersas, ​si hay más variabilidad. 

- ​Un error estándar menor indica ​que las medias muestrales están más cerca unas de ​otras o que hay menos variabilidad. 


​Cuanto menor sea el error estándar, ​más probable es que la media de la muestra ​sea una estimación precisa de la media de la población. ​

Por ejemplo, supongamos que toma ​tres muestras aleatorias de 10 pingüinos cada una, 

- ​el peso medio de la primera muestra es de 3.3 libras. ​La segunda pesa 3,1 libras ​y la tercera 2,9 libras. ​No hay mucha variabilidad ​entre estas tres medias muestrales.

- ​Los valores están todos muy juntos.

- ​El error estándar será relativamente pequeño. 

​Ahora, supongamos que tomas ​otras tres muestras aleatorias de 10 pingüinos cada una.

​El peso medio de la primera muestra es de 2.2 libras, 

​la segunda es de 3.2 libras ​y 

la tercera es de 4.2 libras. 

​Hay más variabilidad entre estas tres medias muestrales. ​Los valores están más dispersos. ​El error estándar será relativamente grande. 



​Tenga en cuenta que el concepto de error estándar se ​basa en la práctica del muestreo repetido. ​En realidad, los investigadores suelen ​trabajar con una sola muestra. ​A menudo es demasiado complicado, caro ​o lento tomar ​muestras repetidas de una población. 

​En cambio, los estadísticos han obtenido ​una fórmula para calcular ​el error estándar basándose en ​la suposición matemática de un muestreo repetido. ​Puede usar la siguiente fórmula para ​calcular el error estándar de la media de la muestra.

​S dividido por la raíz cuadrada de n, ​donde S es la desviación estándar de la muestra ​y n es el tamaño de la muestra. ​



Por ejemplo, en tu estudio del peso de los pingüinos, ​imagina que una muestra de 100 pingüinos tiene un peso medio ​de tres libras y una desviación estándar de una libra.

​Puede calcular el error estándar dividiendo la ​desviación estándar de la muestra 1 por la raíz cuadrada del ​tamaño de la muestra 100,1 dividido por ​la raíz cuadrada de 100 es igual a 0,1. 



​Esto significa que su mejor estimación del ​peso medio real de la población ​de todos los pingüinos es de 3 libras, ​pero debe esperar que ​el peso medio de una muestra a la siguiente ​varíe con una desviación estándar de aproximadamente 0,1 libras. 

**​A medida que aumenta el tamaño de la muestra, ​el error estándar se reduce.**

Esto se debe a que el error estándar mide la diferencia ​entre la media de la muestra y la media real de la población. ​A medida que la muestra crece, la ​media de la muestra se acerca ​a la media real de la población. ​Cuanto más precisa sea la estimación de la media de ​la población, menor será el error estándar. ​Supongamos que recolectas una muestra de 10 000 ​pingüinos en lugar de 100 pingüinos.

​Descubrirá que el peso medio de la muestra es de ​3 libras y la desviación estándar de la muestra es de 1 libra. 

​El error estándar es uno dividido por la raíz cuadrada de ​10.000, lo que equivale a 0,01. ​



Su mejor estimación para la media de la muestra ​seguirá siendo de 3 libras, ​pero ahora puede esperar que el peso medio ​de una muestra de pingüinos a la siguiente ​varíe con una desviación estándar de solo 0,01 libras. 



​En general, puede tener ​más confianza en sus estimaciones a medida que ​el tamaño de la muestra ​aumenta y el error estándar se reduce. ​Esto se debe a que la media de la ​distribución muestral se ​acerca a la media de la población.
