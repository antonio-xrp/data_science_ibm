### Distribución de muestreo de la mediaEstado 

Recientemente, has aprendido cómo los profesionales de los datos utilizan las estadísticas muestrales para estimar los parámetros de la población. Por ejemplo, un profesional de datos puede estimar el tiempo medio que los clientes pasan en un sitio web de venta al por menor, o el salario medio de todas las personas que trabajan en la industria del entretenimiento.

En esta lectura, aprenderá más sobre el concepto de distribución de muestreo y cómo puede ayudarle a representar los posibles resultados de una muestra aleatoria. También discutiremos cómo la Distribución de muestreo de la media muestral puede ayudarle a estimar la media poblacional.


### Distribución de muestreo de la media muestral

Una distribución de muestreo es una distribución de probabilidad de un estadístico muestral. Recordemos que una distribución de probabilidad representa los posibles resultados de una variable aleatoria, como el lanzamiento de una moneda o de un dado. Del mismo modo, una distribución de muestreo representa los posibles resultados de una estadística muestral. Los estadísticos muestrales se basan en datos muestreados aleatoriamente, y sus resultados no pueden predecirse con certeza. Puede utilizar una distribución de muestreo para representar estadísticas como la media, la mediana, la desviación típica, el rango, etc.

Normalmente, los profesionales de los datos calculan estadísticos muestrales como la media para estimar los parámetros poblacionales correspondientes.

Supongamos que desea estimar la media de una población, como la altura media de un grupo de personas, animales o plantas. Una buena forma de entender el concepto de distribución de muestreo es imaginar que se toman muestras repetidas de la población, cada una con el mismo tamaño de muestra, y se calcula la media de cada una de estas muestras. Debido a la variabilidad del muestreo, la media de la muestra variará de una muestra a otra de una forma que no puede predecirse con certeza. La distribución de todas sus medias muestrales es esencialmente la distribución de muestreo. Puede visualizar la distribución de las medias muestrales en un histograma. Los estadísticos llaman a esto la Distribución de muestreo de la media.

**Nota:** En la práctica, debido a la limitación de tiempo y recursos, los profesionales de los datos suelen recoger una sola muestra y calcular la media de esa muestra para estimar la media de la población.

Veamos un ejemplo para tener una idea más concreta de la Distribución de muestreo de la media.

**Ejemplo: Longitud media de la trucha de lago**

Usted es un profesional de los datos que trabaja con un equipo de científicos medioambientales. Su equipo estudia los efectos de la contaminación del agua en las especies de peces. Actualmente, tu equipo está investigando los efectos de la contaminación en la población de truchas del Lago Superior, uno de los Grandes Lagos de Norteamérica. Como parte de esta investigación, te piden que estimes la longitud media de una trucha. Supongamos que hay 10 millones de truchas en el lago. En lugar de recoger y medir millones de truchas, tomas datos de muestras de la población.

Digamos que tomas muestras aleatorias simples repetidas de 100 truchas cada una de la población. En otras palabras, se eligen al azar 100 truchas del lago, se miden y se repite el proceso con un conjunto diferente de 100 truchas. Para la primera muestra de 100 truchas, la longitud media es de 20,2 pulgadas. Para la segunda muestra, la longitud media es de 20,5 pulgadas. Para la tercera muestra, la longitud media es de 19,7 pulgadas. Y así sucesivamente. Debido a la variabilidad del muestreo, la longitud media variará aleatoriamente de una muestra a otra.

Para el propósito de este ejemplo, vamos a suponer que la verdadera longitud media de una trucha en esta población es de 20 pulgadas. Aunque, en la práctica, no lo sabría a menos que midiera todas y cada una de las truchas del lago.

Cada vez que tome una muestra de 100 truchas, es probable que la longitud media de las truchas de su muestra se acerque a la media poblacional de 20 pulgadas, pero no exactamente a 20 pulgadas. De vez en cuando, puede obtener una muestra llena de truchas más cortas que la media, con una longitud media de 16 pulgadas o menos. O puede obtener una muestra llena de truchas más largas que la media, con una longitud media de 24 pulgadas o más.

Puede utilizar una Distribución de muestreo para representar la frecuencia de todas sus diferentes medias muestrales. Por ejemplo, si tomas 10 muestras aleatorias simples de 10 truchas cada una de la población, puedes mostrar la distribución de muestreo de la media como un histograma. El valor que aparece con más frecuencia en los datos de la muestra estará en torno a las 20 pulgadas. Los valores menos frecuentes serán las longitudes más extremas, como 16 o 24 pulgadas.



A medida que aumente el tamaño de la muestra, la longitud media de los datos de la muestra se acercará más a la longitud media de la población. Si tomara muestras de toda la población, es decir, si midiera todas las truchas del lago, la media de la muestra sería la misma que la media de la población.

Sin embargo, no es necesario medir millones de peces para obtener una estimación precisa de la media de la población. Si se toma una muestra lo suficientemente grande de la población (por ejemplo, 1.000 truchas), la media de la muestra será una estimación precisa de la media de la población (20 pulgadas).



### Error estándar

También puede utilizar los datos de la muestra para estimar con qué precisión la longitud media de una muestra determinada representa la media de la población.

Es útil saberlo porque la media muestral varía de una muestra a otra y es probable que cualquier media muestral difiera de la media poblacional real. Por ejemplo, la longitud media de la población de truchas puede ser de 20 pulgadas. La longitud media de una muestra dada de truchas puede ser de 20,2 pulgadas, 20,5 pulgadas, 19,7 pulgadas, y así sucesivamente.

Los profesionales de los datos utilizan la desviación típica de las medias muestrales para medir esta variabilidad. En estadística, la desviación estándar de una estadística muestral se denomina error estándar. El error estándar proporciona una medida numérica de la variabilidad del muestreo. El error estándar de la media mide la variabilidad entre todas las medias muestrales. Un error estándar mayor indica que las medias muestrales están más dispersas, o que hay más variabilidad. Un error estándar menor indica que las medias muestrales están más juntas, o que hay menos variabilidad.

En la práctica, utilizando una única muestra de observaciones, puede aplicar la siguiente fórmula para calcular el error estándar estimado de la media muestral: s / √n. En la fórmula, s se refiere a la desviación típica de la muestra y n al tamaño de la muestra.



Por ejemplo, en su estudio de las longitudes de las truchas, imagine que una muestra de 100 truchas tiene una longitud media de 20 pulgadas y una desviación típica de 2 pulgadas. Puede calcular el error estándar estimado dividiendo la desviación estándar de la muestra, 2, por la raíz cuadrada del tamaño de la muestra, 100:

2 ÷ √100 = 2 ÷ 10 = 0.2

Esto significa que debe esperar que la longitud media de una muestra a la siguiente varíe con una desviación estándar de aproximadamente 0,2 pulgadas.

El error estándar te ayuda a comprender la precisión de tu estimación. En general, puedes confiar más en tus estimaciones a medida que aumenta el tamaño de la muestra y disminuye el error estándar. Esto se debe a que, a medida que aumenta el tamaño de la muestra, la media muestral se acerca más a la media de la población.

### Puntos clave

La estimación de los parámetros de la población mediante muestreo es una potente forma de inferencia estadística. Las distribuciones de muestreo describen la incertidumbre asociada a un estadístico muestral y le ayudan a realizar inferencias estadísticas adecuadas. Esto es importante porque las decisiones de las partes interesadas se basan a menudo en las estimaciones que usted proporciona.
