### Types of statistics 

- ​Las estadísticas descriptivas, como la media y la ​desviación estándar, resumen las principales características del conjunto de datos. ​

- Las estadísticas inferenciales utilizan datos de muestra para sacar conclusiones o ​hacer predicciones sobre una población más grande. 

### La relación entre la muestra y la población

Anteriormente, ha aprendido que la estadística inferencial utiliza datos de muestra para extraer conclusiones o hacer predicciones sobre una población mayor. Los profesionales de los datos utilizan la estadística inferencial para obtener información valiosa sobre sus datos.

En esta lectura, conocerás con más detalle la relación entre muestra y población. También hablaremos de cómo los profesionales de los datos utilizan el muestreo en el trabajo con datos y de la importancia de trabajar con una muestra que sea representativa de la población.

### Población y muestra

**Población frente a muestra**

En estadística, una población incluye todos los elementos posibles que interesa medir, o todo el conjunto de datos sobre el que se quieren extraer conclusiones. Una población estadística puede referirse a cualquier tipo de datos, incluyendo:

- Personas

- Organizaciones

- Objetos

- Eventos

- Y otros

Por ejemplo, una población puede ser el conjunto de:

- Todos los estudiantes de una universidad

- Todos los teléfonos móviles fabricados por una empresa

- Todos los bosques de la Tierra

Una **muestra** es un subconjunto de una población.

Las muestras extraídas de las poblaciones anteriores podrían ser:

- Los estudiantes de matemáticas de la universidad

- Los teléfonos móviles fabricados por la empresa en la última semana

- Los bosques de Canadá

Los profesionales de los datos utilizan muestras para hacer inferencias sobre poblaciones. En otras palabras, utilizan los datos que recogen de una pequeña parte de la población para extraer conclusiones sobre el conjunto de la población.




**Muestreo**

El **muestreo** es el proceso de selección de un subconjunto de datos de una población.

En la práctica, suele ser difícil recopilar datos sobre cada miembro o elemento de una población entera. Una población puede ser muy grande, estar geográficamente dispersa o ser de difícil acceso. En cambio, se pueden utilizar datos de muestra para sacar conclusiones, hacer estimaciones o probar hipótesis sobre el conjunto de la población.

Los profesionales de los datos recurren al muestreo porque:

- A menudo es imposible o poco práctico recopilar datos de toda la población debido a su tamaño, complejidad o falta de accesibilidad.

- Es más fácil, rápido y eficaz recoger datos de una muestra

- Utilizar una muestra ahorra dinero y recursos

- Almacenar, organizar y analizar conjuntos de datos más pequeños suele ser más fácil, rápido y fiable que tratar con conjuntos de datos extremadamente grandes.

**Ejemplo: encuesta electoral**

Imagine que es un profesional de los datos que trabaja en un país con una gran población, como India, Indonesia, Estados Unidos o Brasil. Se van a celebrar unas elecciones presidenciales. Quieres hacer una encuesta electoral para ver qué candidato prefieren los votantes. Supongamos que la población con derecho a voto es de 100 millones de personas. Encuestar a 100 millones de personas sobre sus preferencias de voto llevaría una enorme cantidad de tiempo, dinero y recursos, incluso suponiendo que fuera posible localizar y contactar con todos los votantes y que todos ellos estuvieran dispuestos a participar.

Sin embargo, es realista encuestar a una muestra de 100 o 1.000 votantes extraída de la población más amplia de todos los votantes. Cuando se trata de una población grande, el muestreo puede ayudar a hacer inferencias válidas sobre la población en su conjunto.




**Muestra representativa**

Para hacer inferencias válidas o predicciones precisas sobre una población, la muestra debe ser representativa de la población en su conjunto. Recuerde que una muestra representativa refleja con precisión las características de una población. Las inferencias y predicciones que hagas sobre la población se basan en los datos de la muestra. Si la muestra no refleja con exactitud la población, las inferencias no serán fiables y las predicciones no serán exactas. Y esto puede conducir a resultados negativos para las partes interesadas y las organizaciones.

Métodos estadísticos como el muestreo probabilístico ayudan a garantizar la representatividad de la muestra mediante la recogida de muestras aleatorias de los distintos grupos de una población. Estos métodos ayudan a reducir el Sesgo del muestreo y aumentan la validez de sus resultados. Más adelante aprenderá más sobre los métodos de muestreo.

**Ejemplo: encuesta electoral**

Lo ideal es que la muestra de la encuesta electoral refleje fielmente las características de la población total de votantes. Una población de votantes de un país grande será diversa en cuanto a perspectivas políticas, ubicación geográfica, edad, sexo, raza, nivel educativo, estatus socioeconómico, etc. Su muestra no será representativa si sólo recoge datos de personas que pertenecen a determinados grupos y no a otros. Por ejemplo, si encuestas a personas de un partido político, o que tienen titulaciones superiores, o son mayores de 70 años. Los resultados de una encuesta electoral basada en una muestra no representativa no serán exactos. En general, cualquier afirmación o deducción que se haga sobre cualquier población tendrá más validez si se basa en una muestra representativa.

**Puntos clave**

Los profesionales de los datos trabajan con potentes herramientas estadísticas que pueden modelizar conjuntos de datos complejos y ayudar a generar información valiosa. Pero si los datos de la muestra con la que se trabaja no reflejan con exactitud la población, es decir, si la muestra no es representativa, no importa lo bueno que sea el modelo. Si su modelo predictivo se basa en una mala muestra, sus predicciones no serán exactas.

En última instancia, la calidad de la muestra ayuda a determinar la calidad de los conocimientos que se comparten con las partes interesadas. Para hacer inferencias fiables sobre una población, asegúrese de que su muestra es representativa de la población.
