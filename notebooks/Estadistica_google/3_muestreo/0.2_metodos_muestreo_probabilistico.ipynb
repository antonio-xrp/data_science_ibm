### Métodos de muestreo probabilístico

Anteriormente ha aprendido que existen dos tipos principales de métodos de muestreo: el muestreo probabilístico y el muestreo no probabilístico. El Muestreo probabilístico utiliza la selección aleatoria para generar una [muestra](https://www.statisticshowto.com/sample/). El muestreo no probabilístico suele basarse en la conveniencia, o en las preferencias personales del investigador, en lugar de en la selección aleatoria. El método de muestreo utilizado ayuda a determinar si la muestra es representativa de la población y si está sesgada. El Muestreo probabilístico le ofrece la mejor oportunidad de crear una muestra que sea representativa de la población.

En esta lectura, aprenderá más sobre los diferentes métodos de muestreo probabilístico, y los beneficios e inconvenientes de cada método.

### Métodos de Muestreo probabilístico

Existen cuatro métodos diferentes de muestreo probabilístico:

- Muestreo aleatorio simple

- Muestreo aleatorio estratificado

- Muestreo aleatorio por conglomerados

- Muestreo aleatorio sistemático

Analicemos cada método con más detalle.

### Muestreo aleatorio simple

En un muestreo aleatorio simple, cada miembro de una población se selecciona aleatoriamente y tiene la misma probabilidad de ser elegido. Puede seleccionar aleatoriamente a los miembros mediante un generador de números aleatorios u otro método de selección aleatoria.



Por ejemplo, imagine que desea encuestar a los empleados de una empresa sobre su experiencia laboral. La empresa tiene 10.000 empleados. Puede asignar a cada empleado de la base de datos de la empresa un número del 1 al 10.000 y, a continuación, utilizar un generador de números aleatorios para seleccionar a 100 personas para su muestra. En este caso, cada uno de los empleados tiene las mismas posibilidades de ser elegido para la muestra.

La principal ventaja de las muestras aleatorias simples es que suelen ser bastante representativas, ya que todos los miembros de la población tienen las mismas posibilidades de ser elegidos. Las muestras aleatorias tienden a evitar los sesgos, y las encuestas de este tipo ofrecen resultados más fiables.

Sin embargo, en la práctica, recoger muestras aleatorias simples de gran tamaño suele ser caro y llevar mucho tiempo. Y si el tamaño de la muestra no es suficiente, un grupo específico de personas de la población puede estar infrarrepresentado en la muestra. Si se utiliza una muestra de mayor tamaño, la muestra reflejará mejor la población.

### Muestreo aleatorio estratificado

En un muestreo aleatorio estratificado, se divide la población en grupos y se seleccionan al azar algunos miembros de cada grupo para que formen parte de la muestra. Estos grupos se denominan estratos. Los estratos pueden organizarse por edad, sexo, ingresos o cualquier otra categoría que le interese estudiar.



Por ejemplo, imagine que está realizando un estudio de mercado para un nuevo producto y quiere analizar las preferencias de los consumidores de distintos grupos de edad. Podría dividir su población objetivo en estratos según la edad: 20-29, 30-39, 40-49, 50-59, etc. A continuación, puede encuestar a un número igual de personas de cada grupo de edad y extraer conclusiones sobre las preferencias de los consumidores de cada grupo de edad. Los resultados ayudarán a los responsables de marketing a decidir en qué grupos de edad centrarse para optimizar las ventas del nuevo producto.

Las muestras aleatorias estratificadas ayudan a garantizar que los miembros de cada grupo de la población estén incluidos en la encuesta. Este método ayuda a proporcionar una representación equitativa a los grupos infrarrepresentados y permite extraer conclusiones más precisas sobre cada uno de los estratos. Puede haber diferencias significativas en los hábitos de compra de una persona de 21 años y otra de 51 años. El Muestreo estratificado ayuda a garantizar que ambas perspectivas queden recogidas en la muestra.

Uno de los principales inconvenientes del muestreo estratificado es que puede resultar difícil identificar los estratos adecuados para un estudio si se desconoce la población. Por ejemplo, si desea estudiar los ingresos medios de una población, es posible que desee estratificar la muestra por tipo de empleo, sector, ubicación o nivel educativo. Si no conoce la relevancia de estas categorías para la renta media, le resultará difícil elegir la mejor para su estudio.

### Muestreo de clústeres aleatorio

Cuando se realiza un muestreo aleatorio por conglomerados, se divide una población en conglomerados, se seleccionan aleatoriamente determinados conglomerados y se incluyen en la muestra todos los miembros de los conglomerados elegidos.

El Muestreo de clústeres es similar al Muestreo aleatorio estratificado, pero en éste se eligen aleatoriamente algunos miembros de cada grupo para que formen parte de la muestra. En el Muestreo de clústeres, se elige a todos los miembros de un grupo para que formen parte de la muestra. Los conglomerados se dividen utilizando detalles identificativos, como la edad, el sexo, la ubicación o lo que se quiera estudiar.



Por ejemplo, imagine que desea realizar una encuesta de satisfacción de los empleados en una franquicia mundial de restaurantes utilizando el muestreo de clústeres. La franquicia tiene 40 restaurantes en todo el mundo. Cada restaurante tiene aproximadamente el mismo número de empleados con funciones similares. Seleccione aleatoriamente 4 restaurantes como conglomerados. Incluya en la muestra a todos los empleados de los 4 restaurantes.

Una ventaja de este método es que una muestra de conglomerados incluye a todos los miembros de un conglomerado concreto, lo que resulta útil cuando cada conglomerado refleja la población en su conjunto. Este método es útil cuando se trata de poblaciones grandes y diversas que tienen subgrupos claramente definidos. Si los investigadores quieren saber más sobre la propiedad de la vivienda en los suburbios de Auckland (Nueva Zelanda), pueden utilizar varios suburbios bien elegidos como muestra representativa de todos los suburbios de la ciudad.

Una de las principales desventajas del Muestreo de clústeres es que puede resultar difícil crear conglomerados que reflejen con exactitud la población total. Por ejemplo, por razones prácticas, es posible que sólo tenga acceso a restaurantes de Inglaterra cuando la franquicia tiene establecimientos en todo el mundo. Y los empleados de Inglaterra pueden tener características y valores diferentes de los empleados de otros países.

### Muestreo aleatorio sistemático

En un muestreo aleatorio sistemático, se coloca a cada miembro de una población en una secuencia ordenada. A continuación, se elige un punto de partida aleatorio en la secuencia y se seleccionan miembros para la muestra a intervalos regulares.



Imagine que desea encuestar a los alumnos de un instituto sobre sus hábitos de estudio. Para una muestra aleatoria sistemática, pondría los nombres de los estudiantes en orden alfabético y elegiría al azar un punto de partida: digamos, el número 4. A partir del número 4, se selecciona uno de cada 10 nombres de la lista (4, 14, 24, 34, ... ), hasta obtener una muestra de 100 estudiantes.

Una ventaja de las muestras aleatorias sistemáticas es que suelen ser representativas de la población, ya que cada miembro tiene las mismas posibilidades de ser incluido en la muestra. El hecho de que el apellido del alumno empiece por L o por Q no va a afectar a sus características. El muestreo sistemático también es rápido y cómodo cuando se dispone de una lista completa de los miembros de la población.

Una desventaja del muestreo sistemático es que, antes de empezar, hay que conocer el tamaño de la población que se quiere estudiar. Si no se dispone de esta información, es difícil elegir intervalos coherentes. Además, si hay un patrón oculto en la secuencia, puede que no obtenga una muestra representativa. Por ejemplo, si uno de cada diez nombres de la lista es un estudiante con matrícula de honor, es posible que sólo obtenga información sobre los hábitos de estudio de los estudiantes con matrícula de honor, y no de todos los estudiantes.

### Puntos clave

Los cuatro métodos de muestreo probabilístico que hemos analizado -simple, estratificado, por conglomerados y sistemático- se basan en la selección aleatoria, que es el método de muestreo preferido por la mayoría de los profesionales de datos. Los métodos de muestreo probabilístico le ofrecen la mejor oportunidad de crear una muestra que sea representativa de la población en su conjunto. Y trabajar con una muestra representativa le permite hacer inferencias fiables y predicciones precisas sobre la población que está investigando.
