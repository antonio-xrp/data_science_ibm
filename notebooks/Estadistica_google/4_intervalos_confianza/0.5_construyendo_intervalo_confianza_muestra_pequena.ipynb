### Construya un intervalo de confianza para un tamaño de muestra pequeñoEstado:

Hasta ahora, ha construido intervalos de confianza para tamaños de muestra grandes, que suelen definirse como tamaños de muestra de 30 o más elementos. Por ejemplo, cuando estimó la duración media de la batería de un nuevo teléfono móvil, utilizó una muestra aleatoria de 100 teléfonos. Por otro lado, las muestras pequeñas suelen definirse como aquellas que tienen menos de 30 elementos. Normalmente, los profesionales de los datos intentan trabajar con tamaños de muestra grandes porque ofrecen estimaciones más precisas. Pero no siempre es posible trabajar con una muestra grande. En la práctica, la recogida de datos suele ser cara y requiere mucho tiempo. Si no se dispone del tiempo, el dinero o los recursos necesarios para tomar una muestra grande, se puede acabar trabajando con una muestra pequeña.

En esta lectura, aprenderá a construir un intervalo de confianza para un tamaño de muestra pequeño. Analizaremos paso a paso un ejemplo relacionado con los niveles medios de emisiones del motor de un coche nuevo.

### Tamaños de muestra grandes frente a pequeños

En primer lugar, vamos a discutir brevemente los diferentes métodos que se utilizan para construir intervalos de confianza para tamaños de muestra grandes y pequeños.

### Muestra grande: Puntuaciones Z

Para tamaños de muestra grandes, se utilizan puntuaciones z para calcular el margen de error, al igual que se hizo anteriormente para estimar la duración media de la batería de los teléfonos móviles. Esto se debe al teorema del límite central: para tamaños de muestra grandes, la media de la muestra se distribuye aproximadamente de forma normal. Para una distribución normal estándar, también llamada distribución z, se utilizan puntuaciones z para hacer cálculos sobre los datos.

### Muestra pequeña: Puntuaciones T

Para muestras de pequeño tamaño, es necesario utilizar una distribución diferente, denominada distribución t. Desde el punto de vista estadístico, esto se debe a que hay más incertidumbre en la estimación del error estándar para tamaños de muestra pequeños. No necesita preocuparse por los detalles técnicos, que están fuera del alcance de este curso. Por ahora, sólo sepa que si está trabajando con un tamaño de muestra pequeño, y sus datos se distribuyen aproximadamente de forma normal, debe utilizar la distribución t en lugar de la distribución normal estándar. En una distribución t, se utilizan las puntuaciones t para hacer cálculos sobre los datos.

La gráfica de la distribución t tiene una forma de campana similar a la de la distribución normal estándar. Sin embargo, la distribución t tiene colas más grandes que la distribución normal estándar. Las colas más grandes indican la mayor frecuencia de valores atípicos que se producen en un conjunto de datos pequeño. A medida que aumenta el tamaño de la muestra, la distribución t se aproxima a la distribución normal. Cuando el tamaño de la muestra alcanza 30, las distribuciones son prácticamente iguales y puede utilizar la distribución normal para sus cálculos.




### Ejemplo: Niveles medios de emisión

Ahora que ya conoce un poco la distribución t y las puntuaciones t, vamos a construir un intervalo de confianza para un tamaño de muestra pequeño.

### Contexto

Imagina que eres un profesional de datos que trabaja para un fabricante de automóviles. La empresa produce coches de alto rendimiento que se venden en todo el mundo. Normalmente, los motores de estos coches tienen altas tasas de emisión de dióxido de carbono, oCO2, que es un gas de efecto invernadero que contribuye al calentamiento global. El equipo de ingenieros ha diseñado un nuevo motor para reducir las emisiones del coche más vendido de la empresa.

### Objetivo

El objetivo es mantener las emisiones por debajo de 460 gramos deCO2 por milla. Esto garantizará que el coche cumpla las normas de emisiones en todos los países en los que se venda. Además, el menor índice de emisiones es bueno para el medio ambiente, lo que atraerá a nuevos clientes.

### Pregunta a

El equipo de ingenieros te pide una estimación fiable de la tasa de emisiones del nuevo motor. Debido a problemas de producción, sólo hay un número limitado de motores disponibles para las pruebas. Por lo tanto, trabajarás con una muestra pequeña.

### Muestra

El equipo de ingenieros prueba una muestra aleatoria de 15 motores y recopila datos sobre sus emisiones. La tasa media de emisiones es de 430 gramos deCO2 por milla, y la desviación estándar es de 35 gramos deCO2 por milla.

Es posible que la muestra individual no proporcione la tasa media real de emisiones de cada motor. La media poblacional de emisiones podría estar por encima o por debajo de 430 gramos deCO2 por milla. Aunque sólo disponga de una pequeña muestra de motores, puede construir un intervalo de confianza que probablemente incluya el índice de emisiones real para una gran población de motores. Esto dará a su jefe una mejor idea de la incertidumbre en su estimación. También ayudará al equipo de ingeniería a decidir si necesitan trabajar más en el motor para reducir el índice de emisiones.

### Construir el Intervalo de confianza

Repasemos los pasos para construir un intervalo de confianza:

1. Identifique una estadística de muestra.

2. Elige un Nivel de confianza.

3. Encontrar el Margen de error.

4. Calcular el intervalo.

### Paso 1: Identificar una estadística de muestra

En primer lugar, identifique su estadística muestral. Su muestra representa la tasa media de emisiones de 15 motores. Está trabajando con una media muestral.

### Paso 2: Elija un Nivel de confianza

A continuación, elija un Nivel de confianza. El equipo de ingeniería le pide que elija un nivel de confianza del 95%.

### Paso 3: Encontrar el margen de error

El tercer paso consiste en hallar el margen de error. Para un tamaño de muestra pequeño, se calcula el margen de error multiplicando la puntuación t por el error estándar.

La distribución t se define mediante un parámetro denominado grado de libertad. En nuestro contexto, el grado de libertad es el tamaño de la muestra - 1, o 15-1 = 14. Dado su grado de libertad y su nivel de confianza, puede utilizar un lenguaje de programación como Python u otro software estadístico para calcular su puntuación t.

Basándose en un grado de libertad de 14 y un nivel de confianza del 95%, su puntuación t es 2,145.

Ahora puedes calcular el error estándar, que mide la variabilidad de tu estadística muestral.

Ésta es la fórmula para el error estándar de la media que has utilizado antes:

### Error estándar de la media

 
$SE(x)=s/√(n)$


En la fórmula, la letra s se refiere a la desviación típica de la muestra, y la letra n se refiere al tamaño de la muestra.

Su Desviación estándar muestral es 35, y su tamaño muestral es 15. El cálculo le da un error estándar de aproximadamente 9,04.

El Margen de error es su puntuación t multiplicada por su error estándar. Esto es 2,145 * 9,04 = 19,39.

### Paso 4: Calcular el intervalo

Por último, calcule el intervalo de confianza. El límite superior del intervalo es la media muestral más el margen de error. Esto es 430 + 19,39 = 449,39 gramos deCO2 por milla.

El límite inferior es la media de la muestra menos el margen de error. Esto es 430 - 19,39 = 410,61 gramos deCO2 por milla.

Se tiene un intervalo de confianza del 95% que va desde 410,61 gramos deCO2 por milla hasta 449,39 gramos deCO2 por milla.

### 95 CI [410,61, 449,39]

El intervalo de confianza proporciona al equipo de ingenieros información importante. El límite superior de su intervalo está por debajo del objetivo de 460 gramos deCO2 por milla. Este resultado proporciona una sólida evidencia estadística de que la tasa de emisiones del nuevo motor cumplirá las normas de emisiones.

Nota: Los intervalos de confianza para tamaños de muestra pequeños sólo se refieren a medias de población, no a proporciones de población. La razón estadística de esta distinción es más bien técnica, así que no tienes que preocuparte por ello por ahora.

### Puntos clave

Como profesional de los datos, trabajará tanto con muestras grandes como pequeñas. Aunque las muestras grandes proporcionan estimaciones más precisas que las pequeñas, la recogida de una muestra pequeña suele ser menos costosa y requerir menos tiempo que la recogida de una muestra grande. Saber cómo construir intervalos de confianza para diferentes tamaños de muestra le ayudará a gestionar cualquier conjunto de datos que pueda encontrar en su futura carrera.
