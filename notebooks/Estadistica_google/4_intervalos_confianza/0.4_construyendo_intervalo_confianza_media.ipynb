### Construya un intervalo de confianza para una media

El proceso básico es el mismo que el que utilizas para la ​proporción, pero requiere nuevos cálculos. Repasaré paso a paso un ejemplo ​relacionado con la comercialización de un teléfono móvil nuevo. ​Imagine que es un profesional de datos que trabaja para una empresa que produce teléfonos móviles. ​Recientemente, la compañía desarrolló un teléfono con una batería de mayor duración. 

​Está diseñado para funcionar durante al menos 20 horas sin necesidad de recarga. 

​Esta es una gran mejora en la duración de la batería y aumentará las ventas. ​El equipo de marketing está planificando una campaña publicitaria sobre la nueva batería para ayudar a ​vender el teléfono. La dirección quiere asegurarse de que la afirmación sobre las ​20 horas de duración de la batería sea correcta para que los anuncios se hagan públicos. 

​Le piden que analice los datos y haga una estimación confiable de la duración ​de la batería del nuevo teléfono. ​La compañía ha producido 100 000 teléfonos nuevos. ​El equipo de ingeniería de productos analiza una muestra aleatoria de 100 teléfonos y ​registra los datos sobre la duración de la batería en función de los datos. 


​Usted sabe que la duración media de la batería muestreada es de 20,5 horas y ​la desviación estándar de la muestra es de 1,7 horas.​Y según los datos de la campaña sobre el proceso estándar de fabricación de ​las baterías, ​también sabes que la desviación estándar de la población es de 1,5 horas. 



​La media de la muestra es de más de 20 horas para esta prueba. ​Sin embargo, usted sabe que esta es solo una muestra aleatoria de 100 ​teléfonos de una población grande. ​Si tomaste otra muestra aleatoria de 100 teléfonos, es posible que obtengas resultados diferentes. ​Si tomó una tercera muestra aleatoria, los resultados pueden volver a diferir y así sucesivamente. ​Es posible que tu muestra única no proporcione la duración media real de la batería de ​todos los teléfonos. La duración media de la batería de la población podría ser de 19 horas, ​21 horas o algo más. 



​Puedes usar los datos de tu muestra para construir una revuelta de confianza que probablemente incluya ​la media de la población, en cuanto a la duración de la batería del teléfono. ​Esto le dará al equipo de marketing una mejor idea de la incertidumbre de tu estimación. ​También les ayudará a decidir cómo anunciar el teléfono y ​si pueden afirmar que su batería dura 20 horas o más. 

### Steps for constructing a confidence level

- ​Repasemos los pasos para construir un intervalo de confianza. ​

- En primer lugar, identifique una estadística de muestra. 

​- En segundo lugar, elija un nivel de confianza, 

- En tercer lugar, encuentre el margen de error y, 

- En ​cuarto lugar, calcule el intervalo. 


​En primer lugar, identifique la estadística de la muestra. La ​muestra representa la duración media de la duración de la batería de ​100 teléfonos móviles. En este ejemplo, estás trabajando con la media de la muestra. 

​A continuación, elija un nivel de confianza La solicitud de administración que elija con un ​nivel de confianza del 95%. ​Este es el estándar de la empresa para los nuevos productos. 

El ​tercer paso es encontrar el margen de error, que se refiere al rango de valores ​por encima y por debajo de la estadística de muestra. Puede calcular el margen de error ​multiplicando la puntuación Z por el error estándar. ​Puede recordar que la puntuación Z que utiliza depende de su nivel de confianza. ​Esta tabla muestra las puntuaciones Z que corresponden a los niveles de confianza populares, ​como 90, 95 y 99%. 




​La puntuación Z para un nivel de confianza del 95% es de 1,96. ​Ahora puede calcular el error estándar, ​que mide la variabilidad de la estadística de la muestra. ​¿Muestra en qué medida es probable que la media de la muestra difiera de la ​media real de la población? ​Cuanto mayor sea el error estándar, mayor será la variabilidad. ​La fórmula para el error estándar de la media es la ​desviación estándar de la población dividida por la raíz cuadrada del tamaño de la muestra. 



La ​desviación estándar de la población es de 1.5 y el tamaño de la muestra es de 100. ​Si ingresas los números en la fórmula, obtendrás un error estándar de 0.15. 

​El margen de error es su puntuación Z de 1,96 multiplicado por su error estándar. ​0.15 es igual a 0.294. ​Por último, calcule su intervalo de confianza. 

​El límite superior del intervalo es la media de la muestra más el margen de error. 



​Esto es 20,5 más 0,294, lo que equivale a 20,794 horas o ​aproximadamente 20 horas y 48 minutos. 

​El límite inferior es la media muestral menos el margen de error, es decir, ​20,5 menos 0,294 es igual a 20,206 o ​aproximadamente 20 horas y 12 minutos. 



​Por lo tanto, tiene un intervalo de confianza del 95% para la duración de la batería del teléfono, que ​va de 20 horas y 12 minutos a 20 horas y 48 minutos. 


​El intervalo de confianza proporciona a la dirección de la empresa información importante. 

​El límite inferior de su intervalo de 20 horas y ​12 minutos está por encima del objetivo de 20 horas de la empresa. ​Esto ayuda al equipo de marketing a tener la confianza de anunciar que ​la duración de la batería del teléfono celular será de al menos 20 horas. 



​Presentas tus hallazgos a las partes interesadas de la empresa y ​los resultados satisfacen a todos excepto al jefe de marketing. ​El director de marketing ha dedicado mucho tiempo y ​esfuerzo a desarrollar la campaña publicitaria y quiere tener aún más confianza. 

​El director solicita que analices los datos con un nivel de confianza del 99%. Para ​que el director de marketing esté satisfecho de que recalcules tus resultados. 

​Utilice los mismos datos de muestra pero con un nivel de confianza del 99% en lugar del 95%. ​Su intervalo de confianza ahora se extiende de 20 horas y ​siete minutos a 20 horas y 53 minutos. ​El límite inferior del intervalo sigue siendo superior a 20 horas. ​Este resultado debería dar a la dirección de la empresa aún más confianza en la duración de la ​batería y, con suerte, satisfacer al director de marketing. ​Puede observar que a medida que el nivel de confianza aumenta, el intervalo de confianza ​se amplía. Con un nivel de confianza del 95%, el intervalo cubre 36 minutos, ​con un nivel de confianza del 99%, el intervalo cubre 46 minutos. ​Esto se debe a que es más probable que un intervalo de confianza más amplio incluya ​el parámetro de población real. 




​Tenga en cuenta que en este ejemplo sabemos que la desviación estándar de la población es de ​1,5 horas. ​Sin embargo, en la práctica, la desviación estándar de la población a menudo se desconoce y ​debe estimarse en función de la desviación estándar de la muestra. ​Esto se debe a que es difícil obtener datos completos sobre una población grande si ​no se conoce la desviación estándar de la población. ​Esto cambia los cálculos del intervalo de confianza para obtener más información, no ​dude en consultar la lectura correspondiente. ​
