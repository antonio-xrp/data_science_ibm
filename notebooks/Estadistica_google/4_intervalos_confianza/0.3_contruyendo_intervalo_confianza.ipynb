### Contruyendo un intervalo de confianza para una proporción

​Imagine que es un profesional de datos ​que trabaja para una agencia electoral. ​Próximamente hay una elección para ​gobernador entre dos candidatas, ​Tiffany Davis y Maya Cruz. ​Su agencia representa la campaña de Davis. Faltan ​cuatro semanas para el día de las elecciones. 
​
El equipo de Davis le pide que realice una encuesta ​para saber cómo le va a su candidato. ​Reúne una muestra aleatoria de 100 votantes ​de una población total de 100 000 votantes. ​Les preguntas por qué candidato piensan votar. 

​Los resultados muestran que el ​55 por ciento de los votantes prefería a Davis ​y el 45 por ciento de los votantes prefería a Cruz. 



​La encuesta favorece a su candidato. ​Si Davis obtiene más del 50 por ciento ​de los votos el día de las elecciones, ​es una victoria, por lo que el 55 por ciento ​es un buen resultado. Buenas noticias, ¿verdad? 


​Pero también sabe que se trata ​solo de una muestra aleatoria de ​100 votantes de una población grande. ​Si tomaste otra muestra aleatoria de 100 votantes, ​es posible que obtengas resultados diferentes. ​Si tomó una tercera muestra, ​los resultados pueden volver a diferir y así sucesivamente. 



​En otras palabras, es posible que su muestra única no proporcione ​la proporción o el porcentaje real de la población ​de todos los votantes que votarán por Davis el día de las elecciones. 

​Por ejemplo, el día de las elecciones, ​Davis puede obtener el 52 por ciento, ​que es lo suficientemente bueno para ganar, ​o un 49 por ciento, que no lo es. ​

En lugar de confiar en una estimación puntual ​como prueba de que su candidato ganará las elecciones, ​puede usar los datos de la muestra para ​construir un intervalo de confianza. ​Esto le dará al equipo de campaña ​una mejor idea de la incertidumbre de su estimación ​y de los posibles resultados electorales. 


### Steps for constructing a confidence interval

​​Repasemos los pasos para ​construir un intervalo de confianza.

- ​En primer lugar, identifique una estadística de muestra. 
- ​En segundo lugar, elige un nivel de confianza. 
- ​En tercer lugar, encuentra el margen de error 
- En cuarto lugar, calcula el intervalo. 


- ​En primer lugar, identifique la estadística de la muestra. ​Su polo representa el porcentaje de ​votantes que prefieren a su candidato, ​que es del 55 por ciento. ​Esta es una proporción muestral. 

​- A continuación, elige un nivel de confianza para tu encuesta. ​La mayoría de las encuestas electorales ​arrojan un nivel de confianza del 95 por ciento. 



​La campaña de Davis también solicita que utilices ​un nivel de confianza del 95 por ciento en tus cálculos. El ​tercer paso es encontrar el margen de error. 

**​El margen de error** se refiere al rango de ​valores por encima y por debajo de la estadística de muestra. ​Si trabajas con una distribución normal ​y un tamaño de muestra grande, ​una forma de calcular el margen de error es ​multiplicar la puntuación z por el error estándar. 



​Vamos a desglosar eso. ​Para revisar, una puntuación z mide la distancia de ​un punto de datos desde la media de la población ​y una distribución normal estándar. 

​Por ejemplo: 

- una puntuación z de ​uno es una desviación estándar por encima de la media. 

- ​Una puntuación z de menos 1.5 es ​1.5 desviaciones estándar por debajo de la media. 



​Esta tabla muestra las puntuaciones z que ​corresponden a los niveles de confianza populares. ​1.645 para el 90 por ciento, ​1.96 para el 95 por ciento ​y 2.58 para el 99 por ciento. ​Si elige un nivel de confianza del 95 por ciento, ​utilice una puntuación z de 1,96 para calcular el margen de error. 



​Ahora, necesitas calcular tu error estándar. ​Tal vez recuerde que el error estándar mide ​la variabilidad de la estadística de su muestra. ​Muestra en qué medida es probable que la proporción de la muestra ​difiera de la proporción real de la población. ​Cuanto mayor sea el error estándar, ​mayor será la variabilidad de la muestra. 

​La fórmula para el error estándar de ​la proporción es la raíz cuadrada de la proporción de ​la muestra multiplicada por 1 menos ​la proporción de la muestra dividida por el tamaño de la muestra. 



​La proporción de la muestra es de ​0,55 y el tamaño de la muestra es de 100. ​Si introduce los números en la fórmula, ​obtendrá un error estándar de aproximadamente 0,05. 



​Pongámoslo en conjunto. ​El margen de error es la puntuación z de ​1,96 multiplicada por el error estándar de 0,05. ​Esto equivale a 0,098. Por último, ​el último paso del proceso para construir ​un intervalo de confianza es calcular el intervalo. 



​El límite superior del intervalo es ​la proporción de la muestra más el margen de error, ​o 0,55 más 0,098. ​Esto equivale a 0,648 o 64,8 por ciento. ​El límite inferior es ​la proporción de la muestra menos el margen de error. 



​Esto es 0.55 menos 0.098 igual a 0.452 o 45.2 por ciento. ​Por lo tanto, tiene un ​intervalo de confianza del 95 por ciento que va ​del 45,2 al 64,8 por ciento. 



​Si bien su intervalo de confianza ​se sitúa en su mayoría por encima del 50 por ciento, ​esto no es necesariamente una razón para ser ​optimista con respecto a las próximas elecciones, ​ya que el límite inferior del 45,2 ​por ciento cae por debajo del 50 por ciento. ​



Según el intervalo de confianza, ​perder las elecciones sigue siendo una posibilidad. ​Es posible que el equipo de la campaña quiera invertir más en ​publicidad en televisión o redes sociales para garantizar la victoria. 

​O si el equipo de campaña quiere ​una estimación más precisa de los resultados de las elecciones, ​puede solicitar otra encuesta con un tamaño de muestra más grande. ​Esta estimación será más precisa ​porque incluye a más votantes. 

​Supongamos que realizas otra encuesta con ​una muestra de 1000 votantes. 

​La nueva encuesta informa que el ​54 por ciento de los votantes prefería al candidato Davis. ​

Si calcula un intervalo de confianza del 95 por ciento ​con estos números, ​su integral pasará del ​50,9 por ciento al 57,1 por ciento. 



​El límite inferior del intervalo ahora está por encima del 50 por ciento. ​Esto debería dar al equipo de Davis ​más confianza sobre las próximas elecciones. ​Por supuesto, todavía existe la posibilidad de que ​su candidato pierda las elecciones, ​ya que el nivel de confianza es del ​95 por ciento, no del 100 por ciento.

​Puede observar que a medida que aumenta el tamaño de la muestra, ​el intervalo de confianza se reduce. 



​Con una muestra de 100, ​el intervalo abarca 19,6 puntos porcentuales. ​Con un tamaño de muestra de 1000, ​el intervalo abarca 6,2 puntos porcentuales. ​

Esto se debe a que, a medida que aumenta el tamaño de la muestra, el ​margen de error disminuye. Si pudieras muestrear a todos los miembros de ​la población, el margen de error sería cero. 

​Pero, por supuesto, a menudo es ​demasiado caro y lleva mucho tiempo ​muestrear a toda una población ​o tomar muestras repetidas. 

​Los profesionales de datos suelen trabajar con ​una única muestra aleatoria de una población grande. Los ​intervalos de confianza ayudan a los profesionales de datos a ofrecer ​estimaciones más confiables basadas en los datos disponibles ​y, según sus datos, ​es probable que su candidato gane las elecciones.
