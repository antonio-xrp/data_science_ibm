### Introducción a los intervalos de confianza

​Anteriormente, hablamos sobre cómo los profesionales de datos ​hacen **estimaciones puntuales** sobre los parámetros de la población. ​

Por ejemplo:

- basándose en una muestra de 100 pingüinos, ​un profesional de datos podría estimar que el peso medio de ​una población de 10 000 pingüinos es de 31 libras.

- basándose en una encuesta de 100 votantes, ​un profesional de datos podría estimar que el 55% de los 100 ​000 votantes prefieren a un candidato determinado ​en una próxima elección. 

### Point estimate

​Una estimación puntual utiliza ​un único valor para estimar un parámetro de población. 

### Interval estimate

​Por el contrario, una estimación de intervalo utiliza ​un rango de valores para estimar un parámetro de población. 

​Un intervalo de confianza es un tipo de estimación de intervalo.

​Por ejemplo:

- para el peso de los pingüinos, ​puedes construir un intervalo de confianza del 95% ​entre 28 y 32 libras,

- ​o para la encuesta electoral, ​puedes construir un ​intervalo de confianza del 99% entre el 51 y el 57%. 

### Penguin data

​Por lo general, los profesionales de datos utilizan ​intervalos de confianza en lugar de ​estimaciones puntuales para compartir los resultados. ​Una estimación puntual puede ser útil, ​pero un valor único, como 30 libras, no ​expresa la incertidumbre incorporada en ninguna estimación. ​Esta incertidumbre se debe al método de muestreo aleatorio. 

​Para el propósito de nuestro ejemplo, ​imaginemos que el peso medio de los ​10 000 pingüinos es de 31 libras, ​aunque no lo sabrías ​a menos que pesaras todos los pingüinos. 

​En la práctica, los profesionales de datos suelen seleccionar ​una muestra aleatoria porque el ​muestreo aleatorio repetido ​suele ser caro y lento. ​Como la muestra es aleatoria, es ​probable que la media de cualquier muestra determinada ​no sea igual a la media real de la población. ​Por ejemplo, es posible que peses ​una muestra de pingüinos que ​recientemente han tenido dificultades para encontrar comida. Solo pesan 28 libras, ​o puede pesar una muestra de pingüinos ​que se alimentaron recientemente de un buffet de pescado ​y pesan más de la media, 32 libras. ​

De cualquier manera, la estimación de la muestra no será ​igual a la media de la población de 31 libras. ​

Si solo proporcionas una estadística de muestra o una estimación puntual, ​no será tan precisa. Los ​intervalos de confianza brindan ​a los profesionales de datos una forma de expresar ​la incertidumbre causada por ​la aleatoriedad y proporcionan una estimación más confiable. 

### Confidence interval

​Junto con la estadística muestral, ​un intervalo de confianza incluye ​un margen de error y un nivel de confianza. 

​Exploremos nuestro ejemplo de pingüinos para tener ​una mejor idea de cada componente. ​Empezaremos con nuestra estadística de muestra. 


​La media muestral de nuestro grupo de pingüinos es de 30 libras. ​

A continuación, determinaremos el intervalo para nuestra estimación, ​que se define por la estadística de la muestra ​más o menos el margen de error. ​

**El margen de error** representa ​la diferencia máxima esperada entre ​un parámetro de población y una estimación de la muestra. ​En otras palabras, esta es ​la cantidad que un profesional de datos ​espera que su estimación ​varíe con respecto a la cantidad real. 

**Penguins interval**

​Si nuestro escalón de muestra para nuestros pingüinos es de ​30 libras y nuestro margen de error es de más o ​menos dos libras, significa que el límite inferior ​del intervalo es de 28 libras y ​el límite superior es de 32 libras. ​

El límite superior del intervalo es 30+2 = 32 libras. ​Este rango de valores expresa ​la incertidumbre en su estimación debido al muestreo aleatorio.



​El cálculo del **margen de error** implica ​multiplicar el error estándar por una puntuación z. 



​Recuerde que la puntuación z mide la distancia entre ​un punto de datos y la media de la población ​en una distribución normal estándar. ​Por lo general, usará una computadora para estos cálculos. ​Junto con una estadística muestral y un margen de error, ​un intervalo de confianza también incluye un nivel de confianza. ​

### Confidence level

El nivel de confianza describe ​la probabilidad de que un método de muestreo en particular ​produzca un intervalo de confianza ​que incluya el parámetro de población. 

​Por ejemplo, supongamos que utiliza un nivel de confianza del 95% para ​calcular un intervalo de confianza entre 28 y 32 libras. ​Técnicamente, esto significa que si ​se toman 100 muestras aleatorias de ​la población de pingüinos y se calcula ​un intervalo de confianza del 95% para cada muestra, ​aproximadamente 95 de los 100 intervalos, ​o el 95% del total, ​contendrán la media real de la población. ​Uno de esos intervalos será el rango ​de valores entre 28 y 32 libras. 




​Como profesional de datos, ​puede elegir su propio nivel de confianza ​en función de la precisión deseada de su estimación. ​Los niveles de confianza comunes son 90, 95 ​y 99%. El 95% es una opción popular. 



​Por ejemplo, la mayoría de las encuestas electorales ​muestran un nivel de confianza del 95%, ​y la mayoría de las pruebas A/B recomiendan ​usar un nivel de confianza del 95%. 

​Ten en cuenta que el 95% no tiene nada de mágico. ​Esta es una elección basada en la tradición y la ​investigación estadística y la educación. ​Puede ajustar el nivel de confianza para ​cumplir con los requisitos de su análisis. 

​Exploremos otro ejemplo. ​Imagina que eres un ​profesional de datos que trabaja para una empresa de moda. ​Su gerente le pide que calcule ​los ingresos por ventas de la nueva línea de ropa de primavera. 

​Cuando te reúnas con las partes interesadas, puedes decir: ​«Creo que venderemos 1 millón de dólares» o puedes decir: ​«Basándome en un nivel de confianza del 95%, ​calculo que nuestros ingresos por ventas oscilarán ​entre 950 000 y 1 050.000 dólares. 

​La primera afirmación ofrece una estimación puntual. 

​La segunda afirmación proporciona un nivel de confianza y ​una estimación de intervalo y ​comunica la incertidumbre de la estimación. 

​Brinda a las partes interesadas ​más información y les ayuda a ​tomar decisiones más informadas sobre ​cuestiones relacionadas con los ingresos por ventas futuros. ​Como profesional de datos, ​también debe asegurarse de que ​las partes interesadas entiendan sus resultados, ​por lo que es su trabajo ​comunicar claramente cómo interpretar un intervalo de confianza. ​Hablaremos más sobre la interpretación más adelante.
