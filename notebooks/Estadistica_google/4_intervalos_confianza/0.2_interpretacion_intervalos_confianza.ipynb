Recientemente, ha aprendido que los profesionales de los datos utilizan intervalos de confianza para describir la incertidumbre que rodea a una estimación. Para comprender mejor los datos y comunicar eficazmente los resultados a las partes interesadas, es importante saber interpretar correctamente un intervalo de confianza.

En esta lectura, revisaremos la forma correcta de interpretar un intervalo de confianza. También discutiremos algunas formas comunes de interpretación errónea y cómo evitarlas.

### Interpretación correcta

### Ejemplo: peso medio

Analicemos un ejemplo para comprender mejor cómo interpretar un intervalo de confianza. Imagina que quieres estimar el peso medio de una población de 10.000 pingüinos. En lugar de pesar todos y cada uno de los pingüinos, seleccionas una muestra de 100 pingüinos. El peso medio de la muestra es de 30 libras. Basándote en los datos de tu muestra, construyes un intervalo de confianza del 95% entre 28 libras y 32 libras.

iC 95 [28, 32]

### Interpreta el intervalo de confianza

Anteriormente, aprendiste que el Nivel de confianza expresa la incertidumbre del proceso de estimación. Analicemos lo que significa el 95% de confianza desde una perspectiva más técnica.

Técnicamente, el 95% de confianza significa que si toma muestras aleatorias repetidas de una población y construye un intervalo de confianza para cada muestra utilizando el mismo método, puede esperar que el 95% de estos intervalos capturen la media de la población. También se puede esperar que el 5% del total no capte la media de la población.

El nivel de confianza se refiere a la tasa de éxito a largo plazo del método, o del proceso de estimación basado en el muestreo aleatorio.

A efectos de nuestro ejemplo, imaginemos que el peso medio de los 10.000 pingüinos es de 31 libras, aunque no lo sabríamos a menos que pesáramos realmente a cada pingüino. Entonces, toma una muestra de la población.

Imagina que tomas 20 muestras aleatorias de 100 pingüinos cada una de la población de pingüinos y calculas un intervalo de confianza del 95% para cada muestra. Se puede esperar que aproximadamente 19 de los 20 intervalos, o el 95% del total, contengan el peso medio real de la población de 31 libras. Uno de estos intervalos será el rango de valores entre 28 libras y 32 libras.



En la práctica, los profesionales de los datos suelen seleccionar una muestra aleatoria y generar un intervalo de confianza, que puede contener o no la media real de la población. Esto se debe a que el muestreo aleatorio repetido suele ser difícil, caro y lento. Los intervalos de confianza ofrecen a los profesionales de los datos una forma de cuantificar la incertidumbre debida al muestreo aleatorio.

### Interpretaciones incorrectas

Ahora que ya sabe cómo interpretar correctamente un intervalo de confianza, repasemos algunas interpretaciones erróneas habituales y cómo evitarlas.

**Interpretación errónea 1: El 95% se refiere a la probabilidad de que la media de la población se encuentre dentro del intervalo construido.**

Una afirmación incorrecta que suele hacerse sobre un intervalo de confianza con un nivel de confianza del 95 % es que existe una probabilidad del 95 % de que la media de la población se encuentre dentro del intervalo construido.

En nuestro ejemplo, esto significaría que hay un 95% de probabilidad de que el peso medio de la población de pingüinos caiga en el intervalo entre 28 libras y 32 libras.

Esto es incorrecto. La media poblacional es una constante.

Como cualquier parámetro poblacional, la media poblacional es una constante, no una variable aleatoria. Mientras que el valor de la media muestral varía de una muestra a otra, el valor de la media poblacional no cambia. La probabilidad de que una constante se encuentre dentro de un intervalo de valores determinado es siempre del 0% o del 100%. O bien se encuentra dentro del intervalo de valores, o bien no.

Por ejemplo, cualquier muestra aleatoria de 100 pingüinos puede tener una media de peso diferente: 32,8 libras, 27,3 libras, 29,6 libras, etcétera. Puede utilizar una distribución de muestreo para asignar una probabilidad específica a cada una de las medias muestrales, ya que se trata de variables aleatorias. Sin embargo, el peso medio de la población se considera una constante. En nuestro ejemplo, si pesas los 10.000 pingüinos, verás que la media poblacional es de 31 libras. Este valor es fijo y no varía de una muestra a otra.



Por lo tanto, no es estrictamente correcto decir que hay un 95% de posibilidades de que el intervalo de confianza capture la media de la población, porque esto implica que la media de la población es variable. Los intervalos cambian de una muestra a otra, pero el valor de la media poblacional que se intenta capturar no cambia.

Lo que se puede decir es que si se toman muestras aleatorias repetidas de la población y se construye un intervalo de confianza para cada muestra utilizando el mismo método, se puede esperar que el 95% de los intervalos capturen la media poblacional.

**Consejo práctico:** Recuerde que un nivel de confianza del 95% se refiere a la tasa de éxito del proceso de estimación.

**Interpretación errónea 2: el 95% se refiere al porcentaje de valores de los datos que caen dentro del intervalo**


Otro error común es interpretar un intervalo de confianza del 95% como si el 95% de todos los valores de datos de la población estuvieran dentro del intervalo. Esto no es necesariamente cierto. Un intervalo de confianza del 95% muestra un rango de valores que probablemente incluye la media real de la población. No es lo mismo que un intervalo que contiene el 95% de los valores de datos de la población.

Por ejemplo, su intervalo de confianza del 95% para el peso medio de los pingüinos está entre 28 libras y 32 libras. Puede que no sea exacto decir que el 95% de todos los valores de peso caen dentro de este intervalo. Es posible que más del 5% de los pesos de los pingüinos de la población estén fuera de este intervalo, ya sea menos de 28 libras o más de 32 libras.



**Interpretación errónea 3: El 95% se refiere al porcentaje de medias muestrales que caen dentro del intervalo.**

Una tercera interpretación errónea habitual es que un intervalo de confianza del 95% implica que el 95% de todas las medias muestrales posibles se encuentran dentro del intervalo. Esto no es necesariamente cierto. Por ejemplo, su intervalo de confianza del 95% para el peso medio de los pingüinos está entre 28 libras y 32 libras. Imagina que tomas muestras repetidas de 100 pingüinos y calculas el peso medio de cada muestra. Es posible que más del 5% de las medias de las muestras sean inferiores a 28 libras o superiores a 32 libras.



### Puntos clave

Saber interpretar correctamente los intervalos de confianza le permitirá comprender mejor su estimación y le ayudará a compartir información útil y precisa con las partes interesadas. Es posible que también tenga que explicar las interpretaciones erróneas más comunes y por qué son incorrectas. No querrá que las partes interesadas basen sus decisiones en una interpretación errónea. Entender cómo comunicar eficazmente tus resultados a los interesados es una parte clave de tu trabajo como profesional de los datos.
