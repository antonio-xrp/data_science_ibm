### Estadística descriptiva vs estadisticas por inferencia

### métodos estadísticos
1. Descriptivos 
2. Inferenciales.

### Descriptive Statistics

Las estadísticas descriptivas describen o resumen las principales características de un conjunto de datos. 

- Las estadísticas descriptivas son muy útiles porque permiten comprender rápidamente una gran cantidad de datos.

- Por ejemplo, supongamos que tiene datos sobre la estatura de 10 millones de personas. Probablemente no desee escanear 10 millones de filas de datos para analizarlos para su informe.  Incluso si lo hiciera, sería difícil interpretar los datos. Sin embargo, si resume los datos, puede hacer que sean significativos al instante. Encontrar la altura media o promedio le brinda información útil sobre los datos, además, leer un resumen es mucho mejor que mirar millones de filas de datos.


### Hay dos formas comunes de estadísticas descriptivas: 

- Las imágenes, como gráficos y tablas
- las estadísticas resumidas. 

Anteriormente, aprendió cómo los gráficos y las tablas pueden ayudarlo a explorar , visualizar y compartir sus datos. Probablemente esté familiarizado con las visualizaciones de datos, como los histogramas, los gráficos de dispersión y los diagramas de caja.

> Las estadísticas resumidas le permiten resumir los datos con un solo número. 

Un ejemplo común es el valor medio o promedio.

### Hay dos tipos principales de estadísticas resumidas: 
- medidas de tendencia central 
- medidas de dispersión. 

### Las medidas de tendencia central,
- como la media, le permiten describir el centro de su conjunto de datos,

- las medidas de dispersión, como la desviación estándar, le permiten describir la dispersión de su conjunto de datos o la cantidad de variación en sus puntos de datos. 


### Inferential statistics 

Las estadísticas como la media y la desviación estándar se utilizan para describir y resumir los datos. Sin embargo, los profesionales de datos hacen más que solo describir sus datos.

También sacan conclusiones y hacen predicciones basadas en datos. Para ello, utilizan estadísticas inferenciales. 

- Las estadísticas inferenciales permiten a los profesionales de datos hacer inferencias sobre un conjunto de datos basándose en una muestra de los datos. 

- El conjunto de datos del que se extrae la muestra se denomina población.

- La población incluye todos los elementos posibles que le interese medir.

- una muestra es un subconjunto de una población. 

Por ejemplo: 
- una población puede ser el conjunto de todos los residentes de un país, todos los planetas de nuestro sistema solar o todos los resultados de 1000 lanzamientos de monedas.

- Una muestra es un grupo o subconjunto más pequeño de cualquiera de estas poblaciones. Las muestras pueden ser residentes , planetas o resultados del lanzamiento de monedas. 

Veamos un ejemplo. Supongamos que quieres investigar las preferencias musicales de todos los estudiantes universitarios de los Estados Unidos para averiguar si prefieren el pop, el rap, el country, la música clásica u otro género musical. 

Hay alrededor de 20 millones de estudiantes universitarios en los Estados Unidos y sería demasiado caro y lento recopilar datos de cada persona. En su lugar, puede usar una muestra y encuestar solo a un subconjunto de los 20 millones de estudiantes. 

Por ahora, imaginemos que decides encuestar a 1000 estudiantes en lugar de a 20 millones. Luego, puede usar los resultados para hacer inferencias sobre las preferencias musicales de todos los estudiantes universitarios. 

Tenga en cuenta que su muestra debe ser representativa de su población. De lo contrario, las conclusiones que extraiga de su muestra serán poco fiables y posiblemente sesgadas. 

### Representative sample

> Una muestra representativa es una muestra que refleja con precisión la población

por ejemplo, si solo encuestas a estudiantes de matemáticas o solo a estudiantes atletas, tu muestra no será representativa de todos los estudiantes universitarios. 

Finalmente, revisemos dos términos que corresponden a población y muestra, parámetro y estadística. 

### Parameter & Statistic 
- Un parámetro es una característica de una población. 

- Una estadística es una característica de una muestra. 

por ejemplo: 

- la altura promedio de toda la población de jirafas es un parámetro. 

- La altura promedio de una muestra aleatoria de 100 jirafas es una estadística. 

Como mencioné, es difícil recopilar datos sobre cada miembro de una población grande. En este caso, para localizar y medir la altura de cada jirafa del mundo. Usamos el valor conocido de una estadística muestral, el promedio de 100 jirafas, para estimar el valor desconocido de un parámetro de población.
