### Pruebas A/BEstado

Antes ha aprendido que las Pruebas A/B son una forma de comparar dos versiones de algo para averiguar qué versión funciona mejor. Por ejemplo, un profesional de datos puede utilizar las Pruebas A/B para comparar dos versiones de una página web o dos versiones de un anuncio en línea. También ha aprendido que las Pruebas A/B utilizan métodos estadísticos como el muestreo y la prueba de hipótesis.

En esta lectura, aprenderá más sobre el propósito general y el diseño de una prueba A/B y cómo las pruebas A/B utilizan métodos estadísticos para analizar datos.

### Contexto empresarial

Los profesionales de datos suelen utilizar las pruebas A/B para ayudar a las partes interesadas a elegir el mejor diseño para un sitio web o una aplicación con el fin de optimizar el marketing, aumentar los ingresos o mejorar la experiencia del cliente. En la práctica, las Pruebas A/B consisten en seleccionar aleatoriamente una muestra de usuarios y dividirlos en dos grupos (A y B). Los dos grupos visitan versiones diferentes del sitio web de una empresa. Las dos versiones son idénticas salvo por una única característica de diseño. Por ejemplo, el botón "Comprar" de la versión del grupo A puede tener un tamaño, una forma o un color distintos a los del botón "Comprar" de la versión del grupo B. Una prueba A/B utiliza el análisis estadístico para determinar si el cambio en la característica (por ejemplo, un botón más grande) afecta al comportamiento del usuario para una métrica específica. Un profesional de los datos podría utilizar una prueba A/B para analizar una de las siguientes métricas:

- Promedio de ingresos por usuario: ¿Cuántos ingresos genera un usuario para un sitio web?

- Duración promedio de la sesión: ¿Cuánto tiempo permanece un usuario en un sitio web?

- Tasa de clics: Si a un usuario se le muestra un anuncio, ¿hace clic en él?

- Porcentaje de conversiones: Si se muestra un anuncio a un usuario, ¿se convertirá en cliente?

Veamos un ejemplo para comprender mejor cómo funcionan las Pruebas A/B.

### Ejemplo Ingresos medios por usuario

Imagine que es un profesional de datos que trabaja para una empresa de venta de calzado por Internet. La empresa está tratando de hacer crecer su negocio y está investigando los ingresos medios por usuario en su sitio web. Su jefe de equipo le pide que realice una prueba A/B para determinar si el aumento del tamaño del botón "Comprar" tiene algún efecto sobre los ingresos medios. Usted selecciona al azar una muestra de usuarios y los divide en dos grupos, A y B. El grupo A visita la versión estándar del sitio web de la empresa. El grupo B visita una versión del sitio web idéntica a la versión estándar, excepto por el botón "Comprar", que es más grande. Usted realiza la prueba en línea y recoge los datos de la muestra. Los resultados indican que los ingresos medios por usuario son superiores para el grupo B. Por último, realiza una prueba de hipótesis de dos muestras para determinar si la diferencia observada en los ingresos medios es estadísticamente significativa o se debe al azar.




Una prueba A/B típica tiene al menos tres características principales:

1. Diseño de la prueba

2. Muestreo

3. Testeo de hipótesis

Examinemos cada característica con más detalle utilizando nuestro ejemplo.

### Diseño dela  prueba

En primer lugar, analicemos el diseño fundamental de una prueba A/B.

### Experimento controlado aleatorio

Una prueba A/B es una versión básica de lo que se conoce como experimento controlado aleatorio. En un experimento controlado aleatorio, los sujetos de la prueba se asignan aleatoriamente a un grupo de control y a un grupo de tratamiento. El tratamiento es el nuevo cambio que se prueba en el experimento. El grupo de control no está expuesto al tratamiento. El grupo de tratamiento está expuesto al tratamiento. La diferencia en los valores métricos entre los dos grupos mide el efecto del tratamiento en los sujetos de la prueba.

Nota: lo ideal es que la exposición al tratamiento sea la única diferencia significativa entre los dos grupos. Este diseño de prueba permite a los investigadores controlar otros factores que podrían influir en los resultados de la prueba y extraer conclusiones causales sobre el efecto del tratamiento.

En nuestro ejemplo, el grupo A es el grupo de control, el grupo B es el grupo de tratamiento y el tratamiento consiste en mostrar un botón "Comprar" más grande. Los usuarios del grupo de control (A) visitan la versión estándar del sitio web de la empresa. Los usuarios del grupo de tratamiento (B) visitan una versión alternativa con un botón "Comprar" más grande (es decir, están expuestos al tratamiento). Al hacer que las versiones del sitio web de A y B sean idénticas excepto por el tamaño del botón "Comprar", se minimiza la posibilidad de que cualquier diferencia observada en los ingresos medios se deba a otras características como el diseño de la página o el fondo. Esto le permite medir el efecto del botón más grande comparando la diferencia en los ingresos medios por usuario del grupo A y el grupo B.

La aleatorización, o asignación aleatoria de los sujetos de la prueba al grupo de control o al grupo de tratamiento, también ayuda a controlar el efecto potencial de otros factores sobre el resultado del experimento. En la práctica, hay muchos factores que pueden influir en que un usuario pulse o no el botón "Comprar". Por ejemplo, tal vez los usuarios súper ricos sean mucho más propensos a realizar compras en general, independientemente del tamaño del botón. Si el grupo de tratamiento está formado sólo por usuarios muy ricos, no obtendrá resultados válidos. Cualquier aumento observado en los ingresos medios podría deberse a la riqueza, no al mayor tamaño del botón "Comprar" (el factor que le interesa probar). La aleatorización ayuda a minimizar la posibilidad de que otros factores, como la riqueza, influyan significativamente en sus resultados medios.

### Muestreo

La selección aleatoria le ayuda a crear una muestra representativa que refleje las características de la población total de usuarios. En nuestro ejemplo, se trata de la población de clientes en línea de la empresa para la que trabaja. El uso de una muestra representativa para su prueba A/B le dará resultados válidos que son generalizables, o aplicables a la población en general.

También tendrá que elegir un tamaño de muestra que sea apropiado para su prueba A/B. Cuanto mayor sea el tamaño de la muestra, más precisos serán los resultados y más probabilidades tendrá de obtener resultados estadísticamente significativos cuando haya una diferencia entre el grupo A y el grupo B. Sin embargo, trabajar con muestras grandes puede resultar caro y llevar mucho tiempo. Los profesionales de los datos determinan el tamaño de la muestra en función tanto del objetivo del análisis como de su presupuesto disponible.

### Testeo de hipótesis

A efectos de nuestro ejemplo, supongamos que realiza la prueba en línea, recopila los datos y descubre que el grupo B tiene unos ingresos medios por usuario superiores a los del grupo A. Recordemos que el grupo B es el grupo de tratamiento (botón "Comprar" más grande) y el grupo A es el grupo de control. El siguiente paso es determinar si la diferencia observada en los datos es estadísticamente significativa o se debe al azar. Las Pruebas A/B utilizan pruebas de hipótesis de dos muestras para extraer conclusiones sobre la significación estadística. Para determinar si la diferencia observada en los ingresos medios por usuario es estadísticamente significativa, realice una prueba t de dos muestras. Formule sus hipótesis de la siguiente manera:

- H0: No hay diferencia en los ingresos medios por usuario entre A y B

- Ha: Existe una diferencia en los ingresos medios por usuario entre A y B.

### Resultados

Basándose en los resultados de la prueba t, rechaza la hipótesis nula y concluye que el aumento observado en los ingresos medios por usuario es estadísticamente significativo.

Los resultados de su prueba A/B le ayudan a decidir si recomienda o no un cambio de diseño para el sitio web de su empresa. En este caso, cuando presente los resultados a las partes interesadas de la empresa, sugiera implementar el botón "Comprar" más grande para aumentar los ingresos medios por usuario en el futuro.

### Puntos clave

Las Pruebas A/B son una de las aplicaciones más populares de la estadística para fines empresariales. Los profesionales de los datos utilizan las Pruebas A/B para ayudar a los líderes empresariales a optimizar el rendimiento de los productos, mejorar la experiencia del cliente y hacer crecer su negocio en línea. Comprender el propósito general y el diseño de una prueba A/B le será útil en su futura carrera como profesional de los datos.
