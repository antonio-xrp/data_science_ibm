### Introducción a las pruebas de hipótesis

### Hypothesis testing 

Prueba de hipótesis utiliza ​datos de muestra para evaluar ​una suposición sobre un parámetro de población. ​

Por ejemplo, en el caso de un ensayo clínico de un medicamento nuevo, ​una prueba de hipótesis puede ​ayudarte a determinar si el efecto del medicamento en el ​tiempo medio de recuperación de tu grupo de muestra es ​estadísticamente significativo o se debe al azar. 

### Steps for performing a hypothesis test

​- Primero, establezca la hipótesis nula ​y la hipótesis alternativa. 

- ​En segundo lugar, elige un nivel de significación. 

- ​En tercer lugar, encuentre el valor p ​y, en segundo lugar, rechace o no rechace la hipótesis nula. 


Imagina que te dan una moneda para usar en un juego. ​No estás seguro de si la moneda es justa o está manipulada. ​Es decir, no sabes si se ​trata de una moneda estándar o si se ha ​esperado específicamente para afectar al resultado de un lanzamiento, ​por ejemplo, para que siempre caiga en cruz. 

​Antes de usarla en el juego, ​debes averiguar si la moneda es justa o no. ​Decides probar la moneda lanzándola seis ​veces seguidas y registrando los resultados. ​Como recordarás de nuestro ​análisis anterior sobre la probabilidad, ​si la moneda es justa, ​la probabilidad de caer cara o cruz es del ​0,5 o el 50 por ciento en cualquier lanzamiento dado. 

​Si la moneda está manipulada en forma de cruz, ​la probabilidad de caer en la cola en ​cualquier lanzamiento será mucho mayor, ​quizás del 90 o incluso del 100 por ciento. ​Antes de probarlo, ​debe tener un punto de referencia para ​evaluar los resultados de la prueba. ​Por ejemplo, supongamos que los ​dos primeros lanzamientos caen en cruz, ​¿la moneda está manipulada? ​Recuerde que utilizamos la regla de la multiplicación para ​calcular la probabilidad de eventos independientes. 

**Probability of landing on tails for any given toss**

- Fair coin = 50%
- Rigged coin = 90% - 100%



​La probabilidad: 

-  de que la moneda caiga en la ​cola dos veces seguidas es = ​0,5  * 0,5 = 0,25 %. ​

Eso no es improbable. En este punto, ​no se puede concluir razonablemente que la moneda esté manipulada. 

​Ahora, imagina que la moneda cae en la ​cola cuatro veces seguidas. 

- ​La probabilidad de que esto ocurra es 0,5 multiplicado por ​cuatro, lo que equivale a 0,0625 o 6,25%. 

​Es poco probable, pero no imposible. ​Sin embargo, desea sentirse aún más ​seguro de que el resultado no se debe al azar. 

​Usted decide usar el cinco por ciento ​como umbral para determinar si el resultado se debe al azar. ​En otras palabras, si la probabilidad del resultado es ​inferior al cinco por ciento ​bajo el supuesto de que la moneda es justa, ​concluirá que la moneda está realmente manipulada. 

​Por ejemplo, 

- la probabilidad de que ​una moneda justa caiga seis veces seguidas ​es de 0,5 multiplicado por seis, lo que equivale a ​0,0156 o 1,56 por ciento. 

​Esto es demasiado improbable porque ​la probabilidad es ​inferior al umbral del cinco por ciento. ​Si esto ocurre, concluirá que la moneda está manipulada. 

​Ahora está listo para continuar con la prueba. ​Lanza la moneda seis ​veces seguidas y registra los resultados. ​La moneda cae en la cola cada vez que ​llegas a la conclusión de que la moneda está manipulada. ​Desafortunadamente, la moneda no te servirá ​de mucho a menos que estés ​realizando el truco de magia y ​necesites una moneda que siempre caiga en la cola. ​

Este ejemplo es una versión simplificada ​de una prueba de hipótesis. ​Para comprobar si la moneda es justa o no, ​ha realizado cada paso ​del procedimiento de prueba de hipótesis. ​

### Steps for conducting a hypothesis test

- ​Primero, establezca la hipótesis nula ​y la hipótesis alternativa. 
- ​En segundo lugar, elige un nivel de significación. 
- ​En tercer lugar, encuentra el valor p. 
- ​En adelante, rechace o deje de rechazar la hipótesis nula. 

### Null hypothesis 

​La hipótesis nula es una afirmación que se supone ​verdadera a menos que haya pruebas convincentes de lo contrario. ​La hipótesis nula suele ​suponer que los datos observados se producen por casualidad. 

### Alternative hypothesis

​La hipótesis alternativa es ​una afirmación que contradice la hipótesis nula ​y se acepta como verdadera ​solo si hay pruebas convincentes que la respalden. ​La hipótesis alternativa suele suponer que ​los datos observados no se producen por casualidad. 


​En nuestro ejemplo:

-  su hipótesis nula establece ​que la moneda es justa. ​Tener una moneda justa es ​el estado estándar o típico de las cosas. ​La hipótesis nula establece que ​sus observaciones son el resultado de pura casualidad. 

​- Su hipótesis alternativa afirma lo contrario, ​la moneda no es justa. ​La hipótesis alternativa dice que el resultado fue ​el resultado de una manipulación y no ocurrió por casualidad. ​A continuación, elige tu nivel de significación. 

### Significance level

​​El nivel de significancia también es la probabilidad de ​rechazar la hipótesis nula cuando es verdadera. ​

​En nuestro ejemplo:

- utilizas el 5% como ​umbral para determinar si ​el resultado del lanzamiento de la moneda se produjo por casualidad. ​

Por lo general, los profesionales de datos establecen ​el nivel de significancia en un cinco por ciento. ​Tenga en cuenta que el cinco por ciento no tiene nada de mágico, ​esta es una elección basada en la tradición ​y la investigación estadística en educación. ​Puede ajustar el nivel de significancia ​para cumplir con los requisitos de su análisis. ​Otras opciones comunes son el uno por ciento y el 10 por ciento.

### P-value

El valor ​P se refiere a la probabilidad de observar resultados iguales ​o más extremos que los ​observados cuando la hipótesis nula es verdadera. 

​Ya hemos calculado que ​la probabilidad de que una moneda justa caiga ​seis veces seguidas es del 1,56 por ciento. ​Si asumes que la hipótesis nula ​es verdadera y que la moneda es justa, ​entonces el valor p en nuestro ejemplo es del 1,56 por ciento.Cualquier valor inferior a eso significaría que hay ​pruebas más sólidas a favor de la hipótesis alternativa. ​Recuerde que su hipótesis alternativa ​es que la moneda no es justa.

​Por ejemplo, la probabilidad de lanzar siete colas ​seguidas con una moneda justa es del 0,79 por ciento, ​inferior al valor p del 1,56 por ciento. ​Si lanzas siete colas seguidas, ​tendrías pruebas aún más sólidas para ​la hipótesis alternativa de que la moneda no es justa. **​Por último, tienes que decidir si ​rechazas o no la hipótesis nula.**

**A lower p-value means there is stronger evidence for the alternative hypothesis**

​Los estadísticos siempre dicen que ​no rechazan en lugar de excluir. ​Esto se debe a que las pruebas de hipótesis ​se basan en la probabilidad, ​no en la certeza. ​La aceptación implica certeza. ​En general, como profesionales de los datos, ​tratamos de no pretender tener certeza sobre ​los resultados basados en métodos estadísticos. 

### Drawing a conclusion 

​Hay dos reglas principales para sacar ​una conclusión sobre una prueba de hipótesis.

- ​Si su valor p es inferior a su nivel de significancia, ​usted rechaza la hipótesis nula. ​
- Si su valor p es mayor que su nivel de significancia, ​no rechazará la hipótesis nula. 

​En el ejemplo del lanzamiento de la moneda,

- ​su valor p del 1,56 por ciento ​es inferior a su nivel de significancia del cinco por ciento. **​Rechazas la hipótesis nula ​y llegas a la conclusión de que tu resultado de ​seis colas consecutivas es ​estadísticamente significativo y no se debe al azar.** ​

Su decisión de rechazar o no ​rechazar también depende de su nivel de significación. ​Supongamos que, antes de la prueba, ​eligió un nivel de significancia del ​uno por ciento en lugar del cinco por ciento. 



​En ese caso, no rechazaría ​la hipótesis nula porque su valor p del ​1,56 por ciento sería ​mayor que su nivel de significancia del uno por ciento. ​Un resultado estadísticamente significativo no puede demostrar ​con un 100 por ciento de certeza ​que nuestra hipótesis es correcta. ​Como las pruebas de hipótesis se basan en ​la probabilidad, siempre existe ​la posibilidad de sacar una conclusión errónea sobre la hipótesis nula. 

### Types of errors in hypothesis testing

​En las pruebas de hipótesis, ​hay dos tipos de errores que se ​pueden cometer al llegar a una conclusión: 
- ​un error de tipo I y un error de tipo II. ​

- Un error de tipo I, ​también conocido como falso positivo, ​se produce cuando se rechaza ​una hipótesis nula que es realmente cierta. ​En otras palabras, usted concluye que su resultado es ​estadísticamente significativo cuando, de ​hecho, ocurrió por casualidad. ​En nuestro ejemplo, concluir ​que la moneda está manipulada cuando ​en realidad es justa se consideraría un error de tipo I. 

​Aunque tengas seis colas seguidas, ​este resultado podría deberse al azar. ​Es muy poco probable, pero es posible. ​Anteriormente en el vídeo, ​mencioné que su nivel de significancia también es ​la probabilidad de rechazar ​la hipótesis nula cuando es verdadera.

Un nivel de significancia del ​cinco por ciento significa que está dispuesto a aceptar ​un cinco por ciento de probabilidades de ​equivocarse cuando rechaza la hipótesis nula. 

​Para reducir la probabilidad de cometer un error de tipo I, ​elige un nivel de significancia más bajo. ​Recuerda que si eliges ​un nivel de significancia del uno por ciento, no ​rechazas el valor nulo ni ​llegas a la conclusión de que la moneda es justa. ​Sin embargo, 

- si eliges un nivel de significancia más bajo, ​es más probable que cometas un error de tipo II o un falso negativo. Esto ocurre cuando no se rechaza una hipótesis nula, ​que en realidad es falsa. ​En otras palabras, llegas a la conclusión de que tu resultado se produjo por ​casualidad cuando, de hecho, es estadísticamente significativo. ​En nuestro ejemplo, concluirías que ​la moneda es justa cuando en realidad está manipulada. 


​Como profesional de datos, ​es útil estar al tanto de los posibles errores incluidos ​en las pruebas de hipótesis y de ​cómo pueden afectar a los resultados. ​Según la situación y el objetivo de su análisis, ​es posible que desee minimizar el riesgo de ​un error de tipo I o de tipo II. ​Imagina que estás probando la resistencia de ​la tela para un fabricante de paracaídas. ​Debes estar muy seguro de que el material que estás ​utilizando es lo suficientemente fuerte como para un paracaídas funcional. 
​Un error de tipo I o un falso positivo significa ​que usted identifica erróneamente que el material es lo suficientemente fuerte. 

​Obviamente, en este caso, ​desea minimizar el riesgo de un error de tipo I. ​Para ello, elija un nivel de significancia del ​uno por ciento en lugar del cinco por ciento estándar. 

​Este cambio reduce la probabilidad de un error de tipo I ​o un falso positivo del 5 al 1 por ciento. ​En última instancia, es su responsabilidad ​como profesional de datos determinar la ​cantidad de evidencia que necesita para ​decidir que un resultado es estadísticamente significativo. ​¿Qué tan riesgoso es un error de tipo I o un falso positivo? ​No hay una única respuesta correcta ​para todas estas situaciones.
