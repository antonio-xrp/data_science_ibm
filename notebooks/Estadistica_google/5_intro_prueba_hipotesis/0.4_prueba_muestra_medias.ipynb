### Prueba de una muestra para las medias

Anteriormente mencioné que hay ​dos tipos diferentes de pruebas de hipótesis: ​una muestra y dos muestras. 

### One-sample test

​Una prueba de una muestra determina ​si un parámetro de la población, ​como una media o una proporción, ​es igual o no a un valor específico. 

### Two-sample test

​Una prueba de dos muestras determina ​si dos parámetros de la población, ​como dos medias o dos proporciones, ​son iguales entre sí.

### One-sample hypothesis test

​Un profesional de datos puede realizar ​una prueba de hipótesis de ​una muestra para determinar:

- si los ingresos medios por ventas de una empresa ​son iguales a un valor objetivo,
- la ​tasa media de ​éxito de un tratamiento médico es igual a un objetivo establecido ​
- la tasa de ​rendimiento media de una cartera de valores es igual a un índice de referencia del mercado. 


​Hay diferentes tipos de pruebas de hipótesis que ​puede usar según el tipo de ​datos de muestra con los que esté trabajando. 

### One-sample z-test assumptions 

​La prueba z de una muestra parte de las siguientes suposiciones: 

- ​los datos son una muestra aleatoria de ​una población con distribución normal, 

- ​la desviación estándar de la población es conocida. 

​Imagine que es un profesional de datos ​que trabaja para una empresa de entrega en línea. 

​Por lo general, el tiempo medio de entrega de ​un pedido de comida en línea ​es de 40 minutos con una desviación estándar de cinco minutos. ​Sin embargo, recientemente, la dirección de la empresa lanzó ​un nuevo programa de capacitación para hacer que ​el proceso de entrega sea más eficiente. 



​Una vez que los repartidores completaron el programa de capacitación, la ​gerencia rastreó una muestra aleatoria de ​50 entregas para comprender cuánto tarda una entrega. 
​La muestra de 50 partos tuvo un tiempo medio de entrega de ​38 minutos con una desviación estándar de cinco minutos. ​Hay una diferencia observada de ​dos minutos entre la media ​de la población de 40 minutos y la media de la muestra de 38 minutos.



​El equipo de administración le pide que ​determine si la disminución en el ​tiempo promedio de entrega es estadísticamente ​significativa o si se debe a la casualidad. ​Si la disminución es estadísticamente significativa, ​la empresa quiere invertir en el desarrollo e ​implementación del programa de capacitación en otras regiones. ​Decide realizar una prueba z de una muestra ​para analizar los datos. ​Repasemos los pasos para realizar una prueba de hipótesis. 

### Steps for conducting a hypothesis test

- ​Primero, establezca la hipótesis nula ​y la hipótesis alternativa. 
- ​En segundo lugar, elige un nivel de significación. 
- ​En tercer lugar, encuentra el valor p. ​
- En cuarto lugar, rechazar o dejar de rechazar la hipótesis nula. 

​Comience por establecer ​su hipótesis nula y su hipótesis alternativa. ​

### Null hypothesis

Recuerde que la hipótesis nula ​es una afirmación que se supone ​verdadera a menos que haya ​pruebas convincentes de lo contrario. ​

En una prueba z de una muestra, ​la hipótesis nula establece que ​la media de la población es igual a un valor observado. En este caso, la hipótesis nula dice ​que el tiempo de entrega promedio equivale a **40 minutos**, ​40 minutos es el tiempo de entrega promedio estándar. 


### Alternative hypothesis

​La hipótesis alternativa es ​una afirmación que contradice la hipótesis nula. 

​En una prueba de una muestra, ​hay tres opciones principales para ​la hipótesis alternativa: ​

- la media de la población no es igual ​, menor o mayor que un valor observado. En este caso, desea comprobar si la formación ​ha reducido el tiempo medio de entrega. ​Su hipótesis alternativa dice que ​el tiempo medio de entrega es inferior a 40 minutos. 

​A continuación, establezca el nivel de significancia o ​el umbral en el que considerará ​nuestros resultados estadísticamente significativos. ​Esta es también la probabilidad de ​rechazar la hipótesis nula cuando es verdadera.

​Usted elige un nivel de significancia del cinco por ciento, ​que es el estándar de la empresa para el análisis de datos. 

### P-value

​Ahora, encuentra el valor p. ​Recuerde que su valor p es ​la probabilidad de observar una diferencia en ​sus resultados igual o ​más extrema que la diferencia observada ​cuando la hipótesis nula es verdadera. 

​Normalmente, el tiempo medio de entrega es de 40 minutos. ​El tiempo medio de entrega de la muestra es de 38 minutos. ​Su hipótesis nula afirma ​que esta diferencia de dos minutos en el ​tiempo de entrega se debe al azar o a la variabilidad del muestreo. ​Su valor p es la probabilidad ​de observar una diferencia de ​dos minutos o más si la hipótesis nula es verdadera. 



​Si la probabilidad de este resultado es ​muy improbable, en particular, ​si su valor **p es inferior a ​su nivel de significancia del cinco por ciento**, ​rechazará la hipótesis nula. 

​Como profesional de datos, ​casi siempre calculará el valor p en su ordenador ​utilizando un lenguaje de programación como ​Python u otro software estadístico. ​Sin embargo, exploremos brevemente los conceptos involucrados ​en el cálculo para ​comprender mejor cómo funciona. 
​Ser capaz de usar el código para ​los cálculos es importante para tu futura carrera. ​Sin embargo, estar familiarizado con los conceptos en los que se ​basan los cálculos le ayudará a aplicar ​los métodos estadísticos a los problemas del lugar de trabajo. 

​El valor p se calcula a partir ​de lo que se denomina estadística de prueba.

### Test statistic

​En las pruebas de hipótesis, ​la estadística de prueba es un valor que muestra en qué medida ​los datos observados coinciden con ​la distribución esperada en la hipótesis nula, ​por lo que si se supone que la hipótesis nula es ​verdadera y que el tiempo medio de entrega es de 40 minutos, ​los datos de los tiempos de entrega ​siguen una distribución normal. 



​La estadística de prueba muestra dónde se ​ubicarán los datos observados (un tiempo medio de entrega de la muestra de 38 minutos) en esa distribución. ​Como estás realizando una prueba z, la ​estadística de la prueba es una puntuación z. 

### Z-score
​Recuerde que una puntuación z es ​una medida del número de desviaciones estándar por ​debajo o por encima de la media poblacional de un punto de datos. 

​Las puntuaciones Z le indican dónde se ​encuentran sus valores en una distribución normal.

​La siguiente fórmula le proporciona una estadística de prueba ​z basada en los datos de la muestra: ​«Z es igual a la barra X menos ​Mu dividido por el cociente de Sigma ​dividido por la raíz cuadrada de n». Donde X-bar es la media de la muestra, ​Mu es la media de la población, ​Sigma es la desviación estándar de la población ​y n es el tamaño de la muestra. ​Si introduce los números en ​la fórmula y realiza el cálculo, ​obtendrá una puntuación z de -2,82. 



​Veamos dónde se encuentra la puntuación z, ​menos 2.82, en la distribución. 



​Está muy a la izquierda, ​casi tres desviaciones estándar por debajo de la media. ​Para una distribución normal, ​la probabilidad de obtener ​un valor inferior a la puntuación z de ​-2,82 se calcula tomando ​el área bajo la curva a la izquierda de la puntuación z. 
​Esto se denomina prueba de la cola izquierda porque el ​valor p se encuentra en ​la cola izquierda de la distribución. ​El área bajo esta parte de ​la curva es la misma que tu valor p. 



​De nuevo, su valor p es la probabilidad de observar ​una estadística de prueba igual o ​más extrema que la observada ​cuando la hipótesis nula es verdadera. 

​Su hipótesis alternativa establece que ​el tiempo medio de entrega ​disminuyó en función de los datos de la muestra. 
​Por eso nos interesa la probabilidad de obtener un ​valor inferior a tu puntuación z de -2,82. 

​En un escenario de prueba diferente, la ​estadística de la prueba podría ser positiva de 2,45 ​y es posible que le interesen valores ​superiores a la puntuación z de 2,45. ​En ese caso, tu valor p ​estaría ubicado en la cola derecha de ​la distribución y ​realizarás una prueba a la derecha. ​Si calculas el valor p, ​verás que es 0.0023, ​por lo que tu valor p es 0.0023 o 0.23 por ciento. ​



Esto significa que hay una ​probabilidad del 0,23 por ciento de que la diferencia en el ​tiempo medio de entrega sea de 2 minutos o ​más si la hipótesis nula es cierta. ​En otras palabras, es muy ​poco probable que su diferencia se deba al azar. ​

### Draw a conclusion

Para llegar a una conclusión sobre la hipótesis nula, ​compare el valor p con el nivel de significancia. 

​Si tu valor p es inferior al nivel de significancia, ​llegas a la conclusión de que hay ​una diferencia estadísticamente significativa ​en el tiempo medio de entrega. ​En otras palabras, rechazas la hipótesis nula. ​

Si tu valor p es mayor que el nivel de significancia, ​llegas a la conclusión de que no hay ​una diferencia estadísticamente significativa ​en el tiempo medio de entrega. ​En otras palabras, no rechazas la hipótesis nula. ​



Su valor p de 0,0023, ​o 0,23 por ciento, es inferior ​al nivel de significancia de 0,05 o cinco por ciento, por ​lo que rechaza la hipótesis nula y concluye que ​hay una diferencia estadísticamente significativa ​en el tiempo medio de entrega.



​Sus resultados sugieren que ​el tiempo de entrega más rápido probablemente se ​deba a los efectos positivos de la capacitación. ​Su análisis ayudará a los líderes de la empresa ​a decidir si deben realizar una mayor inversión ​en el programa de capacitación en el futuro.
