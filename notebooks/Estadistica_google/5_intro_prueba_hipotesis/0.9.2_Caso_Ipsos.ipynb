Estudio de caso: Ipsos: Cómo una empresa de investigación de mercados utilizó las pruebas A/B para ayudar a los anunciantes a crear anuncios más eficacesEstado

Anteriormente, aprendió que las Pruebas A/B son una forma de comparar dos versiones de algo para averiguar qué versión funciona mejor. Por ejemplo, un profesional de datos puede utilizar las Pruebas A/B para comparar dos versiones de una página web o dos versiones de un anuncio en línea. También ha aprendido que las Pruebas A/B utilizan métodos estadísticos como el muestreo y las pruebas de hipótesis. Este estudio de caso describe cómo Ipsos utilizó las Pruebas A/B para comparar dos formatos diferentes de anuncios online: anuncios presentados en una narrativa secuenciada frente a un anuncio tradicional de 30 segundos presentado varias veces. Aprenderá cómo la investigación de mercado basada en datos revela importantes conocimientos sobre el impacto de los diferentes formatos de anuncios en la eficacia de una campaña publicitaria digital. 



### Antecedentes de la empresa

Ipsos es una empresa de servicios integrales de investigación de mercado. Fundada en Francia en 1975, Ipsos es ahora una compañía global con 18.000 empleados que operan en 90 países. Ipsos ofrece servicios de investigación en numerosos ámbitos de los sectores público y privado. Estos servicios incluyen la creación de marcas, la eficacia de la publicidad, el desarrollo de productos, la reputación, la experiencia del usuario y del cliente, y la gestión de la opinión pública, las elecciones y las crisis. Ipsos utiliza una combinación de fuentes de datos para sus investigaciones, desde la recopilación de datos primarios hasta la escucha social, la movilidad y las imágenes por satélite.



### Antecedentes del proyecto

El cliente de Ipsos para este proyecto era una empresa de medios de comunicación en línea que permite a los usuarios publicar sus propios contenidos de vídeo. La empresa de medios quería ayudar a sus propios clientes -los anunciantes de su plataforma- a crear los anuncios más eficaces. En concreto, querían averiguar si invertir en anuncios secuenciados aumentaría la probabilidad de que los espectadores recordaran un anuncio y compraran un producto. La secuenciación de anuncios en vídeo permite a los anunciantes mostrar los anuncios en un orden basado en las estructuras narrativas más atractivas y memorables. La empresa de medios de comunicación encargó a Ipsos una investigación para medir el impacto de cinco estructuras de secuencia diferentes en la elevación de la marca.

**Nota:** Para el propósito de este caso de éxito, nos centraremos en una sola estructura de secuencia: Tease, Amplify, Echo. Esta secuencia comienza con un anuncio corto para despertar la curiosidad del espectador (Tease); luego, pasa a un anuncio más largo con más información para asegurar el compromiso del espectador (Amplify); por último, termina con un anuncio más corto que recapitula la historia e incita a los espectadores a la acción (Echo).

### Marco del proyecto

Ipsos desarrolló la siguiente pregunta de investigación para guiar su proyecto: ¿Una secuencia de anuncios con la estructura Tease, Amplify, Echo aumenta el recuerdo del anuncio y la intención de compra en comparación con las visualizaciones repetidas de un anuncio tradicional de 30 segundos?

La hipótesis inicial de Ipsos era que los anuncios presentados en una narrativa secuenciada serían más eficaces que los visionados repetidos de un anuncio tradicional. Para probar esta hipótesis para estos dos enfoques publicitarios, Ipsos llevó a cabo una prueba A/B. La prueba A/B estableció un experimento para dos grupos de usuarios: a un grupo se le mostraron los anuncios Tease, Amplify, Echo y al otro grupo se le mostró varias veces un anuncio tradicional. En cada caso, los distintos formatos de anuncio se basaban en el mismo contenido de marca. A continuación se explican los detalles del proceso de prueba.

### Los retos

Al inicio del proyecto, Ipsos identificó dos retos principales. El primero consistía en diseñar correctamente las Pruebas A/B. El segundo reto consistía en crear los anuncios de prueba en el entorno de pruebas adecuado.

### Diseño de la prueba

La principal preocupación de Ipsos era que los resultados de la prueba A/B fueran generalizables, es decir, aplicables a toda la población de usuarios de la empresa de medios. En otras palabras, Ipsos quería hacer inferencias válidas sobre la población de usuarios más amplia basándose en la muestra más pequeña de participantes en la prueba. Para obtener resultados válidos de la prueba, Ipsos necesitaba hacer lo siguiente

1. Crear una muestra representativa de participantes en la prueba que reflejara la población general de usuarios de la empresa de medios.

2. Crear un entorno de pruebas online que reflejara el entorno online de la empresa de medios. Esto también implicaba crear anuncios de prueba de varias marcas para reproducir la diversidad de anuncios que aparecían en la plataforma de la empresa de medios.

### El enfoque

A pesar de estos retos, Ipsos llevó a cabo las Pruebas A/B y alcanzó sus objetivos de investigación. El enfoque exitoso de Ipsos para su proyecto incluyó los siguientes elementos:

- Equipo

- Muestreo

- Proceso de prueba

- Testeo de hipótesis

### Equipo

Para construir un equipo eficaz, Ipsos creó una operación interfuncional, incluyendo la producción de vídeo para crear anuncios de prueba basados en la estructura Tease, Amplify, Echo, y la tecnología para construir un entorno de pruebas en línea realista.

Para facilitar la colaboración entre los participantes en el proyecto, Ipsos esbozó un conjunto claro de normas de edición y organizó un sitio compartido para albergar los enlaces a los vídeos y las notas de edición. Esto permitió una rápida retroalimentación y ajuste a lo largo del proceso de desarrollo. Por último, Ipsos hizo que los gestores de proyectos del servicio de atención al cliente se hicieran cargo y supervisaran el diseño y el flujo de trabajo del proyecto de principio a fin.

### Muestreo

Ipsos utilizó la selección aleatoria de paneles de consumidores para generar una muestra representativa que reflejara con precisión las características de la población general de usuarios. Ipsos también se aseguró de que cada grupo de prueba incluyera la misma proporción de encuestados para categorías clave como edad y género. Además, Ipsos utilizó una muestra relativamente grande de 7.500 encuestados para obtener resultados más precisos.

### Proceso de prueba

Para obtener resultados válidos de las pruebas, Ipsos realizó Pruebas A/B en un entorno online en el que los encuestados utilizaban la plataforma de la empresa de medios como lo harían habitualmente en su vida cotidiana. Para reflejar la diversidad de anuncios en la plataforma, Ipsos desarrolló anuncios de prueba en 30 categorías de marcas, desde billetes de avión a comida rápida o detergente para la ropa.

El proceso de prueba se organizó de la siguiente manera:

Las encuestas se administraron en línea a través de los teléfonos inteligentes de los encuestados en noviembre y diciembre de 2018. Después de la selección inicial, los encuestados fueron llevados a una versión basada en navegador de la plataforma donde eran libres de buscar y ver videos como lo harían normalmente. Ipsos insertó dinámicamente anuncios de prueba al principio de los vídeos seleccionados por los encuestados en el entorno de pruebas en vivo. Después de la sesión de navegación, los encuestados completaron una encuesta para medir el impacto de la marca en el recuerdo del anuncio y la intención de compra.

### Testeo de hipótesis

Los datos de la encuesta indicaron que la secuencia de anuncios "Tease, Amplify, Echo" generó mayores niveles de recuerdo del anuncio e intención de compra entre los encuestados que las visualizaciones repetidas de un anuncio tradicional. Para determinar si los resultados observados eran estadísticamente significativos, Ipsos realizó una prueba t de dos muestras para cada categoría: una para el recuerdo del anuncio y otra para la intención de compra. Formularon sus hipótesis de la siguiente manera:

- H0: No hay diferencia en el recuerdo del anuncio entre los anuncios secuenciados y un anuncio tradicional repetido.

- Ha: Existe una diferencia en la intención de recordar el anuncio entre los anuncios secuenciados y un anuncio tradicional repetido.

- H0: No existe diferencia en la intención de compra entre los anuncios secuenciados y un anuncio tradicional repetido.

- Ha: Existe una diferencia en la intención de compra entre los anuncios secuenciados y un anuncio tradicional repetido.

Para ambas pruebas, Ipsos rechazó la hipótesis nula. Llegaron a la conclusión de que existen diferencias estadísticamente significativas y sustanciales en el recuerdo del anuncio y la intención de compra entre los anuncios secuenciados y un anuncio tradicional repetido.

### Los resultados

Los resultados de las Pruebas A/B indicaron que la secuenciación de anuncios funciona.

La secuencia de anuncios Tease, Amplify, Echo tuvo un impacto significativamente mayor en el recuerdo del anuncio y la intención de compra que las visualizaciones repetidas de un anuncio tradicional. Por ejemplo, en todas las categorías de productos, el 54% de los espectadores recordaron el anuncio después de haber sido expuestos a la secuencia "Tease, Amplify, Echo", en comparación con el 42% con un anuncio tradicional repetido. Además, el 30% de los espectadores expresaron su intención de compra tras la secuencia "Tease, Amplify, Echo", frente al 25% con un anuncio tradicional repetido.

En general, los resultados sugieren que los anunciantes deberían invertir en la secuenciación de anuncios en sus campañas digitales para aumentar el impacto de la marca.

### Conclusión

Este caso de éxito sobre las Pruebas A/B de Ipsos demuestra el poder de la investigación basada en datos para generar perspectivas de negocio clave. Los resultados de las Pruebas A/B muestran claramente cómo la secuenciación de anuncios aumenta el recuerdo del anuncio y la intención de compra en comparación con las visualizaciones repetidas de un anuncio tradicional. La investigación de Ipsos sobre los beneficios de la secuenciación de anuncios ayudó a la compañía de medios a mejorar la experiencia y el rendimiento de los anunciantes en su plataforma y añadió valor a la marca de la compañía de medios.
