### Diferencias entre las hipótesis nula y alternativaEstado: 

Recientemente, ha aprendido que la prueba de hipótesis utiliza datos de muestra para evaluar una hipótesis sobre un parámetro de población. Los profesionales de los datos realizan una prueba de hipótesis para decidir si las pruebas de sus datos de muestra apoyan la hipótesis nula o la hipótesis alternativa.

En esta lectura, repasaremos las principales diferencias entre la hipótesis nula y la hipótesis alternativa, y cómo formular cada hipótesis en diferentes escenarios.

### Hipótesis estadísticas

Repasemos los pasos para realizar una prueba de hipótesis:

1. Enuncie la hipótesis nula y la hipótesis alternativa.

2. Elija un nivel de significación.

3. Determinar el Valor P.

4. Rechazar o no rechazar la hipótesis nula.

El primer paso de cualquier prueba de hipótesis es plantear las hipótesis nula y alternativa. Las hipótesis nula y alternativa se excluyen mutuamente, lo que significa que no pueden ser ciertas al mismo tiempo.

La **Hipótesis nula** es una afirmación que se asume como cierta a menos que haya pruebas convincentes de lo contrario. La hipótesis nula suele suponer que no hay ningún efecto en la población y que los datos observados se producen por casualidad.

La **hipótesis alternativa** es una afirmación que contradice la hipótesis nula y sólo se acepta como cierta si existen pruebas convincentes de lo contrario. La hipótesis alternativa suele suponer que existe un efecto en la población y que los datos observados no se producen por casualidad.

**Nota:** Las hipótesis nula y alternativa son siempre afirmaciones sobre la población. Esto se debe a que el objetivo del Testeo de hipótesis es hacer inferencias sobre una población basándose en una muestra.

Por ejemplo, imagine que es un profesional de datos que trabaja para un concesionario de coches. La empresa implanta un nuevo programa de formación en ventas para sus empleados. Le piden que evalúe la eficacia del programa.

- **Tu hipótesis nula (H0):** el programa no tuvo ningún efecto sobre los ingresos por ventas.

- **Su hipótesis alternativa (Ha):** el programa aumentó los ingresos por ventas.

Analicemos cada hipótesis con más detalle.

### Hipótesis nula

La hipótesis nula tiene las siguientes características:

- En estadística, la hipótesis nula suele abreviarse como H sub cero (H0).

- Cuando se escribe en términos matemáticos, la hipótesis nula siempre incluye un símbolo de igualdad (normalmente =, pero a veces ≤ o ≥).

- Las hipótesis nulas suelen incluir frases como "ningún efecto", "ninguna diferencia", "ninguna relación" o "ningún cambio"."

### Hipótesis alternativa

La hipótesis alternativa tiene las siguientes características:

- En estadística, la hipótesis alternativa suele abreviarse como H sub a (Ha).

- Cuando se escribe en términos matemáticos, la hipótesis alternativa siempre incluye un símbolo de desigualdad (normalmente ≠, pero a veces < o >).

- Las hipótesis alternativas suelen incluir frases como "un efecto", "una diferencia", "una relación" o "un cambio."

### Ejemplos de hipótesis

Normalmente, la hipótesis nula representa el statu quo, o el estado actual de las cosas. La hipótesis nula supone que el statu quo no ha cambiado. La hipótesis alternativa sugiere una nueva posibilidad o una explicación diferente. Veamos algunos ejemplos para tener una mejor idea de cómo escribir las hipótesis nula y alternativa para diferentes escenarios:

### Ejemplo#1: Peso medio

Una empresa de alimentos ecológicos es famosa por su granola. La empresa afirma que cada bolsa que produce contiene 300 gramos de granola, ni más ni menos. Para comprobar esta afirmación, un experto en control de calidad mide el peso de una muestra aleatoria de 40 bolsas.

- H0: μ = 300 (el peso medio de todas las bolsas de granola producidas es igual a 300 gramos)

- Ha: μ ≠ 300 (el peso medio de todas las bolsas de granola producidas no es igual a 300 gramos)

### Ejemplo 2: Altura media

Supongamos que se supone que la altura media de una determinada especie de árbol es de 9 metros. Sin embargo, un ecologista afirma que la altura media real es superior a 9 metros. Para probar esta afirmación, el ecólogo mide la altura de una muestra aleatoria de 50 árboles.

- H0: μ ≤ 30 (la altura media de esta especie de árbol es igual o inferior a 30 pies)

- Ha: μ > 30 (la altura media de esta especie de árbol es superior a 30 pies)

### Ejemplo nº 3: Proporción de empleados

Una empresa afirma que al menos el 80% de sus empleados están satisfechos con su trabajo. Sin embargo, un investigador independiente cree que menos del 80% de los empleados están satisfechos con su trabajo. Para probar esta afirmación, el investigador encuesta a una muestra aleatoria de 100 empleados.

- H0: p ≥ 0,80 (la proporción de todos los empleados que están satisfechos con su trabajo es igual o superior al 80%)

- Ha: p < 0,80 (la proporción de todos los empleados que están satisfechos con su trabajo es inferior al 80%)
