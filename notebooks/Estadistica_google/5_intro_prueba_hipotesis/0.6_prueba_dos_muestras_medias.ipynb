### Pruebas de dos muestras: Medias

​Anteriormente, realizó una prueba de hipótesis de una muestra para analizar los datos ​sobre el tiempo medio de entrega de un servicio de entrega de alimentos en línea. ​Próximamente, exploraremos una prueba de dos muestras para determinar los medios. 

### Test types
​
- Mientras que una prueba de una muestra determina si la media de una población es ​igual a un valor específico, ​

- una prueba de dos muestras determina si las medias de dos poblaciones son iguales entre sí. 

​En el análisis de datos, las pruebas de dos muestras se utilizan con frecuencia para las pruebas A/B. ​En mi carrera como profesional de datos, he realizado una serie de pruebas A/B para ayudar a las ​empresas a mejorar sus negocios en línea. ​Normalmente utilizo una prueba t de dos muestras para analizar los datos. ​Por ejemplo, supongamos que una tienda minorista en línea está considerando cambiar la ​página de destino por su recompensa llamada miembros que son los clientes más leales. 

​La métrica que más le importa a la empresa ​es el tiempo promedio que los usuarios pasan en la página de destino por sesión. ​

Primero, configuraría un experimento para dos grupos de usuarios: el grupo A usa la ​página de destino predeterminada y el grupo B usa una versión rediseñada de la página de destino. ​

Luego usaría una prueba t para comparar el tiempo promedio empleado en cada página de destino ​y determinar si la diferencia entre las dos medias de la muestra es ​estadísticamente significativa. ​En otras palabras, si el grupo B pasa más tiempo en la página de destino que ​el grupo A, la prueba t ayudará a determinar si se debe al azar o ​al nuevo diseño de la página de destino. 

### Two-sample t-test for means assumptions

​En el análisis de datos, la prueba t de dos muestras es el enfoque estándar para ​comparar dos medias.

​La prueba t de dos muestras para determinar las medias hace las siguientes suposiciones. ​

- Las dos muestras son independientes entre sí. 

- ​Para cada muestra, ​los datos se extraen al azar de una población con distribución normal. 

- ​Se desconoce la desviación estándar de la población. 


​Por lo general, los profesionales de datos utilizan una prueba z cuando se ​conoce la desviación estándar de la población y utilizan una prueba t cuando la desviación estándar de la población es desconocida ​y debe estimarse a partir de los datos. ​En la práctica, la desviación estándar de la población suele ser desconocida porque es ​difícil obtener datos completos sobre poblaciones grandes. ​Por eso, los profesionales de datos utilizan la prueba t para aplicaciones prácticas. 

​Si bien la estadística de prueba para una prueba z es una puntuación Z, la estadística de prueba para ​una prueba t es una puntuación T. ​Y mientras que las puntuaciones Z se basan en la distribución normal estándar, las ​puntuaciones T se basan en la distribución t. 

​La gráfica de la distribución t tiene una forma de campana similar a la ​distribución normal estándar, pero ​la distribución t tiene colas más grandes que las de la distribución normal estándar. ​Las colas más grandes indican la mayor frecuencia de valores atípicos que vienen con un ​conjunto de datos pequeño. ​A medida que aumenta el tamaño de la muestra, ​la distribución t se acerca a la distribución normal. ​Para una prueba t, el estadístico de prueba sigue una distribución t bajo la hipótesis nula. ​Exploremos cómo realizar una prueba t de dos muestras.

### Steps for conducting a hypothesis test 

Repasemos los pasos para realizar una prueba de hipótesis. 

- ​En primer lugar, establezca la hipótesis nula y la hipótesis alternativa. 

- ​En segundo lugar, elige un nivel de significación. 

​- En tercer lugar, encuentra el valor p. 

- ​En cuarto lugar, rechazar o dejar de rechazar la hipótesis nula. 

​Primero, exponga su hipótesis nula y su hipótesis alternativa. ​En una prueba t de dos muestras, 

- la hipótesis nula establece que no hay diferencia ​entre las medias de las dos poblaciones. ​Se supone que esto es cierto a menos que haya pruebas convincentes de lo contrario. ​Por lo tanto, para su hipótesis nula, ​usted dice que no hay diferencia entre el tiempo medio empleado en la versión A y ​la versión B. 

- Su hipótesis alternativa afirmará lo contrario: ​hay una diferencia en el tiempo medio empleado en la versión A y la versión B.

​A continuación, establezca el nivel significativo o ​el umbral en el que considerará un resultado estadísticamente significativo. ​Esta es la probabilidad de rechazar la hipótesis nula cuando es cierta, se ​elige un nivel significativo del 5%, que es el estándar de la empresa para las Pruebas A/B.

​Ahora encuentra el valor p; en este caso, tu valor p es la probabilidad ​de observar una diferencia en las medias de la muestra igual o ​más extrema que la diferencia observada cuando la hipótesis nula es verdadera.

​Según los datos de muestra, la diferencia entre el tiempo medio empleado en la versión ​A y la versión B es de cinco segundos. ​Su hipótesis nula afirma que esta diferencia se debe al azar. ​Su valor p es la probabilidad de observar una diferencia absoluta en las ​medias de la muestra de cinco segundos o más, si la hipótesis nula es verdadera. ​Si la probabilidad de este resultado es muy improbable. 

​En particular, si su valor p es inferior a su nivel de significancia del 5% ​, rechazará la hipótesis nula. ​

Como profesional de datos, casi siempre calculará el valor p en su computadora ​utilizando un lenguaje de programación como Python u otro software estadístico. ​Más adelante usarás python para una prueba t de dos muestras.

### Draw a conclusion 

Para llegar a una conclusión, compare su valor p con el nivel de significancia.

-  ​Si su valor p es inferior al nivel de significancia, ​concluye que hay una diferencia estadísticamente significativa en ​las medias entre las dos versiones. ​En otras palabras, usted rechaza la hipótesis nula de que no hay diferencia en ​el tiempo medio empleado en la versión A y la versión B. ​

- Si su valor p es mayor que el nivel de significancia, concluye que ​no hay una diferencia estadísticamente significativa entre las dos versiones. En otras palabras, no rechaza la hipótesis nula de que ​no hay diferencia en el tiempo medio empleado en la versión A y la versión B. ​
