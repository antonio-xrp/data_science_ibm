### Steps for conducting a hypothesis test

​Repasemos los pasos para realizar una prueba de hipótesis.

- ​Primero, establezca la hipótesis nula ​y la hipótesis alternativa. 

- ​En segundo lugar, elige un nivel de significación.

-  ​En tercer lugar, encuentra el valor p.

- ​En cuarto lugar, rechazar o dejar de rechazar la hipótesis nula. ​

Primero, exponga su hipótesis nula ​y su hipótesis alternativa. 
​
En una prueba z de dos muestras, ​la hipótesis nula establece que ​no hay diferencia entre las proporciones de los dos grupos. ​Se supone que esto es cierto a menos que ​haya pruebas convincentes de lo contrario.

-  ​Para su hipótesis nula, ​usted afirma que no hay diferencia en la proporción de ​empleados satisfechos en Londres y Beijing. 

- ​Su hipótesis alternativa ​afirmará la afirmación opuesta. ​Para su hipótesis alternativa, ​usted dice que hay una diferencia en ​la proporción de empleados satisfechos ​en Londres y Beijing. ​

A continuación, defina el nivel de significancia o el umbral en el ​que considerará un resultado ​estadísticamente significativo. ​Es la probabilidad de rechazar ​la hipótesis nula cuando es verdadera. 
​
Usted elige un nivel de significación del cinco por ciento, ​que es el estándar de la empresa para las encuestas a los empleados. 

​Ahora, encuentra el valor p. ​En este caso, el valor p es ​la probabilidad de observar una diferencia en ​las proporciones de la muestra igual o ​más extrema que la diferencia ​observada cuando la hipótesis nula es verdadera. ​Según los datos de su muestra, ​la diferencia entre la proporción de ​empleados satisfechos en Londres y ​Beijing es de 10 puntos porcentuales. ​Su hipótesis nula afirma que ​esta diferencia se debe a una casualidad. El ​valor p es la probabilidad de ​observar una diferencia absoluta en las proporciones de la muestra, ​es decir, 10 puntos porcentuales o ​más si la hipótesis nula es verdadera. ​Si la probabilidad de este resultado es muy improbable. 

​En particular, si su valor p es ​inferior a su nivel de significancia del cinco por ​ciento, rechazará la hipótesis nula. ​

Como profesional de datos, ​casi siempre calculará el valor p en su ordenador ​utilizando un lenguaje de programación como ​Python u otro software estadístico. 

​Para encontrar el valor p, primero ​calcule el estadístico de prueba Z. ​Utilice la siguiente fórmula para calcular el ​estadístico Z en función de los datos de la muestra. ​Z es igual a p_1 hat menos p_2 hat dividido ​por la raíz cuadrada de p_0 hat multiplicado por paréntesis abiertos, ​uno menos p_0 hat cerrar paréntesis ​veces entre paréntesis abiertos, ​1 dividido por n_1 más 1 dividido por n_2 cerrar paréntesis, ​donde p_1 hat y p_2 hat son las ​proporciones muestrales del primer y segundo grupo, n_1 y ​n_2 son los tamaños de muestra del primer y segundo grupo, y p_1 0 Esa es la proporción agrupada. ​La proporción agrupada es un promedio ponderado ​de las proporciones de las dos muestras. ​Esto tiene una fórmula separada por la ​que no tiene que preocuparse ahora mismo. 



​Si introduce los números en ​la fórmula y realiza el cálculo, ​obtendrá una puntuación Z de 1,03. 

​Para una prueba z, ​la estadística de prueba sigue ​una distribución normal bajo la hipótesis nula. ​Recuerde que su ​hipótesis alternativa establece que hay ​una diferencia en la proporción de ​empleados satisfechos en Londres y Beijing. ​La diferencia observada es de 10 puntos porcentuales. 

​Si encuentra una diferencia estadísticamente significativa ​entre las dos proporciones, ​menor o mayor que ​la diferencia observada de 10 puntos porcentuales, ​rechazará la hipótesis nula. ​Como te interesan los valores en ambas direcciones, ​ya sean inferiores o superiores a la estadística de prueba, ​tu valor p es la probabilidad ​de obtener un valor inferior a la puntuación z de 1,03 ​negativo o superior a la puntuación z positiva de 1,03. ​Su valor p corresponde al área bajo la curva ​tanto en la cola izquierda como en ​la cola derecha de la distribución.



​Se trata de una prueba de doble filo. ​Si calculas el valor p, ​verás que es 0,3030 o 30,3 por ciento. 

​Esto significa que hay una ​probabilidad del 30,3 por ciento de que la diferencia absoluta entre ​la proporción de empleados satisfechos ​en Londres y Beijing ​sea del 10 por ciento o ​más si la hipótesis nula es cierta. 

### Draw a conclusion

​Para llegar a una conclusión, ​compare su valor p con el nivel de significancia. ​Si su valor p es inferior al nivel de significancia, ​concluye que hay ​una diferencia estadísticamente significativa ​en las proporciones entre los dos grupos. 

​En otras palabras, rechazas la hipótesis nula. ​Si su valor p es mayor que el nivel de significancia, ​concluye que no hay ​una diferencia estadísticamente significativa ​en las proporciones entre los dos grupos. 

​En otras palabras, no rechazas la hipótesis nula. ​

Su valor p de 0,3030, ​o 30,3 por ciento, es mayor que ​el nivel de significancia de 0,05, o cinco por ciento. ​Por lo tanto, no rechaza ​la hipótesis nula y concluye que no hay ​una diferencia estadísticamente significativa ​entre la proporción de ​empleados satisfechos en la oficina de Londres ​y la oficina de Beijing. ​
