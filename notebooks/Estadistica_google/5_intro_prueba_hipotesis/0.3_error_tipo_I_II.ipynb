### Errores de tipo I y IIEstado: Traducido automáticamente del Inglés

Antes ha aprendido que puede utilizar una prueba de hipótesis para determinar si los resultados son estadísticamente significativos o si se han producido por casualidad. Sin embargo, como las pruebas de hipótesis se basan en la probabilidad, siempre existe la posibilidad de llegar a una conclusión errónea sobre la hipótesis nula. En las pruebas de hipótesis, se pueden cometer dos tipos de errores al extraer una conclusión: un Error de primer tipo y un Error de segundo tipo.

En esta lectura, discutiremos la diferencia entre los errores de tipo I y de tipo II, y los riesgos que implica cometer cada error.

### Errores en la toma de decisiones estadísticas

Repasemos los pasos para realizar una prueba de hipótesis:

1. Enuncie la hipótesis nula y la hipótesis alternativa.

2. Elija un nivel de significación.

3. Hallar el Valor P.

4. Rechazar o no rechazar la hipótesis nula.

Cuando decida rechazar o no rechazar la hipótesis nula, hay cuatro resultados posibles: dos representan elecciones correctas y dos representan errores. Puede:

- Rechazar la hipótesis nula cuando en realidad es cierta(**Error de primer tipo**)

- Rechazar la hipótesis nula cuando en realidad es falsa (Correcto)

- No rechazar la hipótesis nula cuando en realidad es verdadera (Correcto)

- No rechazar la hipótesis nula cuando en realidad es falsa (**Error de segundo** tipo)



**Ejemplo: Ensayo clínico**


Analicemos un ejemplo para comprender mejor los errores de tipo I y II. Las pruebas de hipótesis se utilizan a menudo en ensayos clínicos para determinar si un nuevo medicamento produce mejores resultados en los pacientes. Imagine que es un profesional de los datos que trabaja para una empresa farmacéutica. La empresa inventa un nuevo medicamento para tratar el resfriado común. La empresa prueba una muestra aleatoria de 200 personas con síntomas de resfriado. Sin el medicamento, la persona típica experimenta síntomas de resfriado durante 7,5 días. El tiempo medio de recuperación de las personas que toman el medicamento es de 6,2 días.

Usted realiza una prueba de hipótesis para determinar si el efecto del medicamento sobre el tiempo de recuperación es estadísticamente significativo o se debe al azar.

En este CASO

- Su hipótesis nula (H0) es que el medicamento no tiene ningún efecto.

- Su hipótesis alternativa (Ha) es que el medicamento es eficaz.

**Error de primer tipo**

Un **error de tipo 1**, también conocido como falso positivo, se produce cuando se rechaza una hipótesis nula que en realidad es cierta. En otras palabras, se concluye que el resultado es estadísticamente significativo cuando en realidad se ha producido por casualidad.

Por ejemplo, en un ensayo clínico, si la hipótesis nula es cierta, significa que el medicamento no tiene ningún efecto. Si comete un error de tipo I y rechaza la hipótesis nula, llegará a la conclusión errónea de que el medicamento alivia los síntomas del resfriado cuando en realidad no es eficaz.

La probabilidad de cometer un error de primer tipo se denomina alfa (α). Su nivel de significación, o alfa (α), representa la probabilidad de cometer un error de tipo I. Normalmente, el nivel de significación se establece en 0,05, o el 5%. Un nivel de significación del 5% significa que está dispuesto a aceptar una probabilidad del 5% de estar equivocado cuando rechaza la hipótesis nula.

**Reduzca el riesgo**

Para reducir la posibilidad de cometer un error de primer tipo, elija un nivel de significación más bajo.

Por ejemplo, si desea minimizar el riesgo de un error de tipo I, puede elegir un nivel de significación del 1% en lugar del 5% estándar. Este cambio reduce la posibilidad de cometer un error de tipo I del 5% al 1%.



**Error de segundo tipo**

Sin embargo, reducir el riesgo de cometer un error de primer tipo significa que es más probable cometer un error de segundo tipo, o falso negativo. Un error de tipo II se produce cuando no se rechaza una hipótesis nula que en realidad es falsa. En otras palabras, se llega a la conclusión de que el resultado se produjo por casualidad, cuando en realidad no fue así.

Por ejemplo, en un estudio clínico, si la hipótesis nula es falsa, significa que el medicamento es eficaz. Si comete un error de tipo II y no rechaza la hipótesis nula, llegará a la conclusión errónea de que el medicamento es ineficaz cuando en realidad alivia los síntomas del resfriado.

La probabilidad de cometer un error de tipo II se denomina beta (β), y beta está relacionada con la potencia de una prueba de hipótesis (potencia = 1- β). La potencia se refiere a la probabilidad de que una prueba pueda detectar correctamente un efecto real cuando lo hay.

**Reducir el riesgo**

Puede reducir el riesgo de cometer un error de tipo II asegurándose de que su prueba tiene suficiente potencia. En el trabajo con datos, la potencia suele fijarse en 0,80 u 80%. Cuanto mayor sea la potencia estadística, menor será la probabilidad de cometer un error de tipo II. Para aumentar la potencia, puede aumentar el tamaño de la muestra o el nivel de significación.

**Nota:** Una discusión detallada del concepto de Poder estadístico está fuera del alcance de este curso. La potencia es algo sobre lo que aprenderá más a medida que avance en su carrera como profesional de los datos y aumente sus conocimientos de estadística.

**Riesgos potenciales de los errores de tipo I y II**

Como profesional de los datos, es importante ser consciente de los riesgos potenciales que conlleva cometer estos dos tipos de errores.

Un Error de primer tipo significa rechazar una hipótesis nula que en realidad es cierta. En general, cometer un error de tipo I suele llevar a aplicar cambios que son innecesarios e ineficaces, y que suponen una pérdida de tiempo y recursos valiosos.

Por ejemplo, si comete un Error de tipo I en su ensayo clínico, el nuevo medicamento se considerará eficaz aunque en realidad sea ineficaz. Basándose en esta conclusión incorrecta, se puede prescribir un medicamento ineficaz a un gran número de personas. Además, pueden rechazarse otras opciones de tratamiento en favor del nuevo medicamento.

Un error de tipo II significa no rechazar una hipótesis nula que en realidad es falsa. En general, cometer un error de segundo tipo puede hacer que se pierdan oportunidades de cambio positivo e innovación. La falta de innovación puede resultar costosa para las personas y las organizaciones.

Por ejemplo, si comete un error de tipo II en su ensayo clínico, el nuevo medicamento se considerará ineficaz aunque en realidad sea eficaz. Esto significa que un medicamento útil puede no llegar a un gran número de personas que podrían beneficiarse de él.

**Puntos clave**

Como profesional de los datos, resulta útil ser consciente de los posibles errores que conllevan las pruebas de hipótesis y de cómo pueden afectar a los resultados. Dependiendo de la situación específica, puede optar por minimizar el riesgo de un error de tipo I o de tipo II. En última instancia, es su responsabilidad como profesional de los datos determinar qué tipo de error es más arriesgado en función de los objetivos de su análisis.
