### Determinar si los datos tienen significación estadísticaEstado: 

Recientemente ha aprendido que la significación estadística es la afirmación de que los resultados de una prueba o experimento no pueden explicarse únicamente por el azar. Una prueba de hipótesis puede ayudarle a determinar si los datos observados son estadísticamente significativos o si se deben al azar. Por ejemplo, en un ensayo clínico de un nuevo medicamento, una prueba de hipótesis puede ayudar a determinar si el efecto positivo del medicamento en un grupo de muestra es estadísticamente significativo o se debe al azar.

En esta lectura, aprenderá más sobre el concepto de significación estadística y su papel en la prueba de hipótesis.

### Significación estadística en el Testeo de hipótesis

Los profesionales de los datos utilizan las pruebas de hipótesis para determinar si una relación entre variables o una diferencia entre grupos es estadísticamente significativa.

Veamos un ejemplo para comprender mejor el papel de la significación estadística en la comprobación de hipótesis.

### Ejemplo: Duración media de la batería
Repasemos los pasos para realizar una prueba de hipótesis:

1. Enuncie la hipótesis nula y la hipótesis alternativa.

2. Elija un nivel de significación.

3. Hallar el Valor P.

4. Rechazar o no rechazar la hipótesis nula.

Imagina que eres un profesional de los datos que trabaja para una empresa de informática. La empresa afirma que la duración media de la batería de su portátil más vendido es de 8,5 horas, con una desviación típica de 0,5 horas. Recientemente, el equipo de ingeniería ha rediseñado el portátil para aumentar la duración de la batería. El equipo toma una muestra aleatoria de 40 portátiles rediseñados. La media de la muestra es de 8,7 horas.

El equipo le pide que determine si el aumento de la duración media de la batería es estadísticamente significativo o si se debe al azar. Decides realizar una prueba z para averiguarlo.

### Paso 1: Plantee la hipótesis nula y la hipótesis alternativa

La hipótesis nula suele suponer que los datos observados se producen por casualidad y no son estadísticamente significativos. En este caso, la hipótesis nula dice que no hay ningún efecto real sobre la duración media de la batería en la población de ordenadores portátiles.

La hipótesis alternativa suele suponer que los datos observados no se producen por casualidad y son estadísticamente significativos. En este caso, su hipótesis alternativa dice que hay un efecto sobre la duración media de la batería en la población de ordenadores portátiles.

En este ejemplo, formule las siguientes hipótesis:

- H0: μ = 8,5 (la duración media de la batería de todos los portátiles rediseñados es igual a 8,5 horas)

- Ha: μ > 8,5 (la duración media de la batería de todos los portátiles rediseñados es superior a 8,5 horas)

### Paso 2: Elija un nivel de significación

El nivel de significación, o alfa (α), es el umbral a partir del cual un resultado se considera estadísticamente significativo. El nivel de significación es también la probabilidad de rechazar la hipótesis nula cuando es cierta.

Normalmente, los profesionales de los datos fijan el nivel de significación en 0,05, o el 5%. Esto significa que unos resultados al menos tan extremos como los suyos sólo tienen un 5% de posibilidades (o menos) de producirse cuando la hipótesis nula es cierta.

**Nota:** el 5% es una elección convencional, no un número mágico. Se basa en la tradición de la investigación y la enseñanza estadística. Otras opciones habituales son el 1% y el 10%. Puede ajustar el nivel de significación para satisfacer los requisitos específicos de su análisis. Un nivel de significación más bajo significa que un efecto tiene que ser mayor para considerarse estadísticamente significativo.

**Consejo profesional:** Como práctica recomendada, debe establecer un nivel de significación antes de comenzar la prueba. De lo contrario, podría encontrarse en una situación en la que esté manipulando los resultados a su conveniencia.

En este ejemplo, se elige un nivel de significación del 5%, que es el estándar de investigación de la empresa.

### Paso 3: Encontrar el Valor P

**El Valor P** se refiere a la probabilidad de observar resultados tan o más extremos que los observados cuando la hipótesis nula es cierta.

El Valor P ayuda a determinar si un resultado es estadísticamente significativo. Un Valor P bajo indica alta significación estadística, mientras que un Valor P alto indica baja o nula significación estadística.

Cada prueba de hipótesis tiene sus características:

- Un estadístico de prueba que indica en qué medida sus datos coinciden con la hipótesis nula. Para una prueba z, su estadístico de prueba es una puntuación z; para una prueba t, es una puntuación t.

- Valor P correspondiente que le indica la probabilidad de obtener un resultado al menos tan extremo como el resultado observado si la hipótesis nula es cierta.

Como profesional de los datos, casi siempre calculará el Valor P en su ordenador, utilizando un lenguaje de programación como Python u otro software estadístico. En este ejemplo, está realizando una prueba z, por lo que su estadístico de prueba es una puntuación z de 2,53. Sobre la base de esta estadística de prueba, se calcula un Valor P de 0,0057, o 0,57%.

### Paso 4: Rechazar o no rechazar la hipótesis nula

En una prueba de hipótesis, se compara el Valor P con el nivel de significación para decidir si los resultados son estadísticamente significativos.

Hay dos reglas principales para llegar a una conclusión sobre una prueba de hipótesis:

- Si el Valor P es inferior al nivel de significación, se rechaza la hipótesis nula.

- Si el Valor P es mayor que el nivel de significación, no se rechaza la hipótesis nula.

**Nota:** Los profesionales de los datos y los estadísticos siempre dicen "no rechazar" en lugar de "aceptar" Esto se debe a que las pruebas de hipótesis se basan en la probabilidad, no en la certeza: la aceptación implica certeza. En general, los profesionales de los datos evitan afirmar que los resultados basados en métodos estadísticos son seguros.

En este ejemplo, el Valor P del 0,57% es inferior al nivel de significación del 5%. Su prueba proporciona pruebas suficientes para concluir que la duración media de la batería de todos los portátiles rediseñados ha aumentado de 8,5 horas. Rechaza la hipótesis nula. Determina que los resultados son estadísticamente significativos.

### Puntos clave

Como profesional de los datos, es importante comprender el concepto de significación estadística para realizar con eficacia una prueba de hipótesis e interpretar los resultados. La información basada en resultados estadísticamente significativos puede ayudar a las partes interesadas a tomar decisiones empresariales más informadas.
