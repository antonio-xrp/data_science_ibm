### Pruebas de una y dos colasEstado

Antes ha aprendido que una prueba de hipótesis puede ser de una o de dos colas. Una cola en las pruebas de hipótesis se refiere a la cola en cualquiera de los extremos de una curva de distribución.

En esta lectura, repasaremos las principales diferencias entre las pruebas de una cola y de dos colas, y discutiremos el procedimiento para realizar cada prueba.

### Pruebas de una y dos colas

En primer lugar, analicemos las diferencias entre las pruebas de una y dos colas.

Una prueba de una cola resulta cuando la hipótesis alternativa establece que el valor real de un parámetro de la población es menor o mayor que el valor de la hipótesis nula.

Una prueba de una cola puede ser de cola izquierda o de cola derecha. Una prueba de cola izquierda resulta cuando la hipótesis alternativa establece que el valor real del parámetro es menor que el valor de la hipótesis nula. Una prueba de colas derechas resulta cuando la hipótesis alternativa establece que el valor real del parámetro es mayor que el valor de la hipótesis nula.

Una prueba bilateral resulta cuando la hipótesis alternativa establece que el valor real del parámetro no es igual al valor de la hipótesis nula.

Por ejemplo, imagine una prueba en la que la hipótesis nula afirma que el peso medio de una población de pingüinos es igual a 30 libras.

- En una prueba de colas izquierdas, la hipótesis alternativa podría afirmar que el peso medio de la población de pingüinos es inferior a ("<") 30 lbs.

- En una prueba de colas a la derecha, la hipótesis alternativa podría afirmar que el peso medio de la población de pingüinos es mayor que (">") 30 lbs.

- En una prueba de dos colas, la hipótesis alternativa podría afirmar que el peso medio de la población de pingüinos no es igual ("≠") a 30 lbs.

Veamos un ejemplo más detallado para comprender mejor la diferencia entre las pruebas de una y dos colas.

### Ejemplo: Pruebas de una cola

Imagine que es un profesional de datos que trabaja para una empresa de venta al por menor por Internet. La empresa afirma que al menos el 80% de sus clientes están satisfechos con su experiencia de compra. Usted encuesta a una muestra aleatoria de 100 clientes. Según la encuesta, el 73% de los clientes dicen estar satisfechos. Basándose en los datos de la encuesta, realice una prueba z para evaluar la afirmación de que al menos el 80% de los clientes están satisfechos.

Repasemos los pasos para realizar una prueba de hipótesis:

1. Enuncie la hipótesis nula y la hipótesis alternativa.

2. Elija un nivel de significación.

3. Hallar el Valor P.

4. Rechazar o no rechazar la hipótesis nula.

En primer lugar, plantee las hipótesis nula y alternativa:

- H0: P >= 0,80 (la proporción de clientes satisfechos es mayor o igual al 80%)

- Ha: P < 0,80 (la proporción de clientes satisfechos es inferior al 80%)

**Nota**: se trata de una prueba de una cola, ya que la hipótesis alternativa contiene el signo menos ("<").

A continuación, elija un nivel de significación de 0,05, o 5%.

A continuación, calcule el Valor P basándose en el estadístico de la prueba. Recuerde que el Valor P es la probabilidad de observar resultados tan o más extremos que los observados cuando la hipótesis nula es cierta. En el contexto de las pruebas de hipótesis, "extremo" significa extremo en la(s) dirección(es) de la hipótesis alternativa.

Su estadístico de prueba es una puntuación z de 1,75 y su Valor P es 0,04.

Dado que se trata de una prueba de cola izquierda, el Valor P es la probabilidad de que la puntuación z esté a menos de 1,75 unidades estándar de la media hacia la izquierda. En otras palabras, es la probabilidad de que la puntuación z sea inferior a -1,75. La probabilidad de obtener un valor inferior a su puntuación z de -1,75 se calcula tomando el área bajo la curva de distribución a la izquierda de la puntuación z. Esto se llama una prueba de cola izquierda, porque su Valor P se encuentra en la cola izquierda de la distribución. El área bajo esta parte de la curva es igual a su Valor P: 0,04.



Por último, saca una conclusión. Dado que su Valor P de 0,04 es inferior a su nivel de significación de 0,05, rechaza la hipótesis nula.

**Nota:** En un escenario de prueba diferente, su estadístico de prueba podría ser positivo 1,75, y usted podría estar interesado en valores tan grandes o mayores que la puntuación z 1,75. En ese caso, su valor p es igual o mayor que el valor z. En ese caso, su Valor P se situaría en la cola derecha de la distribución, y estaría realizando una prueba de cola derecha.



**Ejemplo: Pruebas bilaterales**

Imaginemos que nuestro ejemplo anterior tiene una configuración ligeramente diferente. Supongamos que la empresa afirma que el 80% de sus clientes están satisfechos con su experiencia de compra. Para comprobar esta afirmación, realiza una encuesta a una muestra aleatoria de 100 clientes. Según la encuesta, el 73% de los clientes dicen estar satisfechos. Basándose en los datos de la encuesta, realice una prueba z para evaluar la afirmación de que el 80% de los clientes están satisfechos.

En primer lugar, plantee las hipótesis nula y alternativa:

- H0: P = 0,80 (la proporción de clientes satisfechos es igual al 80%)

- Ha: P ≠ 0,80 (la proporción de clientes satisfechos no es igual al 80%)

**Nota:** se trata de una prueba de dos colas, ya que la hipótesis alternativa contiene el signo de no igual ("≠").

A continuación, elija un nivel de significación de 0,05, es decir, del 5%.

A continuación, calcule el Valor P basándose en el estadístico de la prueba. Su estadístico de prueba es una puntuación z de 1,75. Como se trata de una prueba bilateral, el Valor P es la probabilidad de que la puntuación z sea menor que -1,75 o mayor que 1,75. Tenga en cuenta que el Valor P para una Prueba bilateral es siempre dos veces el Valor P para una prueba de una cola. Por lo tanto, en este caso, su Valor P = 0,04 + 0,04 = 0,08. En una prueba de dos colas, el Valor P corresponde al área bajo la curva tanto en la cola izquierda como en la cola derecha de la distribución.




Por último, saca una conclusión. Como su Valor P de 0,08 es mayor que su nivel de significación de 0,05, no rechaza la hipótesis nula.

### Una cola frente a dos colas

Puede utilizar pruebas de una y dos colas para examinar diferentes efectos.

En general, una prueba de una cola puede proporcionar más potencia para detectar un efecto en una sola dirección. Sin embargo, antes de realizar una prueba de una cola, debe considerar las consecuencias de no detectar un efecto en la otra dirección. Por ejemplo, imagine que una empresa farmacéutica desarrolla un nuevo medicamento que considera más eficaz que otro ya existente. Como profesional de los datos que analiza los resultados del ensayo clínico, es posible que desee elegir una prueba de una cola para maximizar su capacidad de detectar la mejora. Al hacer esto, no comprueba la posibilidad de que el nuevo medicamento sea menos eficaz que el existente. Y, por supuesto, la empresa no quiere dar a conocer al público un medicamento menos eficaz.

Una prueba de una cola puede ser adecuada si las consecuencias negativas de pasar por alto un efecto en la dirección no probada son mínimas. Por ejemplo, imagine que la empresa desarrolla un nuevo medicamento menos caro que cree que es al menos tan eficaz como el medicamento existente. El precio más bajo da al nuevo medicamento una ventaja en el mercado. Por tanto, sólo quieren asegurarse de que el nuevo medicamento no es menos eficaz que el existente. Comprobar si es más eficaz no es una prioridad. En este caso, una prueba de una cola puede ser apropiada.

### Puntos clave

Comprender las diferencias entre una prueba de una cola y una de dos colas es una parte importante de la realización de una prueba de hipótesis. Dependiendo del contexto de su análisis, es posible que desee utilizar una prueba de una cola para examinar los efectos en una sola dirección, o una prueba de dos colas para examinar los efectos en ambas direcciones.
