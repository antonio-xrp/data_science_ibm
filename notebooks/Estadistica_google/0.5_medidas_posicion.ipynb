### Percentiles

- Un percentil es el valor por debajo del cual caen los datos porcentuales.

- Los percentiles muestran la posicin relativa o el rango de un valor determinado en un conjunto de datos. 

Algunas universidades piden a los solicitantes que realicen exmenes estandarizados. Por ejemplo, en los Estados Unidos, el SAT y el ACT son exmenes comunes. Cuando un estudiante recibe la calificacin de su examen, por lo general tambin recibe el percentil correspondiente. 

Por ejemplo, supongamos que la puntuacin de una prueba cae en el percentil 99. Esto significa que el puntaje es superior al 99% de todos los puntajes de las pruebas. Si un puntaje cae en el percentil 75, el puntaje es superior al 75% de todos los puntajes de las pruebas.

Si la puntuacin cae en el percentil 50. El puntaje es superior a la mitad o el 50% de todos los puntajes de las pruebas, etc.



- Los percentiles son tiles para comparar valores. 

- Por ejemplo, los distintos exmenes pueden tener diferentes sistemas de puntuacin. Los puntajes del SAT oscilan entre 400 y 1600, los puntajes del ACT oscilan entre 1 y 36 y un examen escolar tpico de matemticas o historia puede oscilar entre 0 y 100. Si solo conoce los puntajes brutos de cada examen, digamos 1000 para el SAT, 20 para el ACT y 70 para el examen escolar, no tiene forma de hacer una comparacin significativa. Si sabe que los puntajes de los tres exmenes se encuentran en el percentil 50, puede comparar de manera significativa el desempeo de los estudiantes en los diferentes exmenes. 



### Quartile

Puede usar cuartiles para obtener una comprensin general de la posicin relativa de los valores. 

- Un cuartil divide los valores de un conjunto de datos en cuatro partes iguales. 
- Los cuartiles permiten comparar valores en relacin con los cuatro trimestres de los datos. 
Cada trimestre incluye el 25% de los valores de su conjunto de datos.

- El primer cuartil es el valor medio de la primera mitad del conjunto de datos. El primer cuartil, Q1, tambin se denomina cuartil inferior, el 25% de los puntos de datos estn por debajo del Q1 y el 75% por encima de l.

- El segundo cuartil es la mediana del conjunto. El segundo trimestre es la mediana, el 50% de los puntos de datos estn por debajo del segundo trimestre y el 50% por encima de l.

- El tercer cuartil es el valor medio de la segunda mitad del conjunto de datos, el 75% de los puntos de datos estn por debajo del tercer trimestre y el 25% por encima.




Tenga en cuenta la relacin entre los cuartiles y los percentiles.

- Q1 se refiere al percentil 25,

- Q2 al percentil 50 y 

- Q3 al percentil 75. 

Por ejemplo, supongamos que eres el director de un equipo deportivo. Tienes datos que muestran cuntos goles marc cada jugador de tu equipo a lo largo de toda una temporada. - En tercer lugar, encuentra la mediana de la mitad inferior de tu conjunto de datos. Este es el cuartil inferior, Q1, Q1 es igual a 13.

- Por ltimo, encuentra la mediana de la mitad superior de tu conjunto de datos. Este es el cuartil superior, Q3, Q3 es igual a 25.

### Interquartile Range (IQR)

- El 50% medio de los datos se denomina rango intercuartlico o IQR. 

- El rango intercuartil es la distancia entre el primer cuartil, Q1, y el tercer cuartil, Q3. 

> Tcnicamente, el IQR es una medida de dispersin porque mide la dispersin o el 50% medio de los datos. Es la misma que la distancia entre los percentiles 25 y 75, o entre el primer y el tercer trimestre. 

- El IQR tambin es til para determinar la posicin relativa de los valores de los datos. 

IQR es igual a Q3 menos Q1. En este caso, Q3 es igual a 25 y Q1 es igual a 13, por lo que IQR es igual a 12.
