Es un algoritmo de reducción de dimensionalidad lineal:
- Assume que las características del conjunto de datos están correlacionadas linealmente.
- Simplifica los datos, reduce la dimensionalidad, reduce el ruidi y a la vez que minimiza la pérdida de información
- Puede transformar características en un nuevo conjunto de variables no correlacionadas denominadas componentes principales y, al mismo tiempo, conservar la mayor varianza posible.

### Algoritmos de reducción dimensional

Los algoritmos de reducción de dimensionalidad reducen la cantidad de características del conjunto de datos sin sacrificar la información crítica del conjunto de datos. Los datos de alta dimensión suelen ser muy difíciles de analizar y visualizar. 

Los algoritmos de reducción de dimensionalidad simplifican el conjunto de datos para los modelos de aprendizaje automático. 

Los algoritmos de análisis de componentes principales (o PCA), incrustación de vecinos estocásticos distribuidos en T (o t-SNE) y aproximación y proyección de variedades uniformes (o UMAP) transforman las dimensiones originales para crear nuevas características. 



### Principle Component Analysis (PCA)

El análisis de componentes principales, o PCA, es un algoritmo de reducción de dimensionalidad lineal que asume que las características del conjunto de datos están correlacionadas linealmente. 

Simplifica los datos, reduce la dimensionalidad y reduce el ruido, a la vez que minimiza la pérdida de información. 

El PCA puede transformar las características en un nuevo conjunto de variables no correlacionadas denominadas componentes principales y, al mismo tiempo, conservar la mayor varianza posible. 



### Principle Component Analysis (PCA)

Estos componentes principales son ortogonales entre sí y definen un nuevo sistema de coordenadas para el espacio de características. 

Los componentes principales se organizan en orden decreciente de importancia, es decir, en qué medida explican la varianza del espacio entre características.

Los primeros componentes suelen contener la mayor parte de la información, mientras que el resto tiende a representar ruido. 



### T-distributed stochastic neighbor embedding (t-SNE)

La incrustación de vecinos estocásticos distribuidos en T, o t-SNE, mapea puntos de datos de alta dimensión a un espacio de menor dimensión. 

Es bueno para encontrar clústeres en datos complejos y de alta dimensión que se pueden visualizar en dos o tres dimensiones y funciona bien con datos como imágenes y texto. 

t-SNE se centra en preservar la similitud de los puntos que están cerca y menos en los puntos distantes. 

La similitud se mide como proximidad, utilizando la distancia entre pares de puntos.

Desafortunadamente, t-SNE no se escala bien y puede ser difícil de ajustar, ya que es sensible a sus hiperparámetros. 



### Uniform Manifold Approximation and Projection (UMAP)

La aproximación y proyección de variedades uniformes, o UMAP, también es un algoritmo de reducción de dimensionalidad no lineal, que se utiliza a menudo como una alternativa al t-SNE.

Construye una representación gráfica de alta dimensión de los datos basada en la teoría de variedades, que asume que los datos se encuentran en una variedad de dimensiones inferiores incrustada en un espacio de dimensiones superiores. 

A continuación, UMAP optimiza una estructura gráfica de baja dimensión que preserva mejor las relaciones entre los puntos de los datos originales. 

UMAP escala mejor que t-SNE y, además de la estructura más baja de los datos, preserva la estructura global y, a menudo, proporciona un rendimiento de agrupamiento más alto que t-SNE.




### Dimension Reduction scenario

Considere los gráficos 3D que muestran datos simulados utilizando la función MakeBlobs en Scikit-Learn. 

Los gráficos muestran los mismos datos 3D desde dos perspectivas diferentes. 

Hay un poco de superposición entre los cúmulos amarillos y morados, mientras que las otras dos manchas están claramente separadas de todas las manchas. 

Aplicemos los algoritmos de reducción de dimensiones PCA, t-SNE y UMAP para proyectar este conjunto de datos en dos dimensiones y comparar los resultados. 




### PCA result

La PCA ha separado las manchas de manera efectiva. Todas las manchas están distribuidas normalmente y las únicas diferencias entre ellas están en sus medias y varianzas.

Esto significa que las manchas están correlacionadas linealmente, por lo que se espera que el PCA funcione bien en los datos simulados.




### t-SNE result

t-SNE ha agrupado los datos en cuatro grupos y, en su mayoría, ha separado bien las manchas. 

El algoritmo identificó cuatro grupos muy distintos, con varios puntos mal etiquetados dentro del grupo morado de los grupos verde y amarillo. 

Se espera esta mezcla, ya que dos de los grupos tenían una ligera superposición. 



### UMAP result

Si bien UMAP no identificó perfectamente algunos de los puntos de mancha en las manchas, puede ver que tres de los clústeres no se separaron por completo. 

En particular, los racimos amarillo y verde tienen una ligera superposición con el racimo morado. 

Este debería ser el caso de los clústeres amarillo y morado, ya que estos dos clústeres no estaban completamente separados al principio en los datos de entrada 3D. 

La UMAP tuvo un desempeño ligeramente mejor que la t-SNE en este sentido, ya que la t-SNE identificó cuatro grupos muy distintos.
