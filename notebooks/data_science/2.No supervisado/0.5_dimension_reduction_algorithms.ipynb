Principal componets Analysis (PCA)

Es un algoritmo de reduccin de dimensionalidad lineal:
- Assume que las caractersticas del conjunto de datos estn correlacionadas linealmente.
- Simplifica los datos, reduce la dimensionalidad, reduce el ruidi y a la vez que minimiza la prdida de informacin
- Puede transformar caractersticas en un nuevo conjunto de variables no correlacionadas denominadas componentes principales y, al mismo tiempo, conservar la mayor varianza posible.





T-distributed stochastic neighbor embedding (t-SNE)

- Mapea puntos de alta dimensin a un espacio de menor dimensin
- Es bueno para encontrar clsteres en datos complejos y de alta dimensin que se pueden visualizar en dos o tres dimensiones y funciona bien con datos como imgenes y texto. 
- Se centra en preservar la similitud de los puntos que estn cerca y menos en los puntos distantes. 
- La similitud se mide como proximidad, utilizando la distancia entre pares de puntos. 
- Desafortunadamente, no escala bien y puede ser dificil de ajustar, ya que es sensible a sus hiperparmetros. 


Uniform Manifold Approximation and Projection (UMAP)

Es un algoritmo de reduccin de dimensionalidad no lineal, que se utiliza a menudo como una alternativa al T-SNE

- Construye una representacin grfica de alta dimensin de los datos basada en la teora de variedades, que asume que los datos se encuentran en una variedad de dimensiones inferiores incrustrada en un espacio de dimensiones superiores
- Optimiza una estructura grfica de baja dimensin que preserva mejor las relaciones en tre los puntos de los datos originales. 
- Escala mejor que t-SNE
- Preserva la estructura global y a menudo proporciona un rendimiento de agrupamiento ms alto que t-SNE.
