### HOW SIMPLE LINEAR REGRESSION WORKS

- Considerando el mismo conjunto de datos, vamos a trazar el tamao del **motor como una variable independiente y las emisiones de CO2 como el valor objetivo que queremos predecir.**

### REPRESENTING THE SIMPLE LINEAR REGRESSION MODEL

- Puede predecir el valor objetivo **(las emisiones de CO2)** , representado como la variable de respuesta, y-hat.

- La variable independiente, que en este caso es **el tamao del motor**, est representada por la variable predictora nica, x1.

- El Modelo se representa como la ecuacin de una recta, aqu. **y-hat** es la respuesta pronosticada expresada en trminos de **x1** usando una interseccin en **y, theta cero**, y una pendiente, **theta uno**.

- **Theta cero y theta** uno se denominan coeficientes del modelo de regresin lineal, seleccionados para determinar la lnea que mejor se ajusta.



### FINDING THE BEST FIT

- Dado que un automvil con un tamao de motor x1 equivale a 5.4, su emisin real de CO2 es de 250, mientras que su emisin prevista es y-hat es igual a 340.

- Al comparar el valor real con el previsto, hay una discrepancia de 90 unidades.

- El error residual es la distancia vertical desde el punto de datos hasta la lnea de regresin ajustada.

- El promedio de todos los errores residuales mide qu tan mal se ajusta la lnea de regresin a los datos.

- Matemticamente, podra mostrarse mediante la ecuacin, error cuadrtico medio, que se muestra como MSE.

**La regresin lineal tiene como objetivo encontrar la lnea para minimizar la media de todos estos errores residuales.**

- Esta forma de regresin se conoce comnmente como **regresin de mnimos cuadrados ordinarios o regresin OLS.**
