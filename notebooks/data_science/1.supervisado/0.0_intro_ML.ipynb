<ins>**1.1 AI, Machine Learning & deep learning.**<ins>

**Machine Learning**
- Uses algorithms
- Requires feature engineering

**Deep learning:**
- Uses multi-layered neural networks
- Automatically extracts features from big data.

<ins>**How machine learning works?**<ins>

*Teaches computer to:*
- Learn from data
- Identify patterns
- Make decisions

*Algorithms:*
- Use computational methods for learning
- Does not rely on a fixed algorithm. 


<ins>**Machine learning paradigms:**<ins>

1. Supervised learning models train on labeled data (Para aprender hacer inferencias a partir de datos nuevos, permite predecir etiquetas que de otro modo serían desconocidos)
2. Unsupervised learning works without labels (Encuentran patrones en los datos )
3. Semi-supervised learning works iteratively (Entrena con un subconjunto relativamente pequeño de datos que ya esta etiquetado y se reentrena iteractivamente añadiendo nuevas etiquetas que genera con una confianza razonablemente alta.)
4. Reinforcement learning learns from feedback.

<ins>**Machine learning Techniques:**<ins>

1. Classification predicts class or category of a case - Supervised
2. Regression/estimation predicts continuous values - Supervised
3. Clustering groups similar data (clientes similares) - Unsupervised  
4. Association finds items or events that co-occur (eventos que a menudo coexisten: Articulos de abarrotes que un cliente o segmento de mercado en particular suele comprar juntos)
5. Anomaly detection discovers abnormal and unusual cases (Fraudes con tarjeta de credito)
6. Sequence mining predicts next event (Analisis del flujo de clics en paginas web)
7. Dimension reduction reduce size of data 
8. Recommendation systems associate people´s preferences (Preferencia de personas con las otras personas que tienen gustos similares. Books, movies etc)

Un dia en la vida de un ingenierio en ML

---

1. Problem definition. "Beuty product customer, I would like to receive recommendation for other products based on my purchase history"

2. Data Collection.: User data, product data, other data (saved products, liked products, search history) transform to one central source. 

3. Data prepration: Preparing a final version of the data.
    _ Clean data
    _ Remove extreme values
    - Missing values should be removed or replaced
    - Each data column should be in the proper format
    - Create new features
    - Validate the data
    - Exploratory Data Analysis
    - Identify how to split up the data into training and test sets

4. Model development.
    - Define the technique (Content-based filtering)
    - Collaboritive filtering

5. Model evaluation
    - Evaluate model on never-before-seen data (asking feedback)

<ins>Types of machine learning tools<ins>

**1. Data proccesing and analytics:** Process, store, and interact with data. 

Example:
- PostgreSQL: Relational database management system
- Hadoop: Batch processing solution for big data
- Spark: Distributed data processing framework
- Apache Kafka: Distributed real-time streaming analytics
- Pandas: Data wrangling and preprocessing
- Numpy: Numerical computing tools

**2. Data visualization:** Undestand and visualize data structure

Example:
- Matplotlib: Customizable plots and interactive visualizations
- Seaborn: Attractive statistical graphics
- ggplot2: Building and adding elements in layers 
- Tableau: Interactive data visualization dashboards

**3. Machine learning ecosystem:** Create nad tune ML models

Example:
- Numpy: Numerical computation on data arrays
- Pandas: Data analysis, visualization, cleaning and preparation
- Scipy: Computing for optimizacion, integration, and linear regression
- Scikit-learn: Suite of classification, regression, clustering, and dimensionality reduction.

**4. Deep learning:** Desing, training and testing neural network-based models

Example:
- Tensorflow: Numerical computing and large-scale ML
- Keras: Implementing neural networks
- Theano: Defining, optimizing, and evaluating mathematical expressions involving arrays
- PyTorch: Computer vision, NLP, and experimentation

**5. Computer vision tools**

- Object detection
- Image classification
- Facial recognition
- Image segmentation

*Computer vission application*

Example:
- OpenCV: Real-time computer vision applications
- Scikit-image: Image processing algorihms
- TochVision: Popular data sets, image loading, pre-trained deep learning architectures, and image transformations.

**6. Natural language processing:** Help build applications that understand, interpret, and generate human language

Example: 
- NLTK: Text processing, tokenization, and stemming
- TexBlob: Part-of-speech tagging, noun phrase extraction, sentiment analysis, and translation
- Stanza: Pre-trained models for tasks such as part-of-speech tagging, named entity recognition, and dependency parsing. 

**7. Generative AI:** Leverage AI to generate new content based on input data

- Huggling face transformers: Text generation, language, translation, and sentiment analysis
- ChatGPT: Text generation, building chatbots, etc
- DALL-E: Generating images from text descriptions
- PyTorch: Uses deep learning to create generative models such as GANs and transformers
