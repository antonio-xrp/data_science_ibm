<ins>1. Linear regression<ins>
- Supervised learning model
- Models a relationship between a continuous target variable and explanatory features.

<ins>Types of regression models<ins>

- **Simple regression:** When a single independant variable estimates independent variable
    - Simplea linear regression
    - Simple nonlinear regression

- **Multiple regression:**  More a one independable variable is present, the process is called multiple regression. (depending on the relation between dependent and independent variables )
    - Multiple linear regression
    - Multiple nonlinear regression


<ins>Applications of regression<ins>

We use regression when we want to estimate a continuous value. 

- Sales forcasting
- Price estimation
- Predictive maintenance 
- Employment income 
- Rainfall estimation
- Wildfire probability and severity
- Spread of infectious disease 
- Risk of chronic disease


<ins>Regression algorithms<ins>
- Linear and polynomial
- Random forest
- Extreme Gradient Boosting (XGBoost)
- K-nearest neighbors (KNN)
- Support vector machines (SVM)
- Neural Networks



<ins>2. Multiple Linear Regression<ins>

- Extension of simple linear regression model
- Uses two or more independent variables to estimate a dependent variable
- Measures strenght of each independent variable's effect on a dependent variable

- Multiple linear regression better model than simple linear regression
- Too many variables can cause overfitting
- To improve prediction, convert categorical independent variables into numerical variables. 

<ins>Applications of multiple linear regression<ins>

- Used in education to predict outcomes and explain relationships
- Used to predict the impact of changes in "what-if" scenarios.

<ins>"What-if" scenario pitfalls<ins>

Inaccurate findings:
- Considering impossible scenarios
- Extrapolating scenarios
- Model might depend on a gruop of correlated or colinear variables.

<ins>Correlation pitfalls<ins>

- Correlated variable are no longer "independent variables"
- If variable correlated with another feature, "what-if" scenario not possible. 

<ins>Solution to correlation pitfalls<ins>
- Remove redundant variables
- Select variables which are:
    - Most understood
    - Controllable
    - Most  correlated with target

<ins>3. Introduction to Polynomial and Nonlinear Regression<ins>

- Statistical method for modeling relationship between dependent variable and independent variables. 
- Represented by a nonlinear equation
- Equation could be polynomial, exponential, logarithmic, or a nonlinear function
- Useful for a complex relationship
- Example: Data set with exponential growth pattern

 

 - Data has a background trend that follows a smoothed curve
 - A smooth, nonlinear curve does a better job at approximating data
 - The straight line "under-fits" the data 

<ins>3. Introduction to Polynomial and Nonlinear Regression<ins>

- Polynomial regression uses an ordinary linear regression to indirectly fit data to polynomial expressions 
- Nonlinear regression bases inputs on functions of given features 

 <ins>Polynomial regression<ins>



- Relationship between independent variable X and the dependent variable y is modelled as an nth degree polynomial in X 



<ins>Overfitting with polynomials<ins>

- Polynomial regression model "memorizes" everything including noise or variations 
- Pick a regression that fits data without overfitting 



<ins>Applications of nonlinear regression<ins>

Polynomial regression: 
- Nonlinear dependence on input features but linear dependence on regression coefficients
- Can be transformed into a linear regression problem
