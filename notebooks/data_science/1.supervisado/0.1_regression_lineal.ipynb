- Puede predecir el valor objetivo **(las emisiones de CO2)** , representado como la variable de respuesta, _y-hat_.

- Al comparar el valor real con el previsto, hay una discrepancia de 90 unidades. 

- El error residual es la distancia vertical desde el punto de datos hasta la lnea de regresin ajustada. 

- El promedio de todos los errores residuales mide qu tan mal se ajusta la lnea de regresin a los datos. 

- Matemticamente, podra mostrarse mediante la ecuacin, error cuadrtico medio, que se muestra como **MSE**. 

**La regresin lineal tiene como objetivo encontrar la lnea para minimizar la media de todos estos errores residuales.**

<ins>2. Multiple Linear Regression<ins>

- Extension of simple linear regression model
- Uses two or more independent variables to estimate a dependent variable
- Measures strenght of each independent variable's effect on a dependent variable

- Multiple linear regression better model than simple linear regression
- Too many variables can cause overfitting
- To improve prediction, convert categorical independent variables into numerical variables. 

<ins>Applications of multiple linear regression<ins>

- Used in education to predict outcomes and explain relationships
- Used to predict the impact of changes in "what-if" scenarios.

<ins>"What-if" scenario pitfalls<ins>

Inaccurate findings:
- Considering impossible scenarios
- Extrapolating scenarios
- Model might depend on a gruop of correlated or colinear variables.

<ins>Correlation pitfalls<ins>

- Correlated variable are no longer "independent variables"
- If variable correlated with another feature, "what-if" scenario not possible. 

<ins>Solution to correlation pitfalls<ins>
- Remove redundant variables
- Select variables which are:
    - Most understood
    - Controllable
    - Most  correlated with target

|La regresin multiple tiene como tiene como objetivo minimizar la ecuacion MSE
