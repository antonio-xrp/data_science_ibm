Clasification 

Metodo supervisado que se usa para predecir etiquetas categoricas de datos nuevos

Application

Muchos problemas se pueden expresar como asociaciones entre varibles de caracteristicas y de destino. 

- Email filtering
- Speech-to-text
- Handwriting recognition
- Biometric identification
- churn prediction
- customer segmentation
- advertising 
- loan default (cliente incurra en el incumplimento de su deuda)
- multiclass drug prescription  

Classification algorithms

- Naive bayes
- Logistic regression
- K-nearest neighbors
- Support vector machines
- Decision Trees
- Neural netweok

| Algoritmoss como regresin logstica, KNN y los rboles de decision pueden aprender a distinguir varias classes. 

Estrategia para extender los clasificadores binarios para manejo de multiplicas clases son:

- One vs all (Is it this? independent binary classifiers)= Interactua una etiqueta contra el resto de etiquetas. Si existe K clases, contribuiran exactamente a k clasificadores binarios. (Es posible que un punto no pertenezca a una etiqueta (noise o outliers))




- One vs one (is it this or is it that?): Considera todo los pares de conjunto posible y se entra un clasificador en el subconjunot de datos y decide a que clase pertenece cada punto.( se decide por sistema de votacion, en caso de empate usar one vs all o sistema mejorado)
