### One-versus-all strategy

- El esquema de uno contra todos implementa un conjunto de clasificadores binarios independientes, uno para cada etiqueta de clase del conjunto de datos.

- A cada clasificador se le asigna una nica etiqueta que define su clase objetivo. 

- La tarea de cada clasificador es hacer una prediccin binaria para cada punto de datos sobre si tiene la etiqueta dada, un clasificador de uno contra el resto. 

- Como puede ver, si hay k clases, contribuirn exactamente k clasificadores binarios.



Comprendamos cmo la estrategia de uno contra todos descompone un conjunto de puntos de datos.

El algoritmo funciona en cada etiqueta de clase, una a la vez, con un resultado entrenado que muestra los puntos que se prev que tengan un color determinado. 

Tenga en cuenta tambin que es posible que un punto de datos determinado no pertenezca a ninguna de las clases, ya que es posible que ninguno de los clasificadores individuales lo recoja. 

Estos puntos no clasificados pertenecen a otra clase. 

Esta propiedad puede resultar til para identificar valores atpicos o ruidos. 



### One-versus-one strategy


- Con la estrategia uno contra uno, en lugar de que cada clasificador decida si un punto pertenece a una clase, la pregunta cambia de Es esto? a Es esto o aquello? 

- Dado un conjunto de clases, considere todos los pares de clases posibles.

- Para cada par de etiquetas, se entrena un clasificador en el subconjunto de datos correspondiente a las dos etiquetas y decide a qu clase pertenece cada punto. 

- El proceso contina hasta que todos los clasificadores estn entrenados.



La etiqueta de clase final asignada a cada punto puede decidirse mediante un sistema de votacin.
