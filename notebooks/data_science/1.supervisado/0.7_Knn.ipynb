KNN

- Algoritmo supervisado, que toma un un grupo de puntos de datos etiquetados y luego los usa para aprender a etiquetar otrs puntos de datos.
- Knn se usa tanto para regresin y clasificacin
- A los puntos cercanos entre si se les conoce como vecinos





- En regresion, la prediccin se realizar utilizando el promedio o la mediana de los valores objetivos. 

Finding the optimal K

- Probar un rango de valores con un conjuto de datos de prueba etiquetado y medir la precisin.
- Una ves completado, elige K = 1, usa la parte de entrenamiento para modelar y calcula la precisin.
- Repite este proceso aumentando la K y comprube que K es mejor para tu modelo.
