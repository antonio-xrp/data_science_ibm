Suport Vector Machine (SVM)

- SVM tecnica de aprendizaje supervisado para crear modelos de clasificacin y regresin.

- SVM clasifica los datos de entrada identificando el hiperplano (linea que segrega y clasifica el conjunto de datos) lo que diferencia claramente dos clases. 

- Cuando mayor sea el margen, mejor ser la precision del modelo en datos nuevos e invisibles. 

- El equilibro entre maximizar el margen y minimizar el numero de clasificaciones se controla mediante un parmetro C, una C ms pequea permite mas clasificaciones errneas, mientras una C ms grande obliga a una separacion mas estrcta. 







Scikit-learn function for SVM

1. Linear: Nucleo predeterminado y corresponde al modelo SVM habitual
2. Polynomial: funcin parablica
3. Funcin Base Radial (RBF): Obtiene una puntuacin alta para puntos cercanos entre s y una puntuacin que disminuye potencialmente a medida que los puntos se alejan. 
4. Sigmoid: La misma funcin para regresin logstica. 



Epsilon: es un parmetro de algoritmo de aprendiza SVR que puede seleccionar para definir un margen alrededor de la curva de prediccin. 




SVM pros y contras

Pros:
- Eficaz en espacios de alta dimension
- Resistente al sobreajuste(overfitting)
- Destaca en datos separables lineales
- Funcion con datos dbilmente separables mediante la opcin dbil. 

Limitaciones

- Lento para el entrenamiento en grandes conjuntos de datos
- Sencible al ruido y a la superposicin de clases
- Es sencible a la eleccin de los parmetros del ncleo de regularizacin (kernel)

Cuando usar SVM? 

- SVM es buena para tareas de anlisis de imagen, como clasificacin de imgenes y el reconomiento de dgitos escritos a mano.

- Tambin es bueno para el parsing, deteccin de span y anlisis de opiniones Como reconocimiento de voz, deteccin de anomalas y noise filtering.
