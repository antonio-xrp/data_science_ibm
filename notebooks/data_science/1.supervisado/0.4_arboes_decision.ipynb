### Training a decision tree

Un rbol de decisiones se entrena creciendo de la siguiente manera.

- Comienza con un nodo semilla y datos de entrenamiento etiquetados.

- Entrena el nodo en sus datos asignados encontrando la caracterstica que mejor divide los datos en sus clases preetiquetadas, segn un criterio de divisin preseleccionado. 

- Cada divisin divide los datos de entrada del nodo, y cada particin se pasa a lo largo de su rama a un nuevo nodo.

- Repite el proceso para cada nuevo nodo, usando cada caracterstica slo una vez. 



El rbol crece hasta que todos los nodos contienen una sola clase cada uno, o se agotan las caractersticas para seleccionar, o se cumple un criterio de parada preseleccionado.

Alternativamente, tambin puede detener el crecimiento de un rbol cortando las ramas que no mejoren significativamente el rendimiento del sistema.

### Why prune?

Hay varias razones por las que podra querer podar un rbol de decisiones.

- Si el rbol es demasiado complejo, podra estar ajustndolo en exceso a los datos de entrenamiento. 

- Si tiene demasiadas clases y caractersticas, el rbol podra estar capturando ruido y detalles irrelevantes.

- La poda simplifica el modelo de rbol de decisin y lo hace susceptible de generalizacin.

- Un rbol podado es ms conciso y fcil de entender. 

- La poda tambin mejora la precisin predictiva.

### What is entropy?

La entropa es la medida del desorden de la informacin, o aleatoriedad en un conjunto de datos. Mide lo aleatorias que son las clases de un nodo, o lo incierto que es el resultado de la divisin de caractersticas. 

En rboles de decisiones, se buscan rboles que tengan la entropa ms pequea en sus nodos.

Puede calcularse la entropa de un nodo utilizando la frmula de entropa, donde pA y pB son, respectivamente.

Las proporciones de pacientes del frmaco A y del frmaco B en el nodo, si las clases son completamente homogneas, la entropa es 0, y si estn igualmente divididas, la entropa es 1.

Por ejemplo, si pA es igual a pB es igual a 1 mitad, entonces la frmula da 1 mitad por 1 negativo, menos 1 mitad por 1 negativo es igual a 1.

### What is information gain?

La ganancia de informacin es la entropa de un rbol antes de la divisin menos la entropa ponderada despus de la divisin por una caracterstica. 

Por ejemplo, utilizando el colesterol como caracterstica para dividir a todos los pacientes se obtiene una ganancia de informacin de 0,042. 

Puede considerar la ganancia de informacin y la entropa como opuestas. 

A medida que disminuye la entropa, aumenta la ganancia de informacin, o la cantidad de certeza. 

Construir un rbol de decisiones consiste en encontrar caractersticas que proporcionen la mayor ganancia de informacin.
