Un arbol de decision deja de crecer cuando se cumple un criterio de parada. Por ejemplo, se puede establecer creterios de parada para su modelo cuando se cumpla el siguiente critero:

1. Se alcanza la profundidad mxima de arbol
2. Se ha superado el numero minimo de puntos de datos en un nudo. 
3. Se ha superado el numero minimo de muestras de una hoja
4. Ha alcanzado el nmero mximo de nodos de hoja.

El rbol de decisin selecciona una caracteristica que devida mejor los datos, para ello selecciona un criterio de devisin:

- Ganancia de informacion (information gain) Entropia

Entropia
- Mide el desorde de informacin
- Mide que tan random las clases estan en un nodo
- Si las clases son completamente homogeneo, la entropia es igual 0
- Si las clases estan divididas igualmente, la entropia es 1

Information gain.
- Es la entropia de un arbol antes de la division menos la entropia ponderada despues de la division por una caracteristica. 
- Se puede considerar como puesta a la entropia
- a medida que disminuye la entropia aumenta la ganancia de informacion
