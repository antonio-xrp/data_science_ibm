Introduction to Logistic Regression

---

<ins>What is logistic regression?<ins>

- Is a statistical modeling technique that predicts the probability of an observation belonging to one of two classes such True or False
- In Machine Learning (ML), it refers to a binary classifier based on statistical logistic regression 



<ins>When is logistic regression a good choice?<ins>

1. When the target in data is binary for example yes/no, True/false
2. When the probability of an outcome is needed: Logistic regression returns a probability score of a sample and maps it to the appropriate class
3. If the data is linearly separable, the decision boundary of logistic regression is a line, a plane, or a hyperplane
4. To understand the impact of an independent feature for example: Select features based on model coefficient size of weights



<ins>Logistic regression applications<ins>

1. You can use it to predict the probability that a person will have a heart attack risk
2. Diagnosing patients based on a set of characteristcs 
3. Predicting whether a customer will purchase a product halt a subscription 
4. Predicting product failure probability 
5. Predicting mortgage default likelihood

Training a logistic regression model

---

<ins>Recap<ins>
1. Logistical regression training aims to predict classes with minimal error
2. Training proccess seeks to find parameters/theta that minimize cost function
3. Optimization step used to find best parameters
4. Log-loss: Cost function which needs to be minimized
5. Log-loss favors confident, correct classifications and penalize confident, incorrect predictions
6. Gradient descent: Iterative approach to find the minimum of a function
7. SGD: Scalable variation of gradient descent, which uses a random data subset


<ins>Logistic regression training<ins>
- Identify parameters the map input to target outcomes
- Objective: Predict classes with minimal error
- Find paramets/theta that minimizes cost function

Steps:

1. Choose a set of parameters or theta
2. Predict probability that class = 1
3. Calculate prediction error (cost function)
4. Update theta to reduce prediction error
5. Repeat unti:
    - Reach small log-loss value or
    - Targated number of iterations

<ins>Minimizing cost function with gradient descent<ins>

How?

- Stop training when log-loss is satisfactory
- Use a gradient descent 

What is gradient descent?
- Iterative approach to finding the minimum of a function
- Adjusts parameters values using log-loss derivative
- Depends on a specified learning rate
- Controls how far it's allowed to step the parameters 

<ins>Gradient descent on a surface<ins>

Goal: Change parameter values and find path to optimal parameters to minimize the const function. 



<ins>Gradient descent<ins>
- Calculated over the entire data set
- Large data set = Slow descent
- Converge less likely as steps too big to notice minima
- Gradient can be approximated using a random subset 

<ins>Stochastic gradient descent (SGD)<ins>
- Variation of the gradient descent, it's faster, but can be less accurate
- Uses a random data subset and scales well
- Likely to overlook local and find global minima 
- Converge quickly toward a global minimum 

Convergence can be improved by
- Decreasing learning rate
- Gradually increasing sample size
