### Validación cruzada y técnicas avanzadas de validación de modelos.


​Bienvenido a las técnicas de validación cruzada y validación avanzada de modelos. ​Después de ver este vídeo, podrá definir la validación del modelo. ​También podrás explicar qué es el espionaje de datos y cómo evitarlo. ​Por último, podrá analizar las estrategias clave para la validación del modelo. 

### Model Validation 

​La validación del modelo consiste en hacer todo lo posible para optimizar el modelo sin poner en peligro ​su capacidad de predecir correctamente con datos invisibles. 

​Le ayuda a evitar el sobreajuste al seleccionar la mejor configuración de modelo ajustando los hiperparámetros. 



### What's wroing with this picture?

​Considere el método básico de evaluación por trenes o pruebas divididas, en el que el conjunto de datos se divide en dos partes. 

​Se usa un conjunto de entrenamiento para entrenar el modelo y un conjunto de prueba para evaluar o estimar ​la capacidad del modelo para predecir los resultados a partir de datos invisibles.

​La mayoría de los modelos de aprendizaje automático tienen configuraciones de parámetros opcionales, denominadas hiperparámetros del modelo, ​que afectan a la forma en que el modelo se ajusta a los datos utilizados para entrenarlo. 

​¿Qué pasa si prueba diferentes hiperparámetros para su modelo ​y, a continuación, elige el que mejor funciona en los datos de prueba? ​¿No ajustaría eficazmente el modelo a los datos de prueba, no a los datos de entrenamiento? ​Esto provocaría un sobreajuste. ​Es probable que su modelo no se generalice bien a datos invisibles, lo que los invalidaría. ​Comprobar el rendimiento de los datos de prueba antes de terminar de optimizar el modelo se denomina ​espionaje de datos, una forma de lo que se conoce como filtración de datos. 



### Avoiding data snooping 

​¿Qué puede hacer para validar su modelo y asegurarse de que no se ajusta demasiado a sus datos de prueba? ​Es necesario desvincular el ajuste del modelo de la evaluación final. La ​validación significa ajustar el modelo con los datos de entrenamiento, pero solo probarlo con ​datos de prueba invisibles una vez que esté seguro de que está bien entrenado. ​No se trata de husmear. 



### Validation sets

​Esta es una estrategia de validación de modelos que implica segmentar los datos en al menos tres partes. 

​Un conjunto de entrenamiento, que se utiliza para entrenar el modelo, incluida la optimización de sus hiperparámetros. 

​Uno o más conjuntos o subconjuntos de validación de los datos de entrenamiento utilizados durante el ​proceso de optimización del modelo para evaluar el rendimiento de un modelo de aprendizaje automático. 

​Un conjunto de pruebas que se almacena, los datos invisibles se utilizan para la evaluación final ​después del entrenamiento y la validación del modelo. 



### Cross-validation algorithm

​La validación cruzada permite el ajuste de hiperparámetros. ​Este es el algoritmo de validación cruzada para el ajuste y la validación del modelo.

​Divida sus datos en datos de entrenamiento y datos de pruebas.

​Divide aún más tus datos de entrenamiento en un conjunto de entrenamiento y un conjunto de validación. ​

Optimice los hiperparámetros de su modelo entrenándolo repetidamente en el conjunto de entrenamiento ​y midiendo su rendimiento en el conjunto de validación. ​
  
Elija su mejor conjunto de hiperparámetros ​y evalúe el mejor modelo resultante a partir de sus datos de prueba completamente invisibles. 

​Ahora tiene un modelo validado y una estimación de qué tan bien se generalizará en datos nuevos e invisibles. 




### Is There Still a problem?

​Algunos posibles problemas de validación radican en la selección de un conjunto de validación único y específico. ​

Su modelo podría estar sobreajustado a este conjunto de datos específico.

​Si su modelo necesita muchos datos con los que entrenar, es posible que no le queden suficientes datos ​para fines de validación y prueba.

​Esto significa que es posible que sus datos de entrenamiento y validación no sean representativos de la población de la muestra. ​

Es posible que su modelo no esté aprendiendo detalles, como el ruido en este conjunto en particular. 

​Es posible que el rendimiento de su modelo no sea estable en los diferentes conjuntos de validación. 



### A solution: K-fold cross-validation

​Una solución para evitar el sobreajuste de los datos de prueba al intentar optimizar los hiperparámetros del modelo ​consiste en unos pocos pasos clave.

​Divida los datos en K pliegues de igual tamaño para usarlos como subconjuntos de validación. ​

Para cada modelo de prueba o conjunto de hiperparámetros y para cada pliegue, ​entrene un modelo en los K menos 1 pliegues restantes. ​Pruebe el modelo en el pliegue seleccionado y guarde la puntuación de este modelo. ​Calcula una puntuación agregada de los pliegues totales. ​
Seleccione el conjunto de hiperparámetros que condujeron al mejor modelo. 

​Tenga en cuenta que todos los puntos de datos se utilizan tanto para el entrenamiento como para la validación, lo que ​aumenta considerablemente la utilización de los datos que tiene a mano. 

​La validación cruzada por K, normalmente de 5 a 10 veces, proporciona una técnica más sólida para estimar la ​generalización del modelo a datos invisibles del mundo real. 




### K-fold cross-validation advantages

​Variar el conjunto de validación tiene varias ventajas. 

​Aumenta considerablemente los datos con los que se entrena y prueba el modelo.

​Reduce el sobreajuste porque suaviza los detalles no deseados que son específicos de ​un subconjunto de entrenamiento elegido. 

​En consecuencia, mejora su capacidad para evaluar qué tan bien se generalizará su modelo a datos invisibles. 



### Handling imbalanced data

​En los problemas de clasificación, es posible que tenga muchas observaciones en una clase ​y muy pocas en otra. ​

Significa que se trata de un problema de clasificación desequilibrado. 

La ​validación cruzada estratificada garantiza que la distribución de clases se conserve en cada pliegue de validación, lo que ​evita sesgos en el proceso de evaluación.
 
​La analogía con los datos desequilibrados en los problemas de regresión ocurre cuando el objetivo está muy sesgado. ​Muchos modelos asumen que su objetivo está distribuido normalmente. ​

Afortunadamente, puede transformar la variable objetivo mediante métodos como las transformaciones logarítmicas o box-cox ​para reducir la asimetría y ajustar el modelo al objetivo transformado.
