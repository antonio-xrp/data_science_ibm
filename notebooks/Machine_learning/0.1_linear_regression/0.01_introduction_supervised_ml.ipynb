### What is Machine Learning?

> Machine Learning permite a las computadoras aprender de los datos. 

- Con los modelos estadsticos tradicionales, a menudo sabemos bastante sobre el proceso subyacente. Por lo tanto, podemos elegir un modelo de anlisis que se aproxime mucho a su proceso subyacente y usarlo para desarrollar un modelo predictivo.

- Machine Learning, por otro lado, a menudo sabemos menos sobre el proceso subyacente o simplemente es demasiado complicado de expresar. En cambio, aprendemos o aproximamos la ecuacin a partir de los datos. La aproximacin de funciones permite predecir valores futuros mediante el modelo de aprendizaje automtico.

### Model: A Learning Algorithm

Qu hay de un modelo? Qu entendemos por modelo? 

- Un modelo es una cosa pequea que captura una cosa ms grande.

- Un buen modelo omitir detalles sin importancia y conservar lo que es importante. 

As que aqu tenemos un mapa que va a ser un modelo que representa una realidad mucho ms amplia. Al elegir la representacin o el modelo correctos, siempre reducimos la cantidad de informacin incluida, pero debemos hacerlo de manera que se conserven las caractersticas o relaciones que nos interesan. 

- Y un buen modelo es capaz de captar las relaciones importantes y, al mismo tiempo, reducir la complejidad de un fenmeno del mundo real lo suficiente como para que podamos asimilarlo y comprender y representar ese mundo real.
