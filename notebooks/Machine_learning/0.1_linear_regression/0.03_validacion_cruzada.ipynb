### Validación Cruzada K-Fold

### Introducción a la validación cruzada y la validación cruzada K-Fold

La validación cruzada es una técnica estadística empleada en el aprendizaje automático para evaluar el rendimiento y la generalizabilidad de un modelo predictivo. El objetivo principal de la validación cruzada es dividir un conjunto de datos en subconjuntos, entrenar el modelo en algunos de estos subconjuntos y evaluar su rendimiento en el subconjunto restante. Repetimos este proceso múltiples veces y promediamos las métricas de rendimiento del modelo a lo largo de estas iteraciones.

La validación cruzada ofrece varias ventajas, incluyendo una evaluación de rendimiento más confiable al reducir el impacto de la variabilidad en una única división aleatoria de datos. También optimiza la utilización de los datos disponibles, asegurando que cada punto de datos se use para la validación exactamente una vez. La forma más prevalente de validación cruzada es la validación cruzada k-fold.

En la validación cruzada k-fold, el conjunto de datos se divide aleatoriamente en k pliegues o subconjuntos de igual tamaño.

Posteriormente, el modelo se entrena en k-1 pliegues y se valida en el pliegue restante. Este ciclo se repite k veces, con cada iteración empleando un pliegue único como el conjunto de prueba. La métrica de rendimiento final es típicamente el promedio de las métricas obtenidas en cada iteración.

Por ejemplo, si eliges k = 5, creas cinco pliegues. En este enfoque, entrenas tu modelo 5 veces. En cada iteración, un pliegue diferente es el conjunto de validación, y entrenas tu modelo en los otros cuatro pliegues.
En consecuencia, cada punto de datos puede ser parte del conjunto de validación exactamente una vez.

### Tipos de validación cruzada K-Fold

### Validación cruzada K-Fold estándar

En la validación cruzada K-Fold tradicional, el conjunto de datos se divide aleatoriamente en K pliegues de tamaño aproximadamente igual. En cada iteración, un pliegue es el conjunto de validación, mientras que los K - 1 pliegues restantes constituyen el conjunto de entrenamiento. Repetimos este procedimiento K veces, asegurando que cada pliegue se utilice exactamente una vez como conjunto de validación.

### Validación cruzada K-Fold estratificada

La validación cruzada K-Fold estratificada tiene como objetivo mantener una distribución consistente de clases en cada pliegue, alineándose con la proporción general observada en el conjunto de datos. Este proceso resulta particularmente valioso para conjuntos de datos con desequilibrios donde ciertas clases son escasas. El objetivo es garantizar una evaluación del modelo sin sesgos al representar con precisión la distribución de clases en cada pliegue.

Aunque la selección de datos no es completamente aleatoria, hay un elemento de aleatoriedad en la elección de muestras dentro de cada clase.

Por ejemplo, considera un escenario de clasificación binaria con un conjunto de datos desequilibrado, donde el 90% de las muestras pertenecen a Clase A y el 10% a Clase B. En la validación cruzada K-Fold estratificada de 5 pliegues, cada pliegue incluirá una representación proporcional de muestras de Clase A y Clase B, preservando consistentemente la proporción de desequilibrio a través de los pliegues.

### Validación cruzada K-Fold por grupos

La validación cruzada K-Fold por grupos se emplea al trabajar con conjuntos de datos donde las muestras exhiben interdependencias, como datos de series temporales o datos con correlaciones espaciales. Este método garantiza que las muestras del mismo grupo estén completamente dentro del conjunto de entrenamiento o del conjunto de validación, evitando la filtración de datos entre pliegues.

A diferencia de ciertas otras técnicas de validación cruzada, no dividimos los datos aleatoriamente. En su lugar, utilizamos la información de agrupamiento para construir pliegues que preserven la cohesión de los grupos.

Como ilustración, considera un conjunto de datos que comprende múltiples secuencias de series temporales, cada una representando datos recopilados a lo largo del tiempo de un individuo. Con la validación cruzada K-Fold por grupos, las secuencias (grupos) permanecen intactas a través de los pliegues, asegurando que el modelo evite aprender de datos futuros al hacer predicciones.

### Comprendiendo la complejidad del modelo en la regresión lineal a través de la validación cruzada K-Fold

Según la complejidad, podemos categorizar los modelos de aprendizaje automático de la siguiente manera:

- Modelos de baja complejidad
- Modelos de alta complejidad

**Modelos de baja complejidad:** Estos son modelos relativamente simples entrenados y construidos con un número limitado de características. Son fáciles de entender e interpretar y funcionan bien con conjuntos de datos pequeños. Estos modelos no pueden capturar relaciones complejas y no son flexibles para adaptarse a patrones de datos diversos. Son propensos al subajuste, lo que significa que no proporcionan buena precisión en los datos de entrenamiento y prueba.

**Modelos de alta complejidad:** Estos son modelos sofisticados que se construyen y entrenan con más características. Son más flexibles al adaptarse a patrones de datos diversos. Esta flexibilidad también puede causar sobreajuste, donde el modelo puede memorizar ruido o patrones específicos que no se generalizan bien al probar los datos.

La validación cruzada K-fold es una técnica poderosa utilizada para evaluar el rendimiento de los modelos de aprendizaje automático. Al dividir el conjunto de datos en K subconjuntos o pliegues y rotarlos sistemáticamente como conjuntos de entrenamiento y validación, la validación cruzada K-fold ofrece una evaluación más resistente del rendimiento del modelo.

Aquí hay algunas observaciones generales:

**k más pequeño (k = 2 o 3)**

Un k más pequeño (k = 2 o 3) resulta en una mayor varianza entre los pliegues porque cada iteración entrena en una porción más pequeña del conjunto de datos. Dado que los datos de entrenamiento son más limitados en cada iteración, el rendimiento del modelo fluctúa más dependiendo del subconjunto específico de datos utilizado para el entrenamiento y la validación. Esta mayor varianza puede llevar a estimaciones del modelo menos estables y a un mayor riesgo de sobreajuste a conjuntos de entrenamiento más pequeños.

**k más grande (k = 10 o 20)**

Un k más grande resulta en una menor varianza entre los pliegues porque cada iteración entrena en una porción más grande del conjunto de datos. Dado que el conjunto de entrenamiento es casi el conjunto completo de datos en cada iteración, el rendimiento del modelo es más estable entre diferentes pliegues. Esto ayuda a mejorar la generalización y reduce el sobreajuste, ya que el modelo aprende de un conjunto de datos más diverso en cada iteración.

Una menor varianza significa que el modelo se desempeña de manera más consistente entre los pliegues, con fluctuaciones mínimas en métricas de rendimiento como el error cuadrático medio. Sin embargo, el costo computacional aumenta a medida que k crece, ya que el modelo debe ser entrenado y evaluado k veces.

Ejemplo: Consideremos un escenario del mundo real donde construimos un modelo de regresión para predecir los precios de las casas basado en factores como la superficie en pies cuadrados, el número de dormitorios y el vecindario. Al usar k = 10, el modelo se entrena en el 90% de los datos y se prueba en el 10% en cada iteración. Dado que cada conjunto de entrenamiento es grande y diverso, el modelo se generaliza bien a datos no vistos.

- **Pliegue 1:** Entrenamos el modelo en un subconjunto donde las casas con gran superficie en pies cuadrados y muchos dormitorios son predominantes. El modelo aprendido puede priorizar estas características como cruciales para predecir los precios de las casas.

- **Pliegue 2:** Luego entrenamos el modelo en un subconjunto diferente donde las casas en un vecindario diverso, independientemente de la superficie en pies cuadrados, tienden a tener precios más altos. En este caso, el modelo aprendido asigna más importancia al vecindario que en el Pliegue 1.

Por lo tanto, estos dos subconjuntos demuestran una variación considerable.
